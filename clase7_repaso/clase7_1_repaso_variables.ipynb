## Enunciado del Ejercicio 2: Cálculo de Promedio de Notas
Los ayudantes deben calcular el promedio de dos pruebas de Programación. Escribe un algoritmo en Python que solicite las dos notas, calcule el promedio y lo despliegue por pantalla con instrucciones claras al usuario.

### Análisis y Estrategia de Resolución:
1. **Tipos de Datos:** Las notas no siempre son enteros (ej. un 6.5). Por ello, debemos convertir la entrada de `input()` (que por defecto es texto) a `float` (número decimal).
2. **Matemática en Programación:** Para calcular el promedio $(n1 + n2) / 2$, es fundamental usar **paréntesis**. Sin ellos, Python dividiría solo la segunda nota por 2 debido a la prioridad matemática de los operadores.
3. **Estructura:** Pedir datos -> Calcular -> Mostrar.

### Resolución Paso a Paso:

```python
# Bloque 1: Ingreso de datos con conversión a float
nota1 = float(input("Ingresa la nota de la Prueba 1: "))
nota2 = float(input("Ingresa la nota de la Prueba 2: "))
```

```python
# Bloque 2: Cálculo aritmético
# El uso de paréntesis asegura que la suma ocurra antes que la división
promedio = (nota1 + nota2) / 2
```

```python
# Bloque 3: Salida de resultados
print("El promedio final de las pruebas es:", promedio)
```

### Código Completo
A continuación, el algoritmo en una sola celda para su ejecución:

```python
nota1 = float(input("Ingresa la nota de la Prueba 1: "))
nota2 = float(input("Ingresa la nota de la Prueba 2: "))

promedio = (nota1 + nota2) / 2

print("El promedio final de las pruebas es:", promedio)
```

## Enunciado del Ejercicio 3: División Entera y Resto (Módulo)
El profesor de matemáticas te ha enseñado que cuando se divide un número entero, el resultado de la división tiene una parte entera y un resto. Escribe el código en Python de un algoritmo que solicite por pantalla el valor del número entero A, y luego el del número entero B. Despliega por pantalla la parte entera de A/B y el resto de A/B.

### Análisis y Estrategia de Resolución:
1. **Requerimiento de Enteros:** Usaremos `int(input())` para asegurar que trabajamos estrictamente con números sin decimales.
2. **Operador de Cociente (`//`):** En Python, la doble barra inclinada devuelve la parte entera de una división (cuántas veces cabe el divisor en el dividendo de forma exacta).
3. **Operador de Módulo (`%`):** El signo de porcentaje devuelve el "sobrante" o residuo matemático de la división.

### Resolución Paso a Paso:

```python
# Bloque 1: Captura de números enteros
a = int(input("Ingresa el valor del número entero A: "))
b = int(input("Ingresa el valor del número entero B: "))
```

```python
# Bloque 2: Cálculo de parte entera y resto
parte_entera = a // b
resto = a % b
```

```python
# Bloque 3: Despliegue con formato
print("La división es", parte_entera, ", y el resto es", resto)
```

### Código Completo
A continuación, el algoritmo en una sola celda para su ejecución:

```python
a = int(input("Ingresa el valor del número entero A: "))
b = int(input("Ingresa el valor del número entero B: "))

parte_entera = a // b
resto = a % b

print("La división es", parte_entera, ", y el resto es", resto)
```

## Enunciado del Ejercicio 4: Distancia entre dos puntos
Escribe el código en Python del algoritmo que solicita al usuario las coordenadas de dos puntos en el plano cartesiano, $(x_1,y_1)$ y $(x_2,y_2)$, calcula la distancia lineal entre ambos puntos y despliega el resultado por pantalla.

### Análisis y Estrategia de Resolución:
1. **Fórmula Geométrica:** Aplicaremos la distancia euclidiana: $d = \sqrt{(x2-x1)^2 + (y2-y1)^2}$.
2. **Potenciación en Python:** El operador para elevar a una potencia es `**`.
3. **Raíz Cuadrada:** Matemáticamente, elevar una expresión a $0.5$ (o $1/2$) es exactamente lo mismo que calcular su raíz cuadrada.

### Resolución Paso a Paso:

```python
# Bloque 1: Entrada de coordenadas (4 variables independientes)
x1 = float(input("Coordenada x1: "))
y1 = float(input("Coordenada y1: "))
x2 = float(input("Coordenada x2: "))
y2 = float(input("Coordenada y2: "))
```

```python
# Bloque 2: Aplicación de la fórmula geométrica
# Calculamos la suma de los cuadrados de las diferencias y sacamos la raíz (**0.5)
distancia = ((x2 - x1)**2 + (y2 - y1)**2)**0.5
```

```python
# Bloque 3: Despliegue del resultado final
print("La distancia lineal entre los puntos es:", distancia)
```

### Código Completo
A continuación, el algoritmo en una sola celda para su ejecución:

```python
x1 = float(input("Coordenada x1: "))
y1 = float(input("Coordenada y1: "))
x2 = float(input("Coordenada x2: "))
y2 = float(input("Coordenada y2: "))

distancia = ((x2 - x1)**2 + (y2 - y1)**2)**0.5

print("La distancia lineal entre los puntos es:", distancia)
```

## Enunciado del Ejercicio 5: Contador de Días (Año Cero)
Escribe el código en Python del algoritmo que solicita al usuario una fecha (solicite día, mes y año) y muestre por pantalla la cantidad de días que han pasado desde el año cero a esa fecha. Asuma que cada año tiene 365 días y cada mes tiene 30 días.

### Análisis y Estrategia de Resolución:
1. **Lógica de Acumulación:** Para calcular los días totales, debemos multiplicar los años completos por 365, los meses por 30, y sumar los días sueltos ingresados.
2. **Consideración lógica:** Generalmente se asume que el año o mes ingresado está "en curso", por lo que la fórmula que usaremos será estrictamente acumulativa sobre lo ingresado.
3. **Entradas:** Necesitamos 3 capturas numéricas de tipo entero.

### Resolución Paso a Paso:

```python
# Bloque 1: Captura de fecha separada por partes
dia = int(input("Ingrese el día: "))
mes = int(input("Ingrese el mes: "))
anio = int(input("Ingrese el año: "))
```

```python
# Bloque 2: Conversión a la unidad mínima (días)
# Multiplicamos cada unidad de tiempo por su equivalente en días según el enunciado
total_dias = (anio * 365) + (mes * 30) + dia
```

```python
# Bloque 3: Resultado al usuario
print("Han pasado", total_dias, "días desde el año cero.")
```

### Código Completo
A continuación, el algoritmo en una sola celda para su ejecución:

```python
dia = int(input("Ingrese el día: "))
mes = int(input("Ingrese el mes: "))
anio = int(input("Ingrese el año: "))

total_dias = (anio * 365) + (mes * 30) + dia

print("Han pasado", total_dias, "días desde el año cero.")
```

## Enunciado del Ejercicio 6: Física - Movimiento de Proyectil
Cree un programa en Python, que asuma $x_0=0$ e $y_0=0$, y solicite la velocidad inicial del objeto a lanzar (asuma valores correctos correspondientes a $m/s$). Determine la posición en la que cae en el eje X e Y tras 10 segundos.

Ecuaciones dadas:
* $x(t) = x_0 + v_{x,0} \cdot t$
* $y(t) = y_0 + v_{y,0} \cdot t - 0.5 \cdot g \cdot t^2$

### Análisis y Estrategia de Resolución:
1. **Asignación de Constantes:** El problema nos da variables con valores estáticos: gravedad ($g = 9.8$), tiempo ($t = 10$), y las posiciones iniciales.
2. **Captura de Variables Dinámicas:** Necesitamos que el usuario proporcione las velocidades iniciales para cada eje (`vx0` y `vy0`). Usaremos `float`.
3. **Traducción de Ecuaciones a Código:** Escribiremos las fórmulas matemáticas utilizando los operadores de Python (`*` para multiplicar y `**` para exponente).

### Resolución Paso a Paso:

```python
# Bloque 1: Configuración de constantes y entrada de usuario
g = 9.8
t = 10
x0 = 0
y0 = 0

vx0 = float(input("Ingrese velocidad inicial en el eje X (m/s): "))
vy0 = float(input("Ingrese velocidad inicial en el eje Y (m/s): "))
```

```python
# Bloque 2: Cálculos de trayectoria parabólica
# Posición horizontal (movimiento rectilíneo uniforme)
xt = x0 + vx0 * t

# Posición vertical (movimiento uniformemente acelerado)
# Usamos 0.5 para representar 1/2 y t**2 para el tiempo al cuadrado
yt = y0 + vy0 * t - 0.5 * g * (t**2)
```

```python
# Bloque 3: Informe de posición
print("Tras 10 segundos, el objeto estará en:")
print(xt, "metros en el eje X")
print(yt, "metros en el eje Y")
```

### Código Completo
A continuación, el algoritmo en una sola celda para su ejecución:

```python
g = 9.8
t = 10
x0 = 0
y0 = 0

vx0 = float(input("Ingrese velocidad inicial en el eje X (m/s): "))
vy0 = float(input("Ingrese velocidad inicial en el eje Y (m/s): "))

xt = x0 + vx0 * t
yt = y0 + vy0 * t - 0.5 * g * (t**2)

print("Tras 10 segundos, el objeto estará en:")
print(xt, "metros en el eje X")
print(yt, "metros en el eje Y")
```
