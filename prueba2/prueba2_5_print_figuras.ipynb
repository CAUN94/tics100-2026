# Guía Práctica: Creación de Figuras Geométricas por Consola

En esta guía practicaremos la lógica espacial y matemática utilizando el doble ciclo `for`.

**IMPORTANTE:** En **ninguno** de estos ejercicios es necesario utilizar listas anidadas (matrices). Todo el trabajo se realiza evaluando exclusivamente las coordenadas de fila (`i`) y columna (`j`), e imprimiendo los caracteres directamente por pantalla utilizando `print(..., end='')`.

---
### Letra X en Cuadrícula

**Enunciado:**
Cree un código con una variable `n` escrita directamente en el programa. Utilice un doble ciclo `for` para generar una cuadrícula de tamaño `n x n` que dibuje una letra **X**. Utilice el carácter `#` para la letra y un espacio en blanco `' '` para el fondo.

**Ejemplo de ejecución (n = 5):**
```text
#   #
 # # 
  #  
 # # 
#   #
```

### Letra Z en Cuadrícula

**Enunciado:**
Cree un código con una variable `n` escrita directamente. Utilice un doble ciclo `for` para generar una cuadrícula que dibuje la letra **Z**. La letra está formada por la primera fila, la última fila y la diagonal inversa.

**Ejemplo de ejecución (n = 5):**
```text
#####
   # 
  #  
 #   
#####
```

### Letra I Especial con Grosor Adaptable

**Enunciado:**
Cree un código con una variable `n` escrita en el código. Debe dibujar una letra **I** muy particular: el techo y la base están formados por `#`. El cuerpo central está rodeado de ceros ` `, pero su "pilar" central está hecho de `#`.

El código debe adaptarse mágicamente: si `n` es un número impar, el pilar central tiene 1 carácter de grosor. Si `n` es un número par, el pilar central debe tener 2 caracteres de grosor en el medio.

**Ejemplo de ejecución (n = 6, par):**
```text
######
  ##  
  ##  
  ##  
  ##  
######
```

**Ejemplo de ejecución (n = 5, impar):**
```text
#####
  #  
  #  
  #  
#####
```

### Cuadrado Relleno

**Enunciado:**
Haga un código donde el largo del lado de un cuadrado (`n`) esté escrito en el programa. Debe imprimir una figura donde el perímetro del cuadrado esté compuesto por `#` y todo el interior esté relleno con el símbolo `-`.

**Ejemplo de ejecución (n = 5):**
```text
#####
#---#
#---#
#---#
#####
```

---
### Cuadrado con Borde Ajustable

**Enunciado:**
Haga un código que tenga el largo del cuadrado (`n`), pero que además tenga un segundo valor numérico llamado `borde`. 

Este segundo valor indicará qué tan "adentro" se debe dibujar el marco de `#`. Por ejemplo, si el borde es `0`, el borde es el contorno final exterior. Pero si el borde es `1`, el marco de `#` se dibujará un paso hacia adentro, dejando el exterior y el interior rellenados completamente con `-`.

**Ejemplo de ejecución (n = 7, borde = 1):**
```text
-------
-#####-
-#---#-
-#---#-
-#---#-
-#####-
-------
```

**Ejemplo de ejecución (n = 10, borde = 2):**
```text
----------
----------
--######--
--#----#--
--#----#--
--#----#--
--#----#--
--######--
----------
----------
```

### Marco Decorado con Esquinas

**Enunciado:**
Cree un cuadro vacío de tamaño `n x n`. Utilice el símbolo `+` exclusivamente para las 4 esquinas. Utilice el símbolo `-` para el techo y el piso, y el símbolo `|` para las paredes izquierda y derecha.

**Ejemplo de ejecución (n = 5):**
```text
+---+
|   |
|   |
|   |
+---+
```

### Triángulo Inverso Bajo la Diagonal

**Enunciado:**
Imprima un cuadrado de `n x n` que esté dividido diagonalmente. Todo lo que esté por encima de la diagonal inversa será representado con espacios en blanco, y todo lo que toque o esté por debajo de la diagonal inversa se rellenará con `#`.

**Ejemplo de ejecución (n = 5):**
```text
    #
   ##
  ###
 ####
#####
```

### La Letra Y

**Enunciado:**
Imprima la letra **Y** en un recuadro de `n x n`. La figura consiste en que, desde la mitad hacia arriba, se debe dibujar como una X (ambas diagonales). Desde la mitad hacia abajo, solo se debe dibujar un pilar recto en la columna central.

**Ejemplo de ejecución (n = 5):**
```text
#   #
 # # 
  #  
  #  
  #  
```

### Ventana Dividida en Cuadrantes

**Enunciado:**
Imprima una figura similar a una ventana clásica de vidrio. Tiene un marco completo en todo el borde, y además cuenta con una división sólida horizontal y vertical que cruzan exactamente por el centro, asuma que el n es impar siempre.

**Ejemplo de ejecución (n = 7):**
```text
#######
#  #  #
#  #  #
#######
#  #  #
#  #  #
#######
```
