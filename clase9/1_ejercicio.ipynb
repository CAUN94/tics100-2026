En este documento resolveremos un ejercicio complejo dividiéndolo en pequeños fragmentos de código. Veremos cómo el programa evoluciona y crece bloque a bloque, explicando el razonamiento detrás de cada nueva capa de lógica.

## Enunciado

Crea un programa que permita generar una "sopa de letras" aleatoria. El programa debe seguir estas reglas:
1. Pedir al usuario que ingrese la cantidad de filas y columnas que tendrá la sopa de letras.
2. Crear una matriz (lista anidada) del tamaño especificado, rellenándola con letras elegidas completamente al azar desde una lista predefinida: `['a', 'b', 'c', 'd']`.
3. Mostrar la matriz por pantalla en formato de cuadrícula.
4. Analizar la matriz creada y mostrar un reporte con la proporción (cantidad) de veces que aparece cada letra **por cada fila**.



---
Creación de la Matriz Aleatoria
Primero, nos enfocaremos únicamente en recibir las medidas del usuario y crear la lista anidada en la memoria del computador.

```python
import random

# 1. Solicitamos las dimensiones al usuario
filas = int(input("Ingrese la cantidad de filas: "))
columnas = int(input("Ingrese la cantidad de columnas: "))

letras_posibles = ['a', 'b', 'c', 'd']
sopa_de_letras = []

# 2. Construcción de la matriz con doble for
for i in range(filas):
    fila_temporal = []
    for j in range(columnas):
        # Elegimos una letra al azar de nuestra lista de opciones
        azar = random.randint(0, len(letras_posibles) - 1)
        letra_azar = letras_posibles[azar]
        fila_temporal.append(letra_azar)
        
    # Guardamos la fila completa en la matriz principal
    sopa_de_letras.append(fila_temporal)

print("Matriz creada en memoria con éxito (oculta por ahora).")
```

### Mostrar

Ahora que la matriz `sopa_de_letras` existe en la memoria, necesitamos que el usuario pueda verla de forma amigable.

Para mostrarla, no basta con hacer un `print(sopa_de_letras)` porque se vería como una lista desordenada en una sola línea. Debemos usar **otro doble ciclo `for`** que recorra las filas y columnas de la matriz que acabamos de crear.

Usaremos el truco de `end=" "` en el `print` interior para imprimir las letras una al lado de la otra, y un `print()` vacío en el ciclo exterior para realizar el salto de línea al terminar cada fila.

---
### Parte 1 + Parte 2: Creación y Visualización

```python
import random

filas = int(input("Ingrese la cantidad de filas: "))
columnas = int(input("Ingrese la cantidad de columnas: "))

letras_posibles = ['a', 'b', 'c', 'd']
sopa_de_letras = []

# --- PARTE 1: CREACIÓN ---
for i in range(filas):
    fila_temporal = []
    for j in range(columnas):
        # Elegimos una letra al azar de nuestra lista de opciones
        azar = random.randint(0, len(letras_posibles) - 1)
        letra_azar = letras_posibles[azar]
        fila_temporal.append(letra_azar)
        
    # Guardamos la fila completa en la matriz principal
    sopa_de_letras.append(fila_temporal)

# --- PARTE 2: VISUALIZACIÓN ---
print("\n--- SOPA DE LETRAS ---")
for i in range(filas):
    for j in range(columnas):
        # Imprimimos la letra actual separada por un espacio
        print(sopa_de_letras[i][j], end=" ")
    # Salto de línea al finalizar la fila
    print()
```

### Razonamiento de la Parte 3

Ya logramos crear la sopa y mostrarla. Ahora se nos pide un análisis estadístico: ¿Qué proporción de letras hay **por cada fila**?

Para resolver esto, necesitaremos un **tercer bloque con un doble ciclo `for`** que vuelva a recorrer la matriz. La clave aquí es dónde colocar nuestros contadores.

Como nos piden el análisis *por fila*, debemos inicializar los contadores de `a`, `b`, `c` y `d` en `0` **justo al entrar al primer `for` (el de las filas)**, pero antes del segundo `for` (el de las columnas). De esta forma, cada vez que bajemos a una nueva fila, los contadores se reiniciarán.

Dentro del ciclo de las columnas usaremos condicionales `if / elif` para detectar qué letra es y sumar 1 al contador correspondiente.

---
### Parte 1 + Parte 2 + Parte 3: Código Completo y Final

```python
import random

filas = int(input("Ingrese la cantidad de filas: "))
columnas = int(input("Ingrese la cantidad de columnas: "))

letras_posibles = ['a', 'b', 'c', 'd']
sopa_de_letras = []

# --- PARTE 1: CREACIÓN ---
for i in range(filas):
    fila_temporal = []
    for j in range(columnas):
        # Elegimos una letra al azar de nuestra lista de opciones
        azar = random.randint(0, len(letras_posibles) - 1)
        letra_azar = letras_posibles[azar]
        fila_temporal.append(letra_azar)
        
    # Guardamos la fila completa en la matriz principal
    sopa_de_letras.append(fila_temporal)

# --- PARTE 2: VISUALIZACIÓN ---
print("\n--- SOPA DE LETRAS ---")
for i in range(filas):
    for j in range(columnas):
        # Imprimimos la letra actual separada por un espacio
        print(sopa_de_letras[i][j], end=" ")
    # Salto de línea al finalizar la fila
    print()

# --- PARTE 3: ANÁLISIS DE PROPORCIONES ---
print("\n--- PROPORCIÓN DE LETRAS POR FILA ---")
for i in range(filas):
    # Reiniciamos los contadores a 0 CADA VEZ que evaluamos una fila nueva
    cuenta_a = 0
    cuenta_b = 0
    cuenta_c = 0
    cuenta_d = 0
    
    for j in range(columnas):
        letra_actual = sopa_de_letras[i][j]
        
        # Clasificamos la letra y aumentamos su contador
        if letra_actual == 'a':
            cuenta_a += 1
        elif letra_actual == 'b':
            cuenta_b += 1
        elif letra_actual == 'c':
            cuenta_c += 1
        elif letra_actual == 'd':
            cuenta_d += 1
            
    # Calculamos el porcentaje dividiendo por el total de columnas
    prop_a = (cuenta_a / columnas) * 100
    prop_b = (cuenta_b / columnas) * 100
    prop_c = (cuenta_c / columnas) * 100
    prop_d = (cuenta_d / columnas) * 100
            
    # Imprimimos el reporte de la fila actual antes de pasar a la siguiente
    print(f"Fila {i}:")
    print(f"  - 'a': {cuenta_a} vez/veces ({prop_a}%)")
    print(f"  - 'b': {cuenta_b} vez/veces ({prop_b}%)")
    print(f"  - 'c': {cuenta_c} vez/veces ({prop_c}%)")
    print(f"  - 'd': {cuenta_d} vez/veces ({prop_d}%)")
```

Una versión avanzada usando for para pasar por las letras y contarlas sería:


```python
import random

filas = int(input("Ingrese la cantidad de filas: "))
columnas = int(input("Ingrese la cantidad de columnas: "))

letras_posibles = ['a', 'b', 'c', 'd']
sopa_de_letras = []

# --- PARTE 1: CREACIÓN ---
for i in range(filas):
    fila_temporal = []
    for j in range(columnas):
        # Elegimos una letra al azar de nuestra lista de opciones
        azar = random.randint(0, len(letras_posibles) - 1)
        letra_azar = letras_posibles[azar]
        fila_temporal.append(letra_azar)
        
    # Guardamos la fila completa en la matriz principal
    sopa_de_letras.append(fila_temporal)

# --- PARTE 2: VISUALIZACIÓN ---
print("\n--- SOPA DE LETRAS ---")
for i in range(filas):
    for j in range(columnas):
        # Imprimimos la letra actual separada por un espacio
        print(sopa_de_letras[i][j], end=" ")
    # Salto de línea al finalizar la fila
    print()

# --- PARTE 3: ANÁLISIS DE PROPORCIONES ---
print("\n--- PROPORCIÓN DE LETRAS POR FILA ---")
for i in range(filas):
    lista_contador_letras = []
    for k in range(len(letras_posibles)):
        lista_contador_letras.append(0)
    
    for j in range(columnas):
        letra_actual = sopa_de_letras[i][j]
        
        # Usamos un tercer for para agilizar la revisión de la letra actual contra las opciones posibles
        for k in range(len(letras_posibles)):
            if letra_actual == letras_posibles[k]:
                lista_contador_letras[k] = lista_contador_letras[k] + 1 
        
    print(f"Fila {i}:")

    # Calculamos el porcentaje dividiendo por el total de columnas
    for k in range(len(letras_posibles)):
        prop_letra = (lista_contador_letras[k] / columnas) * 100
        print(f"  - '{letras_posibles[k]}': {lista_contador_letras[k]} vez/veces ({prop_letra}%)")
            

```
