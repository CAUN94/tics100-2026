# Introducción a las Matrices y Doble `for`

Una **matriz** es esencialmente una lista que contiene varias listas en su interior. Podemos imaginarla como una tabla con filas y columnas. 

Para recorrer o construir estas estructuras de dos dimensiones, la herramienta perfecta es el **doble ciclo `for`**.

## 1. Mostrando los valores de una Matriz

Vamos a analizar una matriz numérica de 5x5. Utilizaremos un doble `for` para extraer todos sus valores y veremos dos formas distintas de presentarlos en la pantalla.

```python
# Aquí tenemos una matriz de 5 filas y 5 columnas, con números del 11 al 55.

matriz = [
    [11, 12, 13, 14, 15], 
    [21, 22, 23, 24, 25], 
    [31, 32, 33, 34, 35], 
    [41, 42, 43, 44, 45], 
    [51, 52, 53, 54, 55]
]

# Ahora usaremos un ciclo anidado para recorrer la matriz y mostrar cada número junto con su posición (fila, columna).

print("--- Forma 1: Lista detallada de Coordenadas ---\n")
# El ciclo exterior (i) recorre las filas
for i in range(len(matriz)):
    # El ciclo interior (j) recorre cada columna de la fila actual
    for j in range(len(matriz[i])):
        # Accedemos al número usando el doble índice [fila][columna]
        print(f"posicion {i},{j} = {matriz[i][j]}")
```

La primera forma imprime una lista vertical. Sin embargo, normalmente queremos ver la matriz como una cuadrícula.

Para lograrlo usamos dos trucos con la función `print()`:
* **`end=" "`**: Lo agregamos al print interior para que, tras imprimir el número, deje un espacio en lugar de hacer un salto de línea. Así la fila se dibuja hacia el lado.
* **`print()` vacío**: Lo ponemos en el ciclo exterior, justo cuando termina el ciclo interior. Su única función es actuar como un "Enter" para bajar a la siguiente línea del cuadrado.

```python
print("--- Forma 2: Formato Cuadrícula ---\n")

for i in range(len(matriz)):
    # Imprimimos toda la fila de izquierda a derecha
    for j in range(len(matriz[i])):
        print(matriz[i][j], end=" ")
        
    # Cuando el ciclo 'j' termina, hacemos un salto de línea para bajar al siguiente nivel
    print()
```

## 2. Creando una Matriz Manualmente

¿Cómo construimos una matriz desde cero en código? Lo hacemos definiendo primero una matriz vacía (`[]`) y luego creando y agregando filas (`[]`) una por una mediante la función `.append()`.

```python
# 1. Creamos la matriz principal vacía
matriz_manual = []

# 2. Creamos la primera fila, la llenamos y la agregamos
fila1 = []
fila1.append(1)
fila1.append(2)
fila1.append(3)
matriz_manual.append(fila1) # Guardamos la lista 'fila1' dentro de la matriz

# Usamos el mismo proceso que mostramos arriba para mostrar la matriz, pero ahora con 'matriz_manual'

print("Matriz creada manualmente:")

for i in range(len(matriz_manual)):
    for j in range(len(matriz_manual[i])):
        print(matriz_manual[i][j], end=" ")
    print()

print()

# 3. Creamos la segunda fila, la llenamos y la agregamos
fila2 = []
fila2.append(4)
fila2.append(5)
fila2.append(6)
matriz_manual.append(fila2)

print("Matriz creada manualmente:")

for i in range(len(matriz_manual)):
    for j in range(len(matriz_manual[i])):
        print(matriz_manual[i][j], end=" ")
    print()

print()

# 4. Creamos la tercera fila, la llenamos y la agregamos
fila3 = []
fila3.append(7)
fila3.append(8)
fila3.append(9)
matriz_manual.append(fila3)

print("Matriz creada manualmente:")

for i in range(len(matriz_manual)):
    for j in range(len(matriz_manual[i])):
        print(matriz_manual[i][j], end=" ")
    print()

```

## 3. Construcción Dinámica con Doble `for`

Crear filas manualmente es inviable si necesitamos tableros gigantes. Por eso, utilizamos el doble `for` para automatizar la creación de "listas dentro de listas".

El ciclo exterior (`i`) se encargará de crear una lista vacía nueva en cada vuelta. El ciclo interior (`j`) rellenará esa lista vacía. Finalmente, el ciclo exterior tomará la lista recién rellenada y la guardará en la matriz central.

En este caso lo vamos a rellenar con #

```python
matriz_gato = []

# Queremos generar 5 filas
for i in range(5):
    # En cada vuelta de 'i', creamos una nueva fila vacía temporal.
    # Esto quiere decir que cada fila es una lista nueva, independiente de las otras.
    # Al redefinir la fila se borra lo que había antes, y se empieza a llenar desde cero.
    fila_temporal = []
    
    # El ciclo interior rellena esta fila con 5 símbolos
    for j in range(5):
        # print(f"Rellenando fila {i}, columna {j}")
        fila_temporal.append("#")
        
    # Terminó de rellenarse. Agregamos esta fila lista a la matriz principal
    matriz_gato.append(fila_temporal)

print("Matriz de 5x5 generada dinámicamente con código:\n")
for i in range(len(matriz_gato)):
    for j in range(len(matriz_gato[i])):
        print(matriz_gato[i][j], end=" ")
    print()
```
