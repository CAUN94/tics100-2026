# Clase: Estructuras de Repetición en Python (For y While)

En programación, los **ciclos** (o bucles) nos permiten ejecutar un bloque de código varias veces paso a paso, sin tener que escribirlo manualmente una y otra vez. Python nos ofrece dos herramientas principales para lograr esto: `for` y `while`.

## 1. El Ciclo `for`

El ciclo `for` se utiliza para recorrer (leer paso a paso) una **secuencia** ordenada de datos, como por ejemplo una lista de elementos.

### Concepto de Iteración y Variable Dinámica
Cuando usamos un `for`, la información de la lista se va guardando temporalmente en una **variable de control**. Decimos que esta variable es **dinámica** porque su valor cambia automáticamente en cada iteración (cada "vuelta" del ciclo), tomando el siguiente valor de la lista en estricto orden.

```python
# Definimos una lista de nombres ordenados
estudiantes = ["Ana", "Luis", "Marta"]

# 'nombre' es la variable dinámica que cambiará en cada iteración
for nombre in estudiantes:
    # En la 1ra vuelta valdrá Ana, en la 2da Luis, en la 3ra Marta
    print(f"Iteración actual - La variable 'nombre' ahora vale: {nombre}")
```

```python
# Ejemplo similar: leyendo una lista de colores
colores = ["Rojo", "Azul", "Verde"]

# La variable dinámica ahora se llama 'color'
# Funciona exactamente igual, va cambiando en orden
for color in colores:
    print("Color seleccionado:", color)
```

### Uso de `range()`

El comando `range()` es sumamente útil cuando no tenemos una lista previa creada, pero necesitamos repetir una acción un número exacto de veces.

Básicamente, `range()` **construye una secuencia de números**. Para demostrarlo, podemos envolverlo en el comando `list()`. Esto revelará la lista invisible de números que el ciclo `for` va a leer paso a paso.

Al igual que con las listas de texto, la variable dinámica irá cambiando y tomando cada uno de los números numéricos de ese rango.

```python
# Mostramos qué construye realmente range(5) usando list()
mi_rango = range(5)
print("Lo que hay dentro del rango generado es una lista:", list(mi_rango))

# La variable dinámica 'i' va cambiando su valor numérico en cada iteración
for i in range(1, 6): # Genera números del 1 al 5 (el 6 no se incluye)
    print(f"Contando... paso número {i}")
```

```python
# Ejemplo similar: range con saltos. Estructura: range(inicio, fin, salto)
# Genera números del 0 al 10, saltando de 2 en 2
print("Rango con saltos:", list(range(0, 11, 2)))

# La variable 'numero' toma los valores 0, 2, 4, 6, 8, 10 en orden.
for numero in range(0, 11, 2):
    print(f"Número par generado por range: {numero}")
```

### Formas útiles de usar el `for` combinado con `if`

El ciclo `for` desata todo su potencial cuando introducimos condicionales (`if`) en su interior para analizar los datos en cada vuelta.

#### A. Contar casos (El Patrón Contador)
Consiste en crear una variable inicializada en `0` justo antes del ciclo. Cada vez que el `if` encuentre una condición verdadera, le sumamos `1` a nuestra variable.

```python
numeros = [1, 5, 8, 10, 13, 20]
cuenta_pares = 0 # Variable para guardar el conteo inicializada en 0

for n in numeros:
    # Condición: Si el resto de dividir por 2 es cero, el número es par
    if n % 2 == 0: 
        cuenta_pares = cuenta_pares + 1 # Sumamos 1 al contador

print(f"Total de números pares encontrados: {cuenta_pares}")
```

```python
# Ejemplo similar: Contar cuántas personas son mayores a 10 años
edades = [5, 12, 18, 8, 21, 9]
mayores_de_10 = 0

for edad in edades:
    if edad > 10:
        # += 1 es exactamente lo mismo que hacer: mayores_de_10 = mayores_de_10 + 1
        mayores_de_10 += 1 

print(f"Hay {mayores_de_10} personas mayores a 10 años en la lista.")
```

#### B. Sumar y Multiplicar (El Patrón Acumulador)
Muy parecido al contador, pero en lugar de sumar 1, sumamos (o multiplicamos) el valor actual de la variable dinámica.

```python
valores = [2, 3, 4]
total_suma = 0
total_multiplicacion = 1 # Empezamos en 1 porque si empezamos en 0, multiplicar siempre daría 0

for v in valores:
    total_suma = total_suma + v
    total_multiplicacion = total_multiplicacion * v

print(f"Suma total: {total_suma}")
print(f"Multiplicación total: {total_multiplicacion}")
```

```python
# Ejemplo similar: Sumar los precios de un carrito de compras virtual
precios = [1500, 300, 5000, 200]
total_pagar = 0

for precio in precios:
    total_pagar += precio # Acumulamos cada precio al total

print(f"El total a pagar es: ${total_pagar}")
```

#### C. Bandera o Flag (Confirmar si un evento ocurre)
Se usa para leer toda una lista y descubrir si **al menos un elemento** cumple una condición. Empezamos asumiendo que no se cumple (`False`) y si el `if` la encuentra, "levantamos la bandera" cambiándola a `True`.

```python
numeros_azar = [1, 3, 5, 7, 10, 11]
hay_par = False # Suponemos que no hay ningún par al inicio (Bandera baja)

for n in numeros_azar:
    if n % 2 == 0:
        hay_par = True # Encontramos al menos uno, "levantamos la bandera"

# Evaluamos la bandera al final del ciclo
if hay_par:
    print("¡Se encontró al menos un número par en toda la lista!")
else:
    print("No se encontraron números pares.")
```

```python
# Ejemplo similar: Verificar si hay temperaturas bajo cero
temperaturas = [12, 15, 20, -2, 8]
hubo_helada = False # Bandera baja

for temp in temperaturas:
    if temp < 0:
        hubo_helada = True # Bandera arriba

if hubo_helada:
    print("¡Atención! Hubo temperaturas bajo cero registradas.")
else:
    print("Todo normal, ninguna temperatura fue negativa.")
```

## 2. El Ciclo `while`

A diferencia del `for` (que repite una cantidad de veces definida por los elementos de una lista o un `range`), el ciclo `while` se ejecuta **mientras una condición lógica siga siendo Verdadera (`True`)**.

### La lógica inversa en la condición
Un error muy común de los programadores novatos es poner el objetivo final como condición en el `while`. **¡En realidad hay que poner lo contrario!**

Piensa en un guardia de seguridad. Si queremos detener el ciclo cuando el usuario escriba la palabra `'salir'`, el guardia debe decir: **"Te seguiré reteniendo en el bucle MIENTRAS tu palabra SEA DISTINTA a 'salir'"**.

```python
palabra = ""

# Explicación de la condición:
# Nuestro objetivo es salir cuando la palabra sea 'python'.
# Por lo tanto, el ciclo debe continuar ejecutándose MIENTRAS la palabra NO SEA 'python'.
while palabra != "python":
    palabra = input("Escribe la palabra mágica ('python') para salir: ")
    
    if palabra != "python":
        print("Equivocado, sigue intentando.")

print("¡Acceso concedido! Lograste salir del ciclo while.")
```

```python
# Ejemplo similar: Forzar al usuario a ingresar obligatoriamente un número positivo
numero_ingresado = -1

# Queremos salir del bucle solo cuando ingrese algo mayor a 0 (>0).
# Entonces, el bucle repite MIENTRAS el número ingresado sea menor o igual a 0 (<=0).
while numero_ingresado <= 0:
    numero_ingresado = int(input("Por favor, ingresa un número positivo (>0): "))
    
    if numero_ingresado <= 0:
        print("Ese no es positivo, intenta de nuevo...")

print(f"¡Excelente! Ingresaste el número válido: {numero_ingresado}.")
```
