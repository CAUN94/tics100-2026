### Ejercicio 1: Obtención de índices usando `range(len())`

**Concepto:** Cuando usamos un simple `for elemento in lista`, podemos *ver* los datos, pero no podemos modificarlos en su lugar de origen. Si queremos alterar la lista original, necesitamos saber el número de la casilla exacta (el índice `i`). Para eso combinamos `range()` (que genera números) y `len()` (que nos dice cuántos elementos hay).

**Enunciado:** 
Tienes una lista de precios base. Usa un ciclo `for` con `range(len(lista))` para recorrer la lista entera. Multiplica cada precio por 2 y guarda el resultado exactamente en la misma posición de la lista original. Muestra la lista antes y después.

### Ejercicio 2: Mostrando una Matriz por Pantalla

**Concepto:** Una matriz ya creada en memoria se ve fea si la imprimimos de golpe con un `print(matriz)`. Para verla como una cuadrícula, necesitamos un ciclo que recorra las filas (hacia abajo) y otro ciclo adentro que recorra las columnas (hacia el lado).

**Enunciado:**
Dada una matriz de 3x3 predefinida, utiliza un doble ciclo `for` para mostrarla en pantalla. Recuerda usar `end=" "` para que los números queden al lado, y un `print()` vacío para hacer el salto de línea al terminar la fila.

### Ejercicio 3: Creación de una Matriz Aleatoria

**Concepto:** Para construir una matriz desde cero, necesitamos crear una lista general vacía (`[]`). Luego, con un doble ciclo `for`, creamos listas "temporales" (que serán las filas), las rellenamos con datos y finalmente metemos esa fila completa dentro de la lista general.

**Enunciado:**
Usando la librería `random`, crea una matriz de 4x4. El ciclo exterior creará las filas y el interior generará números al azar entre 0 y 9. Al terminar de construirla, usa otro doble `for` para mostrarla por pantalla.

### Ejercicio 4: ¿Cómo funciona la Revisión de Vecinos?

**Concepto:** Para revisar qué hay alrededor de una coordenada, no escribimos 8 condiciones a mano. Usamos "desplazamientos". Iteramos una lista `[-1, 0, 1]` para las filas y columnas. Al sumar esos desplazamientos a nuestra coordenada central, ¡obtenemos las coordenadas de los vecinos mágicamente!

**Enunciado:**
Tenemos una matriz pequeña de 3x3 y nuestro objetivo está en el centro (Fila 1, Columna 1). Usa un doble `for` iterando sobre `[-1, 0, 1]` para imprimir el valor de todos los vecinos. Incluye la regla para NO leer el centro (nosotros mismos) y la regla para no salirte de los bordes.
