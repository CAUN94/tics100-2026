### Ejercicio: Reporte de Temperaturas Semanales

**Descripción:** Tienes 3 listas paralelas que contienen la información climática de una semana: los días, las temperaturas mínimas y las temperaturas máximas. 

Usando un ciclo `for`, debes calcular la temperatura promedio de cada día. Muestra por pantalla un reporte diario indicando el día, la mínima, la máxima y el promedio. Además, guarda cada promedio calculado en una lista nueva llamada `promedios` e imprímela al final del programa.

**Hint:** Utiliza `range(len(dias))` para generar un índice que te permita acceder a la posición correspondiente en las tres listas al mismo tiempo. Recuerda que el promedio se calcula sumando la mínima y la máxima, y dividiendo el resultado por 2.

```python
dias = ["Lunes", "Martes", "Miércoles", "Jueves", "Viernes", "Sábado", "Domingo"]
temp_min = [12, 10, 11, 14, 15, 13, 12]
temp_max = [22, 20, 24, 26, 28, 25, 23]

promedios = []

print("--- REPORTE CLIMÁTICO DIARIO ---")

# Usamos range(len()) para iterar sobre los índices (del 0 al 6)
for i in range(len(dias)):
    
    # Calculamos el promedio sumando min y max de la posición 'i', y dividiendo por 2
    promedio_dia = (temp_min[i] + temp_max[i]) / 2
    
    # Guardamos el resultado en nuestra nueva lista
    promedios.append(promedio_dia)
    
    # Imprimimos el reporte formateado para el usuario
    print(f"{dias[i]}: Min {temp_min[i]}°C | Max {temp_max[i]}°C -> Promedio: {promedio_dia}°C")


print("\n--- RESUMEN DE PROMEDIOS ---")
print("Lista guardada en memoria:", promedios)
```
