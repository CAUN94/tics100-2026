**Enunciado del Ejercicio:**

En un laboratorio de física, se está realizando un experimento para medir la capacidad térmica del agua. Para ello, se calienta una muestra de agua y se registra la temperatura en diferentes momentos.

Escriba un programa que solicite al usuario la cantidad de mediciones a registrar y luego reciba cada una de las temperaturas. Finalmente, el programa debe mostrar el promedio de las temperaturas, la temperatura máxima y la temperatura mínima registrada durante el experimento.

### Paso 1: Entender el calculo del promedio matematicamente
Antes de procesar muchas temperaturas usando ciclos, veamos como funciona la logica base si tuvieramos solo 3 temperaturas fijas.

Para calcular el promedio, simplemente sumamos todas las temperaturas obtenidas y dividimos ese gran total por la cantidad de mediciones que hicimos.

```python
temp1 = 15.5
temp2 = 22.0
temp3 = 18.2

cantidad_mediciones = 3

suma_total = temp1 + temp2 + temp3
promedio = suma_total / cantidad_mediciones

print("Temperatura 1:", temp1)
print("Temperatura 2:", temp2)
print("Temperatura 3:", temp3)
print("\nEl promedio manual es:", round(promedio, 2))
```

### Paso 2: El concepto de encontrar el mayor y menor en una lista
Para que una computadora encuentre el numero mayor en un grupo de opciones, no puede verlos todos a la vez como nosotros. Debe ir evaluando uno por uno.

Si tenemos las temperaturas guardadas en una lista de 3 opciones, el truco logico esta en **asumir que el primer valor (la posicion 0) es el mas grande y el mas pequeño a la vez** y guardarlo en una variable.

Luego, lo comparamos con el siguiente numero usando un condicional `if`. Si el numero nuevo es mas grande, actualizamos nuestra variable. Si no lo es, lo ignoramos. 

Veamos como se ejecuta esto paso a paso con mensajes explicativos:

```python
temperaturas = [15.5, 22.0, 18.2]
print("Opciones de temperaturas:", temperaturas)

# Asumimos que el primer elemento es el maximo y minimo inicial
temp_maxima = temperaturas[0]
temp_minima = temperaturas[0]
print("\n--- INICIO ---")
print("Asumimos que el primer valor (", temperaturas[0], ") es el mayor y menor inicial.")

# Comparamos con el segundo elemento (posicion 1)
print("\n--- COMPARANDO SEGUNDO VALOR ---")
print("Evaluando el valor:", temperaturas[1])
if temperaturas[1] > temp_maxima:
    print("-> Es MAYOR que el maximo actual (", temp_maxima, "). Actualizamos la variable.")
    temp_maxima = temperaturas[1]
if temperaturas[1] < temp_minima:
    print("-> Es MENOR que el minimo actual (", temp_minima, "). Actualizamos la variable.")
    temp_minima = temperaturas[1]

# Comparamos con el tercer elemento (posicion 2)
print("\n--- COMPARANDO TERCER VALOR ---")
print("Evaluando el valor:", temperaturas[2])
if temperaturas[2] > temp_maxima:
    print("-> Es MAYOR que el maximo actual (", temp_maxima, "). Actualizamos la variable.")
    temp_maxima = temperaturas[2]
if temperaturas[2] < temp_minima:
    print("-> Es MENOR que el minimo actual (", temp_minima, "). Actualizamos la variable.")
    temp_minima = temperaturas[2]

print("\n--- RESULTADO FINAL ---")
print("La temperatura maxima es:", temp_maxima)
print("La temperatura minima es:", temp_minima)
```

### Paso 3: Evitar el trabajo manual usando for y range()
Si el experimento tuviera 100 mediciones, no podriamos escribir 100 bloques `if` a mano. Para solucionar esto, usamos un ciclo `for` con `range()`.

Esta herramienta nos permite repetir una instruccion de pedir datos la cantidad de veces que queramos.

```python
cantidad_mediciones = 3
print("Iniciando la recoleccion de datos...")

# El ciclo repetira el input la cantidad de veces que indique la variable
for i in range(cantidad_mediciones):
    
    # La variable 'i' comienza en 0, asi que le sumamos 1 solo para que en pantalla diga "1, 2, 3"
    medicion_actual = i + 1
    temp = float(input(f"Ingrese la temperatura {medicion_actual}: "))
    
    print("Se ha registrado en el sistema la temperatura:", temp)
```

### Paso 4: El uso de 'i' para detectar la primera iteracion
Al automatizar la recoleccion de datos, surge un problema logico: ¿Como asignamos el primer maximo y minimo si el usuario ingresa los datos uno a uno y ya no tenemos una lista con la posicion 0?

La solucion esta en la variable `i` de nuestro ciclo `for`. Recuerda que `i` siempre comienza valiendo `0` en la primera vuelta.

Por lo tanto, utilizamos la condicion `if i == 0:` para decirle a la computadora: *"Esta es la primera temperatura que ingresa el usuario. Como es la unica que tenemos, guardala automaticamente como la maxima y minima inicial sin comparar con nada"*.

En todas las vueltas siguientes (`else`), la computadora sabra que ya tiene un valor guardado y recien ahi aplicara las comparaciones que vimos en el Paso 2.

```python
print("Demostracion del control de iteraciones:")
cantidad = 3

for i in range(cantidad):
    temp = float(input(f"\nIngrese la temperatura {i + 1}: "))
    
    if i == 0:
        print(" -> Detectamos i=0. Esta es la primera vuelta.")
        print(" -> Asignamos", temp, "como nuestro primer maximo y minimo inicial.")
        temp_max = temp
        temp_min = temp
    else:
        print(" -> Estamos en la vuelta donde i=", i)
        print(" -> Ahora si podemos comparar", temp, "con los valores anteriores.")
        if temp > temp_max:
            temp_max = temp
        if temp < temp_min:
            temp_min = temp
```

### Paso 5: Solucion final uniendo conceptos
Ahora armaremos el sistema definitivo. 

Usaremos el ciclo `for` para pedir las temperaturas una por una. Dentro de ese mismo ciclo, iremos sumando cada temperatura a un acumulador (para calcular el promedio final) y usaremos la logica del `if i == 0:` para manejar el maximo y minimo paso a paso.

```python
# 1. Solicitamos la cantidad total de mediciones
cantidad_mediciones = int(input("Ingrese la cantidad de temperaturas a registrar: "))

# 2. Creamos nuestro acumulador inicial antes de que empiece el ciclo
suma_temperaturas = 0.0
temp_maxima = 0.0
temp_minima = 0.0

# 3. Iniciamos el ciclo de recoleccion y evaluacion
for i in range(cantidad_mediciones):
    temp = float(input(f"Ingrese la temperatura {i + 1}: "))
    
    # Vamos acumulando la suma paso a paso
    suma_temperaturas = suma_temperaturas + temp
    
    # 4. Evaluamos maximo y minimo usando la logica de la iteracion cero
    if i == 0:
        temp_maxima = temp
        temp_minima = temp
    else:
        if temp > temp_maxima:
            temp_maxima = temp
            
        if temp < temp_minima:
            temp_minima = temp

# 5. Al terminar el ciclo for, calculamos el promedio y mostramos los datos finales
promedio = suma_temperaturas / cantidad_mediciones

print("\n--- RESULTADOS DEL EXPERIMENTO ---")
print("Mediciones realizadas:", cantidad_mediciones)
print("Temperatura Promedio:", round(promedio, 2))
print("Temperatura Maxima detectada:", temp_maxima)
print("Temperatura Minima detectada:", temp_minima)
```
