##  Clasificación Básica (If / Else) (Básico)
Tienes las edades de dos personas guardadas en variables separadas. 
Tu objetivo es evaluar cada edad individualmente y usar `.append()` para enviar el dato a la lista que corresponda:
- Si la persona tiene 18 años o más, agrégala a la lista `adultos`.
- Si tiene menos de 18 años, agrégala a la lista `menores`.

### Análisis y Estrategia de Resolución:
Este es el flujo de decisión más sencillo. Solo hay dos caminos posibles.
1. Creamos las listas vacías.
2. Usamos un `if` para preguntar si la edad es >= 18.
3. Usamos un `else` (si no se cumple lo anterior) para el camino contrario.

```python
# Edades individuales
edad_persona_a = 25
edad_persona_b = 14

# Listas contenedoras
adultos = []
menores = []

# Evaluación de la Persona A
if edad_persona_a >= 18:
    adultos.append(edad_persona_a)
else:
    menores.append(edad_persona_a)

# Evaluación de la Persona B
if edad_persona_b >= 18:
    adultos.append(edad_persona_b)
else:
    menores.append(edad_persona_b)

print("Lista de Adultos:", adultos)
print("Lista de Menores:", menores)
```

## Separación por Condición Única

Tienes los saldos de tres cuentas bancarias. Evalúa cada saldo:
- Si el saldo es negativo (menor a 0), envíalo a la lista `cuentas_morosas`.
- Si el saldo es 0 o positivo, envíalo a la lista `cuentas_al_dia`.

### Análisis y Estrategia de Resolución:
Aplicaremos la misma lógica anterior (If/Else), pero repetida tres veces de forma independiente para cada variable.

```python
# Saldos de las cuentas
saldo_cuenta_1 = -15000
saldo_cuenta_2 = 45000
saldo_cuenta_3 = -500

cuentas_morosas = []
cuentas_al_dia = []

# Evaluamos la Cuenta 1
if saldo_cuenta_1 < 0:
    cuentas_morosas.append(saldo_cuenta_1)
else:
    cuentas_al_dia.append(saldo_cuenta_1)

# Evaluamos la Cuenta 2
if saldo_cuenta_2 < 0:
    cuentas_morosas.append(saldo_cuenta_2)
else:
    cuentas_al_dia.append(saldo_cuenta_2)

# Evaluamos la Cuenta 3
if saldo_cuenta_3 < 0:
    cuentas_morosas.append(saldo_cuenta_3)
else:
    cuentas_al_dia.append(saldo_cuenta_3)

print("Saldos morosos (negativos):", cuentas_morosas)
print("Saldos al día (positivos):", cuentas_al_dia)
```

## Comparación Lógica Condicional

Las agencias de Santiago y Mendoza registraron si llovió (1) o no llovió (0) en un día específico. 
Tu objetivo es evaluar el clima de ese día y agregar a una lista llamada `reporte_clima` un texto que describa el patrón de lluvia:
- Si llovió en ambas, agrega: "Lluvia en ambas".
- Si no llovió en ninguna, agrega: "Seco en ambas".
- Si el clima fue distinto, agrega: "Clima diferente".

### Análisis y Estrategia de Resolución:
En este nivel agregamos la instrucción `elif` para manejar más de dos caminos posibles, y usamos `and` para evaluar dos cosas al mismo tiempo.
1. Crearemos nuestra lista vacía `reporte_clima`.
2. Usaremos operadores lógicos (`and`) dentro de un `if/elif/else` para comparar simultáneamente el clima de ambas ciudades.
3. Usaremos `.append()` para guardar el resultado de nuestra conclusión.

```python
# 1. Definición de variables (1 = Lluvia, 0 = No lluvia)
santiago_dia_1 = 1
mendoza_dia_1 = 1

reporte_clima = []

# 2. Evaluación condicional del patrón
if santiago_dia_1 == 1 and mendoza_dia_1 == 1:
    reporte_clima.append("Lluvia en ambas")
elif santiago_dia_1 == 0 and mendoza_dia_1 == 0:
    reporte_clima.append("Seco en ambas")
else:
    reporte_clima.append("Clima diferente")

# 3. Mostrar el resultado
print("El reporte final del día es:", reporte_clima)
```


## Ordenamiento Lógico Manual
La multitienda LeroMart tiene una oferta "Lleve 3 y pague 2", donde el cliente obtiene gratis el producto más barato. 
Tienes 3 productos en tu carro. Evalúa cuál es el más barato usando condiciones lógicas. 
- Agrega el precio del producto gratis a la lista `productos_gratis`.
- Agrega los precios de los otros dos productos a la lista `boleta_cobro`.

### Análisis y Estrategia de Resolución:
No usaremos funciones de ordenamiento (`.sort()`) ni ciclos. Lo haremos con pura lógica deductiva, enviando datos a múltiples listas a la vez.
1. Para saber si el Producto 1 es el más barato, debe ser menor o igual al Producto 2 **Y** menor o igual al Producto 3.
2. Aplicamos esta misma lógica en un bloque `if/elif/else`.
3. Dependiendo de quién gane como el "más barato", hacemos `.append()` distribuyendo los precios en las listas correspondientes.

```python
# Precios de los tres productos llevados por el cliente
precio_prod1 = 4000
precio_prod2 = 1000
precio_prod3 = 2000

productos_gratis = []
boleta_cobro = []

# Evaluamos si el Producto 1 es el más barato
if precio_prod1 <= precio_prod2 and precio_prod1 <= precio_prod3:
    productos_gratis.append(precio_prod1)
    boleta_cobro.append(precio_prod2)
    boleta_cobro.append(precio_prod3)

# Si no lo fue, evaluamos si el Producto 2 es el más barato
elif precio_prod2 <= precio_prod1 and precio_prod2 <= precio_prod3:
    productos_gratis.append(precio_prod2)
    boleta_cobro.append(precio_prod1)
    boleta_cobro.append(precio_prod3)

# Si ninguno de los anteriores fue el más barato, por descarte es el Producto 3
else:
    productos_gratis.append(precio_prod3)
    boleta_cobro.append(precio_prod1)
    boleta_cobro.append(precio_prod2)

print("Te llevas GRATIS el monto de:", productos_gratis)
print("Se te cobrarán los montos de:", boleta_cobro)
```

## Toma de Decisiones de Optimización (Difícil)
Deseas instalar un sistema de calefacción. Has pre-calculado el índice de "rendimiento" (costo vs calor) de tres sistemas: Estufa, Eléctrico y Leña. Mientras MENOR sea el índice, mejor es el sistema.
Tu objetivo es:
1. Guardar los tres índices en una lista llamada `registro_sistemas` usando `.append()`.
2. Evaluar cuál tiene el índice más bajo y agregar **el nombre del sistema ganador** a una lista llamada `recomendacion_final`.

### Análisis y Estrategia de Resolución:
1. Realizamos cálculos matemáticos simples previos.
2. Llenamos nuestra primera lista de forma secuencial.
3. Comparamos las variables con `if/elif/else`.
4. Guardamos un String (texto) en la lista de recomendación.

```python
# Variables pre-calculadas (mientras más bajo, mejor para el bolsillo)
indice_estufa = 10.2
indice_electrico = 7.1
indice_lena = 9.7

registro_sistemas = []
recomendacion_final = []

# 1. Guardamos el registro de los índices evaluados
registro_sistemas.append(indice_estufa)
registro_sistemas.append(indice_electrico)
registro_sistemas.append(indice_lena)
print("Índices registrados:", registro_sistemas)

# 2. Encontramos el sistema óptimo
if indice_estufa < indice_electrico and indice_estufa < indice_lena:
    recomendacion_final.append("Estufa tradicional")
elif indice_electrico < indice_estufa and indice_electrico < indice_lena:
    recomendacion_final.append("Sistema Eléctrico")
else:
    recomendacion_final.append("Calefacción a Leña")

print("El sistema recomendado para comprar es:", recomendacion_final)
```
