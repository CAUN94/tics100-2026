# Guía Práctica: Ejercicios de `def` y Funciones Simples

Las funciones (`def`) son como pequeñas fábricas: les entregamos materiales (argumentos), hacen un trabajo con ellos, y nos devuelden un producto terminado. 

### El `return` (Devolver un valor)

Imagina una función que suma 3 valores (`a`, `b` y `c`). Si dentro de la función usamos `print()`, el resultado aparecerá en la pantalla, pero el programa "olvidará" ese número inmediatamente y no podremos seguir usándolo para otros cálculos matemáticos.

Para solucionar esto usamos **`return`**. Esta sentencia expulsa el resultado hacia afuera de la función. 

**REGLA DE ORO:** Como `return` NO imprime nada en pantalla, cuando llamamos a la función **estamos obligados a guardar esa respuesta en una variable** posterior a su uso para no perderla, como si fuera un input().

```python
# DEFINICIÓN DE LA FUNCIÓN
def sumar_tres_valores(a, b, c):
    resultado = a + b + c
    return resultado

# ==========================================
# USO CORRECTO EN EL PROGRAMA PRINCIPAL
# ==========================================

# Si hacemos esto, el PC calcula el 12, pero se pierde en el vacío porque no lo guardamos:
sumar_tres_valores(5, 5, 2) 

# Lo CORRECTO es almacenar la respuesta que retorna la función en una variable:
mi_suma = sumar_tres_valores(5, 5, 2)

# Ahora podemos usar esa variable para lo que queramos:
print("El valor almacenado es:", mi_suma)
```

## Ejercicio 1: Números Primos

**Enunciado:**
Crea una función llamada `esPrimo(n)` que reciba un número. La función debe evaluar si el número es primo o no. 
Debe utilizar **`return`** para devolver el valor booleano `True` (si es primo) o `False` (si no lo es).

```python
# --- RESPUESTA ---
def esPrimo(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

# ==========================================
# PRUEBAS (NO TOCAR) - Ejecuta para probar
# ==========================================
print("¿7 es primo?:", esPrimo(7))   # Respuesta esperada: True
print("¿10 es primo?:", esPrimo(10)) # Respuesta esperada: False
print("¿13 es primo?:", esPrimo(13)) # Respuesta esperada: True
```

## Ejercicio 2: ¿Quién es el mayor?

**Enunciado:**
Crea una función llamada `mayor(a, b)`. Se le dan dos valores y debe evaluar las condiciones. 

**IMPORTANTE:** A diferencia de las demás, esta función **SÍ usa `print()` y NO retorna nada**. Debe imprimir directamente un mensaje diciendo si `a` es mayor que `b`, si `b` es mayor que `a`, o si son iguales.

```python
# --- RESPUESTA ---
def mayor(a, b):
    if a > b:
        print(f"{a} es mayor que {b}")
    elif b > a:
        print(f"{b} es mayor que {a}")
    else:
        print("Son iguales")

# ==========================================
# PRUEBAS (NO TOCAR) - Ejecuta para probar
# ==========================================
mayor(10, 5)  # Respuesta esperada: 10 es mayor que 5
mayor(3, 8)   # Respuesta esperada: 8 es mayor que 3
mayor(7, 7)   # Respuesta esperada: Son iguales
```

## Ejercicio 3: Clasificador de Triángulos

**Enunciado:**
Crea una función llamada `tipoTriangulo(lado1, lado2, lado3)` en la que se dan los 3 lados de un triángulo.
La función debe evaluar las medidas y utilizar **`return`** para devolver el texto con el nombre correspondiente del triángulo (Equilátero, Isósceles o Escaleno).

```python
# --- RESPUESTA ---
def tipoTriangulo(lado1, lado2, lado3):
    if lado1 == lado2 == lado3:
        return "Equilátero"
    elif lado1 == lado2 or lado2 == lado3 or lado1 == lado3:
        return "Isósceles"
    else:
        return "Escaleno"

# ==========================================
# PRUEBAS (NO TOCAR) - Ejecuta para probar
# ==========================================
print("Lados 5, 5, 5:", tipoTriangulo(5, 5, 5)) # Respuesta esperada: Equilátero
print("Lados 4, 4, 6:", tipoTriangulo(4, 4, 6)) # Respuesta esperada: Isósceles
print("Lados 3, 4, 5:", tipoTriangulo(3, 4, 5)) # Respuesta esperada: Escaleno
```

## Ejercicio 4: Ordenar y Matriz Automática

**Enunciado:**
En este ejercicio deberás crear dos funciones que trabajen juntas en el programa principal:

1. **`ordenar_tres(a, b, c)`**: Recibe 3 números y usa `return` para devolverlos guardados en una lista ordenada del menor al mayor.
2. **`crear_matriz_repetida(lista)`**: Recibe la lista que retornó la función anterior. Debe crear y retornar una matriz de 3x3 donde la primera fila esté compuesta puramente por el primer valor de la lista, la segunda fila por el segundo valor de la lista, y la tercera fila por el tercer valor.

```python
# --- RESPUESTA ---
def ordenar_tres(a, b, c):
    if a <= b and a <= c and b <= c:
        lista = [a, b, c]
    elif a <= b and a <= c and c <= b:
        lista = [a, c, b]
    elif b <= a and b <= c and a <= c:
        lista = [b, a, c]
    elif b <= a and b <= c and c <= a:
        lista = [b, c, a]
    elif c <= a and c <= b and a <= b:
        lista = [c, a, b]
    else:
        lista = [c, b, a]

    return lista

def crear_matriz_repetida(lista):
    matriz = []
    for i in range(3):
        # En la fila 'i', repetimos el elemento que está en la posición 'i' de la lista original
        fila = [lista[i], lista[i], lista[i]]
        matriz.append(fila)
    return matriz

# ==========================================
# PRUEBAS (NO TOCAR) - Ejecuta para probar
# ==========================================
lista_ordenada = ordenar_tres(9, 2, 5)
print("Lista ordenada generada:", lista_ordenada)
# Respuesta esperada: [2, 5, 9]

matriz_final = crear_matriz_repetida(lista_ordenada)
print("\nMatriz generada:")
for fila in matriz_final:
    print(fila)
# Respuesta esperada:
# [2, 2, 2]
# [5, 5, 5]
# [9, 9, 9]
```

## 5. Ejemplos Clásicos de Funciones Simples

Aquí tienes algunos ejemplos cotidianos del uso de `def`. Nota cómo la mayoría se basan en recibir un argumento, procesarlo en un condicional `if`, y usar `return` para entregar la respuesta al sistema.

```python
# Ejemplo 1: Validar Acceso (if simple con retorno)
def validar_acceso(edad):
    if edad >= 18:
        return "Acceso Permitido"
    else:
        return "Acceso Denegado"

print("Usuario de 20 años:", validar_acceso(20))
print("Usuario de 15 años:", validar_acceso(15))


# Ejemplo 2: Calcular Descuento (Matemática simple y booleanos)
def aplicar_descuento(precio, es_estudiante):
    if es_estudiante == True:
        return precio * 0.8  # 20% de descuento
    else:
        return precio        # Precio normal sin descuento

print("\nPrecio $10000 (Estudiante):", aplicar_descuento(10000, True))
print("Precio $10000 (Normal):", aplicar_descuento(10000, False))


# Ejemplo 3: Filtrar una lista (Ciclos y arreglos)
def extraer_positivos(lista_numeros):
    positivos = []
    for num in lista_numeros:
        if num > 0:
            positivos.append(num)
    return positivos

print("\nLista original: [-5, 10, -2, 8, 0]")
print("Filtrada (Solo positivos):", extraer_positivos([-5, 10, -2, 8, 0]))
```
