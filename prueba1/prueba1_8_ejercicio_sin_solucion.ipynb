**Enunciado del Ejercicio:**

Cree un programa en el que la computadora genera un numero secreto y el usuario deba adivinarlo. El programa debe dar pistas para guiar al usuario hacia la respuesta correcta.

Requisitos:
1. Al iniciar, el programa debe generar un numero entero aleatorio entre 1 y 100. No debe mostrarlo.
2. El programa debe solicitar al usuario que ingrese un numero para intentar adivinar el numero secreto.
3. El programa debe repetirse hasta que el usuario adivine el numero. En cada intento, debe ocurrir lo siguiente:
a. Si el numero ingresado por el usuario es menor que el numero secreto, el programa debe mostrar por pantalla el mensaje: `El numero es muy bajo. ¡Intenta de nuevo!`.
b. Si el numero ingresado por el usuario es mayor que el numero secreto, el programa debe mostrar por pantalla el mensaje: `El numero es muy alto. ¡Intenta de nuevo!`.
4. Cuando el usuario finalmente adivine el numero secreto, el programa debe mostrar un mensaje de felicitacion, por ejemplo: `¡Felicidades! Adivinaste el numero secreto.`

### Paso 1: Separacion de la logica con 'if'
Antes de pensar en repetir el juego, debemos construir la logica central: ¿Como sabe la computadora si nuestra adivinanza es correcta o si nos equivocamos?

Para esto, imaginemos que el numero secreto ya esta definido (por ejemplo, es el 42). Usaremos una estructura `if / elif / else` para comparar nuestro intento con ese numero secreto y mostrar la pista correspondiente.

### Paso 2: La necesidad de usar el ciclo 'while'
En el paso anterior, si te equivocas, el programa simplemente termina. Para que sea un juego real, debemos dejar que el usuario siga intentando.

¿Podemos usar un ciclo `for`? **No.** El ciclo `for` sirve cuando sabemos exactamente cuantas veces se repetira algo (ej. `range(10)`). En este juego, un usuario con mucha suerte podria adivinar al primer intento, o podria equivocarse eternamente y necesitar 500 intentos. Como la cantidad de vueltas es **indefinida**, estamos obligados a usar un ciclo `while` (mientras).

El `while` funciona como un `if` que se repite infinitamente mientras una condicion sea verdadera. En nuestro caso, la condicion sera: **mientras el intento sea distinto (`!=`) al numero secreto.**

**La trampa del ciclo infinito:** Si entramos al ciclo `while` porque nos equivocamos, estamos OBLIGADOS a pedirle un nuevo intento al usuario dentro del ciclo. Si no lo hacemos, la variable `intento` nunca cambiara, la condicion de error seguira siendo verdadera, y el programa se quedara pegado repitiendo el mismo mensaje para siempre.

### Paso 3: Combinando el 'while' con nuestras pistas ('if')
Ahora uniremos el Paso 1 con el Paso 2.

Piensalo asi: **El ciclo `while` te atrapa, pero el `if` te guia.** 
- El `while` se encarga unicamente de verificar si te equivocaste y repetir todo el proceso.
- Dentro de esa repeticion, colocamos nuestro bloque `if / else` para analizar *por que* estas equivocado en esa vuelta especifica y darte una pista.

El flujo exacto es:
1. Pedimos el primer intento.
2. El `while` atrapa el error si son distintos.
3. El `if` interior evalua si el error fue por estar muy alto o muy bajo y lo imprime.
4. Al terminar de dar la pista, pedimos el nuevo intento para que el `while` pueda volver a empezar.

### Paso 4: Resolucion final uniendo la aleatoriedad
Ahora armaremos el juego definitivo.

Nuestra logica con la combinacion de `while` e `if` ya es perfecta. Lo unico que falta para cumplir todos los requisitos del enunciado es hacer que el numero secreto sea impredecible.

Para esto, importaremos la herramienta `random` y usaremos la funcion `randint(1, 100)` para que la computadora elija un numero al azar al comenzar el programa. ¡Intenta jugar y ganar en la celda de abajo!
