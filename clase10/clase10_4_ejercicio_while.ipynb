# Ejercicio Paso a Paso: El Limpiador de Ceros

En este documento resolveremos un ejercicio interactivo construyéndolo bloque a bloque. Aprenderemos a controlar un ciclo `while` utilizando **condiciones matemáticas estrictas**, sin depender de atajos o interrupciones forzadas como `break`.

## Enunciado

Crea un programa que pida un número `n` al usuario para crear un tablero de `n x n` lleno de ceros (`0`). 

Luego, utiliza un ciclo `while` para pedirle al usuario que ingrese coordenadas (fila y columna) para ir reemplazando esos ceros por unos (`1`). El programa debe mostrar el tablero actualizado en cada turno.

**Regla estricta:** El ciclo debe terminar **únicamente por sus propias condiciones** (no puedes usar `while True` ni la instrucción `break`). El programa finalizará si ocurre una de dos cosas:
1. Ya no queda ningún cero en el tablero.
2. El usuario ingresa un `-1` en la fila pedida.

Al finalizar por cualquiera de los dos motivos, se debe mostrar el tablero una última vez.

---
### Parte 1: Creación del Tablero
Comenzaremos creando nuestra matriz de `n x n` utilizando un doble ciclo `for` y rellenándola exclusivamente con ceros.

```python
n = int(input("Ingrese el tamaño del tablero (n): "))
tablero = []

# Doble for para crear una matriz de n filas por n columnas
for i in range(n):
    fila = []
    for j in range(n):
        fila.append(0)
    tablero.append(fila)

print("Tablero creado con éxito (oculto por ahora).")
```

### Razonamiento de la Parte 2

Ahora debemos estructurar nuestro ciclo `while`. Como tenemos estrictamente prohibido usar `break`, toda la responsabilidad de detener el ciclo recae en la línea de la condición.

El enunciado nos da dos motivos para salir:
1. **"Hasta que no quede ningún cero"**: Para saber esto sin usar herramientas avanzadas, simplemente calcularemos cuántos ceros hay al principio (`n * n`) y los guardaremos en una variable `ceros_restantes`. Cada vez que cambiemos un cero por un uno, le restaremos `1`. El ciclo debe seguir mientras `ceros_restantes > 0`.
2. **"Si se entrega un -1"**: Crearemos una variable `fila_ingresada = 0` (o cualquier número que no sea -1) antes del ciclo. El ciclo debe seguir mientras `fila_ingresada != -1`.

Juntando ambas con un operador lógico `and`, garantizamos un ciclo perfecto y seguro.

---
### Parte 1 + Parte 2: Creación y Condición Estricta

```python
n = int(input("Ingrese el tamaño del tablero (n): "))
tablero = []

for i in range(n):
    fila = []
    for j in range(n):
        fila.append(0)
    tablero.append(fila)

# Variables de control para la condición del while
ceros_restantes = n * n
fila_ingresada = 0

# Condición ESTRICTA: Ambas cosas deben ser ciertas para que siga preguntando
while ceros_restantes > 0 and fila_ingresada != -1:
    print(f"\nQuedan {ceros_restantes} ceros en el tablero.")
    
    fila_ingresada = int(input("Ingrese la fila (-1 para salir): "))
    
    # Evaluamos con if para no pedir la columna si el usuario acaba de escribir -1
    if fila_ingresada != -1:
        columna_ingresada = int(input("Ingrese la columna: "))
        print("Coordenadas recibidas.")
```

### Razonamiento de la Parte 3

Ya tenemos el motor de nuestro juego listo. Ahora solo nos falta la interfaz visual y la modificación de los datos.

Al principio del `while`, usaremos un doble `for` para imprimir la matriz actual en formato de cuadrícula.
Luego, cuando el usuario ingrese una coordenada válida, comprobaremos si efectivamente hay un `0` en esa posición. Si lo hay, lo cambiamos por un `1` y le restamos uno a nuestros `ceros_restantes`. 

Al salir del `while`, imprimiremos la matriz una última vez tal cual lo pide el enunciado.

---
### Parte 1 + Parte 2 + Parte 3: Código Completo y Final

```python
n = int(input("Ingrese el tamaño del tablero (n): "))
tablero = []

# --- 1. CREACIÓN DEL TABLERO ---
for i in range(n):
    fila = []
    for j in range(n):
        fila.append(0)
    tablero.append(fila)

# --- 2. VARIABLES DE CONDICIÓN ---
ceros_restantes = n * n
fila_ingresada = 0

# --- 3. CICLO PRINCIPAL ESTRICTO ---
while ceros_restantes > 0 and fila_ingresada != -1:
    
    # Visualización del tablero en cada turno
    print("\n--- TABLERO ACTUAL ---")
    for i in range(n):
        for j in range(n):
            print(tablero[i][j], end=" ")
        print() # Salto de línea por fila
        
    print(f"\nTe faltan {ceros_restantes} ceros por limpiar.")
    fila_ingresada = int(input("Ingrese la fila (-1 para rendirse): "))
    
    # Si el usuario no ingresó el código de salida, procedemos
    if fila_ingresada != -1:
        columna_ingresada = int(input("Ingrese la columna: "))
        
        # Medida de seguridad: Validamos que la coordenada exista en la matriz
        if fila_ingresada >= 0 and fila_ingresada < n and columna_ingresada >= 0 and columna_ingresada < n:
            
            # Validamos si todavía hay un cero en ese lugar
            if tablero[fila_ingresada][columna_ingresada] == 0:
                tablero[fila_ingresada][columna_ingresada] = 1
                ceros_restantes -= 1  # Actualizamos nuestra condición matemática!
            else:
                print("[*] Esa coordenada ya fue modificada antes.")
        else:
            print("[!] Cuidado: Coordenada fuera de los límites del tablero.")

# --- 4. CIERRE Y ÚLTIMA VISUALIZACIÓN ---
print("\n=======================")
print("   JUEGO TERMINADO")
print("=======================")

if ceros_restantes == 0:
    print("¡Felicidades, limpiaste todo el tablero!")
else:
    print("Te retiraste antes de tiempo. Así quedó tu avance:")

# Último print del tablero
for i in range(n):
    for j in range(n):
        print(tablero[i][j], end=" ")
    print()
```
