# Ejercicio Paso a Paso: Esculpiendo Letras con Índices

En este documento resolveremos un ejercicio de dibujo en consola dividiéndolo en pequeños fragmentos de código. Veremos cómo el programa evoluciona bloque a bloque, acumulando las funciones mientras explicamos la lógica matemática detrás de cada trazo.

## Enunciado

Crea un programa que dibuje figuras geométricas y letras en la pantalla utilizando el símbolo `#` en una cuadrícula (matriz) de 5x5.
El programa debe construirse paso a paso e imprimir de forma secuencial:
1. Un lienzo base (un cuadrado sólido de 5x5 lleno de `#`).
2. La letra **O**.
3. La letra **U**.
4. La letra **A**.
5. La letra **E**.
6. La letra **I**.

Todo debe lograrse manipulando las condiciones de los índices `i` (filas) y `j` (columnas).

---
### Parte 1: Base
Primero, vamos a crear nuestra herramienta principal: un doble ciclo `for` que dibuje una cuadrícula sólida de 5x5. Este será nuestro bloque de mármol del cual esculpiremos las letras.

### Razonamiento: ¿Cómo dibujamos la letra 'O'?

En la Parte 1, imprimimos el símbolo `#` de forma incondicional. Para dibujar letras, necesitamos usar condicionales `if / else`.

Si pensamos en la grilla, `i` representa las filas (de arriba a abajo) y `j` las columnas (de izquierda a derecha). 
Para dibujar una **O**, necesitamos imprimir `#` únicamente en los **bordes** del cuadrado. Es decir:
- En la primera fila (`i == 0`)
- En la última fila (`i == 4`)
- En la primera columna (`j == 0`)
- En la última columna (`j == 4`)

Si la coordenada coincide con alguno de estos bordes, imprimimos `#`. Si no, imprimimos un espacio vacío `" "` para dejar el centro hueco.

### Razonamiento: Transformando la 'O' en una 'U'

La letra **U** es visualmente idéntica a la letra **O**, con la única diferencia de que le falta el "techo". 

Para programarla, simplemente copiamos la lógica matemática de la letra 'O', pero **eliminamos la condición de la primera fila (`i == 0`)**. De esta forma, la parte superior quedará hueca.

### Razonamiento: Construyendo la 'A' y la 'E'

Cada letra requiere que pensemos en sus "trazos" geométricos en la grilla de 5x5:

**Para la Letra 'A':**
Necesitamos las paredes laterales (`j == 0` y `j == 4`), un techo (`i == 0`) y un cinturón central. En una grilla del 0 al 4, el medio exacto es la fila 2 (`i == 2`).

**Para la Letra 'E':**
Necesitamos una pared sólida en la izquierda (`j == 0`), y tres trazos horizontales: el techo (`i == 0`), el medio (`i == 2`) y el piso (`i == 4`).

### Razonamiento Final: La Letra 'I'

Para completar nuestras vocales base, esculpiremos la letra **I**.
Esta letra requiere líneas horizontales en el techo (`i == 0`) y en el piso (`i == 4`), pero a diferencia de las otras, necesita un pilar vertical que pase exactamente por el centro de las columnas.

En una cuadrícula de 5 columnas (índices 0, 1, 2, 3, 4), el centro perfecto es la columna 2. Por lo tanto, la condición para el pilar vertical será `j == 2`.
