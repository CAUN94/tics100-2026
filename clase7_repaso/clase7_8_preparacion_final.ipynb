## Nivel Facil: Calculo de Promedios y Puntajes

### 1. Enunciado
Un estudiante necesita un programa para calcular su promedio de notas. Ademas, la universidad usa ese promedio para calcular un 'Puntaje de Beca' mediante una formula simple. Tu programa debe calcular el promedio y luego el puntaje.
* Formula a utilizar: Puntaje = (Promedio * 10) + 50

### 2. Ejemplo de funcionamiento
```text
Ingrese la nota 1: 5.0
Ingrese la nota 2: 6.0
Ingrese la nota 3: 7.0
Promedio obtenido: 6.0
Puntaje de beca: 110.0
```

### 3. Analisis del Algoritmo
La logica principal consiste en pedir datos, acumular (sumar) las notas que el usuario ingresa, dividirlas por la cantidad de evaluaciones para obtener el promedio, y finalmente aplicar una operacion matematica simple al resultado.

### 4. Tareas / Puntos Clave
1. Solicitar las notas al usuario.
2. Sumar todas las notas.
3. Calcular el promedio dividiendo la suma por la cantidad de notas.
4. Calcular el puntaje usando la formula indicada.
5. Mostrar los resultados por pantalla.

### 5. Version A (Sin Ciclos)
Resolvamos el problema asumiendo que el estudiante siempre tendra exactamente 3 notas.

````python
# 1. Solicitamos exactamente 3 notas usando variables distintas
nota1 = float(input("Ingrese la nota 1: "))
nota2 = float(input("Ingrese la nota 2: "))
nota3 = float(input("Ingrese la nota 3: "))

# 2. Sumamos las notas
suma_notas = nota1 + nota2 + nota3

# 3. Calculamos el promedio (sabemos que son exactamente 3)
promedio = suma_notas / 3

# 4. Aplicamos la formula matematica para el puntaje
puntaje_beca = (promedio * 10) + 50

# 5. Mostramos los resultados
print(f"Promedio obtenido: {promedio}")
print(f"Puntaje de beca: {puntaje_beca}")
````

### 6. Transicion
La Version A funciona bien para 3 notas. Pero, ¿que pasa si un ramo tiene 10 notas? ¿O 15? Tendriamos que crear muchas variables repetidas. Para evitar esto, usaremos un ciclo `for` que nos permitira repetir la accion de 'pedir y sumar nota' la cantidad de veces que el estudiante decida.

### 7. Version B (Con Ciclos)
Esta version es dinamica y sirve para cualquier cantidad de evaluaciones.

````python
# Le preguntamos al estudiante cuantas notas tiene en total
cantidad_notas = int(input("¿Cuantas notas va a ingresar?: "))

# Esta variable actuara como un 'acumulador'. Comienza en cero.
suma_notas = 0.0

# Usamos un ciclo FOR que se repetira segun la cantidad indicada
for i in range(cantidad_notas):
    # Pedimos la nota (usamos i+1 para que empiece a contar desde 1 en la pantalla)
    nota_actual = float(input(f"Ingrese la nota {i+1}: "))
    
    # Acumulamos el valor ingresado en nuestra variable principal
    suma_notas = suma_notas + nota_actual 

# Una vez finalizado el ciclo, hacemos los calculos finales
promedio = suma_notas / cantidad_notas
puntaje_beca = (promedio * 10) + 50

# Imprimimos los resultados finales
print("--- Resumen Academico ---")
print(f"Promedio obtenido: {promedio}")
print(f"Puntaje de beca: {puntaje_beca}")
````

## Nivel Medio: Clasificacion de Rendimiento

### 1. Enunciado
Un profesor necesita clasificar el rendimiento de los examenes ingresados en el sistema.
* Si la nota es mayor a 6.0: Clasifica como 'Sobresaliente'.
* Si esta entre 4.0 y 6.0 (inclusive): Clasifica como 'Aprobado normal'.
* Si es menor a 4.0: Clasifica como 'Reprobado'.
* Regla critica: Si se ingresa una nota mayor a 7.0, es un error de tipeo. Debe emitir una 'ALERTA: Nota Invalida' y descartar el estado.

Tu programa debe evaluar las notas y reportar cuantas hay de cada tipo.

### 2. Ejemplo de funcionamiento
```text
Nota ingresada: 6.5
Estado: Sobresaliente.
```

### 3. Analisis del Algoritmo
Este problema exige categorizar un valor numerico usando estructuras condicionales (if, elif, else). Es vital ordenar las condiciones adecuadamente, evaluando primero los casos mas excepcionales o de error (como la alerta de > 7.0) para que no interfieran con las reglas normales.

### 4. Tareas / Puntos Clave
1. Obtener la nota.
2. Evaluar primero si es mayor a 7.0 (ALERTA).
3. Si no lo es, evaluar si es mayor a 6.0 (Sobresaliente).
4. Si no lo es, evaluar si es mayor o igual a 4.0 (Aprobado).
5. Si no cumple ninguna, obligatoriamente es menor a 4.0 (Reprobado).
6. Imprimir el estado.

### 5. Version A (Sin Ciclos)
Aqui evaluaremos la clasificacion para un solo examen de forma manual.

````python
# 1. Solicitamos la nota actual
nota = float(input("Ingrese la nota del examen: "))

# 2. Iniciamos el bloque de decisiones (siempre de lo mas restrictivo a lo general)
if nota > 7.0:
    print(f"ALERTA - Nota Invalida: {nota}. Supera el maximo permitido.")
elif nota > 6.0:
    print(f"Nota {nota} - Clasificacion: Sobresaliente.")
elif nota >= 4.0:
    print(f"Nota {nota} - Clasificacion: Aprobado normal.")
else:
    # Si no fue >7.0, ni >6.0, ni >=4.0... obligatoriamente es menor a 4.0
    print(f"Nota {nota} - Clasificacion: Reprobado.")
````

### 6. Transicion
El codigo anterior revisa solo un examen a la vez. En un curso real, el profesor necesita revisar muchos examenes y saber los totales. Para no repetir el codigo manualmente, usaremos un ciclo `for` que simulara la revision de 10 pruebas y contara cuantas clasificaciones hubo de cada tipo.

### 7. Version B (Con Ciclos)
Simulacion y clasificacion de 10 examenes.

````python
import random # Importamos la libreria para generar notas aleatorias de prueba

# Inicializamos variables contadoras en cero para el reporte final
total_sobresalientes = 0
total_aprobados = 0
total_reprobados = 0
alertas_error = 0

print("--- Iniciando Revision de 10 Examenes ---")

# Ciclo FOR que se repetira 10 veces
for alumno in range(1, 11):
    # Simulamos una nota al azar entre 2.0 y 8.0 (para forzar algunos errores)
    nota = round(random.uniform(2.0, 8.0), 1)
    
    # Evaluamos la nota y aumentamos el contador que corresponda
    if nota > 7.0:
        print(f"Examen {alumno}: {nota} -> ALERTA ERROR DE TIPEO.")
        alertas_error += 1
    elif nota > 6.0:
        print(f"Examen {alumno}: {nota} -> Sobresaliente.")
        total_sobresalientes += 1
    elif nota >= 4.0:
        print(f"Examen {alumno}: {nota} -> Aprobado normal.")
        total_aprobados += 1
    else:
        print(f"Examen {alumno}: {nota} -> Reprobado.")
        total_reprobados += 1

# Al terminar, imprimimos el reporte usando los contadores
print("\n--- Reporte Final del Curso ---")
print(f"Sobresalientes: {total_sobresalientes}")
print(f"Aprobados normales: {total_aprobados}")
print(f"Reprobados: {total_reprobados}")
print(f"Errores de tipeo detectados: {alertas_error}")
````

## Nivel Dificil: Control de Saldo en Tarjeta de Casino

### 1. Enunciado
Debes monitorear el porcentaje de saldo de una tarjeta de alimentacion universitaria. El programa pedira registrar el saldo restante despues de cada compra y finalizara cuando se ingrese un `0` (tarjeta vacia).
Existen **Reglas de Negocio estrictas**:
1. El saldo ingresado debe ser un porcentaje entre 0 y 100.
2. Como el dinero se gasta, el porcentaje nuevo no puede ser mayor al ingresado en el turno anterior.
Al finalizar, debes reportar el porcentaje del tiempo total que la tarjeta estuvo en: saldo alto (>80%), saldo medio (20% a 80%) y saldo bajo (<20%).

### 2. Ejemplo de funcionamiento
```text
Compra 1: 90
Compra 2: 95 -> Error: El saldo no puede aumentar.
Compra 2: 85
Compra 3: 0 -> Tarjeta vacia.
Tiempo con saldo alto: 50.0%
```

### 3. Analisis del Algoritmo
Este es un problema clasico de validacion de datos. No sabemos cuantas compras hara el usuario antes de llegar a cero, por lo que usar un ciclo `for` es imposible; necesitamos un `while`. Ademas, necesitamos una variable que guarde la 'memoria' del ultimo saldo valido para compararlo con el nuevo dato y asegurar que la tarjeta no se este 'recargando' sola.

### 4. Tareas / Puntos Clave
1. Establecer una variable para recordar el ultimo registro (iniciada en 101 para permitir el primer 100).
2. Entrar en un ciclo que pida datos continuamente.
3. Validar que el dato este entre 0 y 100.
4. Validar que el dato sea menor o igual al ultimo registro guardado.
5. Si es valido y distinto de 0: clasificar el saldo (alto, medio, bajo) y guardarlo como el nuevo 'ultimo registro'.
6. Si es 0: calcular los porcentajes finales basados en el total de ingresos y mostrar reporte.

### 5. Version A (Sin Ciclos)
Mira como la complejidad del codigo se vuelve inmanejable si intentaramos validar manualmente tan solo 3 compras.

````python
# Simulando un proceso tedioso sin ciclos
ultimo_saldo = 101 # Valor inicial ficticio

# INTENTO DE COMPRA 1
c1 = int(input("Compra 1 - Ingrese % saldo restante: "))
if c1 >= 0 and c1 <= 100:
    if c1 <= ultimo_saldo:
        print("Dato 1 aceptado.")
        ultimo_saldo = c1
    else:
        print("Error: El saldo no puede subir.")
else:
    print("Error: Fuera de rango.")

# INTENTO DE COMPRA 2 (Se vuelve muy confuso si el paso anterior fallo)
c2 = int(input("Compra 2 - Ingrese % saldo restante: "))
if c2 >= 0 and c2 <= 100:
    if c2 <= ultimo_saldo:
        print("Dato 2 aceptado.")
        ultimo_saldo = c2
    else:
        print("Error: El saldo no puede subir.")
else:
    print("Error: Fuera de rango.")

# Repetir esto de forma manual es propenso a errores y poco escalable.
````

### 6. Transicion
La Version A demuestra que validar datos manteniendo 'memoria' del estado anterior requiere una herramienta automatizada, sobre todo porque no sabemos cuando el usuario ingresara el `0`. El ciclo `while` (mientras) es la herramienta perfecta: 'Mientras el saldo no sea 0, sigue pidiendo, validando y contando'.

### 7. Version B (Con Ciclos)
Solucion solida aplicando el ciclo `while` y acumuladores estadisticos.

````python
print("--- Control de Tarjeta del Casino ---")
print("Ingrese '0' cuando la tarjeta se vacie.\n")

# Variables de estado
ultimo_saldo = 101  # Inicia sobre 100 para que el primer ingreso valido pase la regla
saldo_actual = -1   # Se inicia en algo distinto de 0 para que el while pueda comenzar

# Contadores estadisticos
compras_validas = 0
veces_medio = 0 # (20 - 80)
veces_alto = 0  # (> 80)
veces_bajo = 0  # (< 20)

# El ciclo se repetira MIENTRAS el saldo no sea 0
while saldo_actual != 0:
    
    # 1. Solicitamos el dato
    saldo_actual = int(input(f"Compra {compras_validas + 1} - Saldo restante (%): "))
    
    # Condicion de salida: Si es 0, dejamos que el ciclo continue para evaluar y terminar
    if saldo_actual == 0:
        print("\nTarjeta vacia. Generando reporte de uso...")
        continue # Vuelve al inicio del while, que ahora sera falso (saldo_actual != 0)
        
    # 2. Validacion de Rango (0 a 100)
    if saldo_actual < 0 or saldo_actual > 100:
        print(" -> ERROR: El porcentaje debe estar entre 0 y 100.")
        continue # Omite lo de abajo y vuelve a pedir
        
    # 3. Validacion Logica (no puede ser mayor al anterior)
    if saldo_actual > ultimo_saldo:
        print(f" -> ERROR: El saldo no puede aumentar. Tu ultimo saldo fue {ultimo_saldo}%.")
        continue # Omite lo de abajo y vuelve a pedir
        
    # --- Si la ejecucion llega a esta linea, el dato es 100% VALIDO ---
    
    compras_validas += 1         # Aumentamos el contador general
    ultimo_saldo = saldo_actual  # Actualizamos la memoria para la proxima vuelta
    
    # 4. Clasificacion del registro valido
    if saldo_actual > 80:
        veces_alto += 1
    elif saldo_actual >= 20: # Sabemos que es <= 80 por la regla de arriba
        veces_medio += 1
    else:
        veces_bajo += 1

# Fin del while. Verificamos que hayan datos para evitar dividir por cero
if compras_validas > 0:
    # Calculamos porcentajes
    pct_medio = (veces_medio / compras_validas) * 100
    pct_alto = (veces_alto / compras_validas) * 100
    pct_bajo = (veces_bajo / compras_validas) * 100
    
    print("\n--- Reporte Final de Uso de Tarjeta ---")
    print(f"Total de lecturas validas: {compras_validas}")
    print(f"% de tiempo con saldo alto (> 80%): {pct_alto}%")
    print(f"% de tiempo con saldo medio (20% - 80%): {pct_medio}%")
    print(f"% de tiempo con saldo bajo (< 20%): {pct_bajo}%")
else:
    print("No se registraron compras validas antes de vaciar la tarjeta.")
````
