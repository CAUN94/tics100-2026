## Tarea: Distancia en el Plano Cartesiano

**Descripción del problema:**
En geometría, la distancia entre dos puntos se calcula utilizando el Teorema de Pitágoras. Tu objetivo es construir un programa en Python que automatice este cálculo matemático para cualquier par de coordenadas. 

El programa debe recibir los datos numéricos, aplicar la lógica matemática correspondiente a la distancia euclidiana y entregar el resultado exacto. Queda a tu criterio deducir e investigar cómo estructurar la fórmula y cómo representar raíces cuadradas o exponentes utilizando la sintaxis de Python.

**Entrada de datos (Input):**
El programa debe solicitar al usuario cuatro valores por separado, en el siguiente orden estricto:
1. La coordenada X del primer punto.
2. La coordenada Y del primer punto.
3. La coordenada X del segundo punto.
4. La coordenada Y del segundo punto.

**Salida de datos (Output):**
El programa debe imprimir en pantalla un mensaje claro que incluya el resultado final del cálculo de la distancia.
