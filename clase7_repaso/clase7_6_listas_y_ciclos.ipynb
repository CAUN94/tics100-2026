# Simulación y Recorrido de Listas Paralelas (Avanzado)
## Enunciado del Ejercicio:
Las agencias de pluviometría de las ciudades de Santiago y Mendoza sospechan que el cambio climático podría estar alterando el patrón de precipitaciones de estas ciudades, haciéndolas más parecidas en términos de ocurrencias de lluvias. Para verificar esta sospecha ambas agencias registraron la existencia de lluvias para cada día del año 2017.

Escribe un programa que despliegue por pantalla el número de días en que llovió en ambas ciudades simultáneamente y el número de días en que no llovió en ambas ciudades simultáneamente. Emula el contenido de los arreglos generando el contenido de manera aleatoria.

### Análisis y Estrategia de Resolución:
El problema nos pide comparar dos conjuntos de datos (listas) índice por índice. La estrategia es:
1. Generar dos listas de 365 elementos (0s y 1s) usando un ciclo.
2. Crear contadores para los días donde llovió en ambas (`1` y `1`) y donde no llovió en ambas (`0` y `0`).
3. Recorrer las listas usando el mismo índice `i` para comparar los valores paralelos.

```python
import random

# 1. Generamos los datos simulados para 365 días
santiago = []
mendoza = []

for i in range(365):
    santiago.append(random.randint(0, 1))
    mendoza.append(random.randint(0, 1))

print(f"Muestra Santiago (primeros 10 días): {santiago[:10]}")
print(f"Muestra Mendoza (primeros 10 días): {mendoza[:10]}")
```

```python
# 2. Inicializamos los contadores de patrones
dias_ambas_lluvia = 0
dias_ninguna_lluvia = 0

# 3. Recorremos usando los índices para comparar paralelamente
for i in range(len(santiago)):
    # Verificamos si en ambas no llovió (0)
    if santiago[i] == 0 and mendoza[i] == 0:
        dias_ninguna_lluvia += 1
    # Verificamos si en ambas llovió (1)
    elif santiago[i] == 1 and mendoza[i] == 1:
        dias_ambas_lluvia += 1

print("Días con lluvia en ambas ciudades:", dias_ambas_lluvia)
print("Días sin lluvia en ambas ciudades:", dias_ninguna_lluvia)
```

# Título de la Materia: Algoritmos de Machine Learning Clásico (1-NN)
## Enunciado del Ejercicio:
Suponga que tiene dos vectores/listas de igual tamaño de largo 10 que representan puntos en el espacio cartesiano X-Y junto a su etiqueta (sano 0, enfermedad 1). Realice un programa que solicita ingresar un nuevo punto (coordenada X e Y), determine la coordenada más cercana entre los puntos observados y le asigne la etiqueta del punto más cercano.

### Análisis y Estrategia de Resolución:
El algoritmo de los K-Vecinos Más Cercanos (K-NN) requiere medir distancias.
1. Definimos el nuevo punto a predecir.
2. Calculamos la distancia euclidiana entre el nuevo punto y todos los puntos existentes usando un ciclo.
3. Encontramos la distancia mínima y guardamos su índice para saber qué etiqueta histórica asignarle.

```python
import math

# Datos históricos (coordenadas y etiquetas)
X = [2, 3, 6, 6, 4, 1, 9]
Y = [1, 7, 1, 8, 9, 1, 1]
etiqueta = [0, 0, 0, 1, 1, 1, 0]

# Ingreso del nuevo punto a clasificar (valores simulados para evitar input() en Jupyter)
nuevo_X = 5
nuevo_Y = 5
print(f"Clasificando el nuevo punto: ({nuevo_X}, {nuevo_Y})")
```

```python
# Lista para guardar todas las distancias calculadas
distancias = []

# Calculamos la distancia a cada punto histórico
for i in range(len(X)):
    # Fórmula de distancia euclidiana pitagórica
    dist = math.sqrt((X[i] - nuevo_X)**2 + (Y[i] - nuevo_Y)**2)
    distancias.append(dist)

print("Distancias calculadas:", [round(d, 2) for d in distancias])
```

```python
# Buscamos el índice de la distancia más pequeña (el vecino más cercano)
indice_menor = 0

for i in range(1, len(distancias)):
    if distancias[i] < distancias[indice_menor]:
        indice_menor = i

print("El punto histórico más cercano está en el índice:", indice_menor)
print("La etiqueta predicha para el nuevo punto es:", etiqueta[indice_menor])
```

# Título de la Materia: Lógica de Negocios y Ordenamiento de Listas
## Enunciado del Ejercicio:
La gran multitienda LeroMart tiene una oferta lleve 3 y pague 2 la cual consiste en poder comprar 3 elementos cualesquiera y solo pagar 2 (la tienda cobrará los 2 elementos más caros). Raspum desea comprar N elementos y decide entrar las veces que sean necesarias para pagar menos.
Realice un programa que calcule el valor mínimo que podría pagar.

### Análisis y Estrategia de Resolución:
Para minimizar el costo pagando 2 de cada 3 productos:
1. El usuario debe agrupar siempre los artículos más caros de 3 en 3.
2. Por lo tanto, debemos **ordenar la lista** de compras de mayor a menor (`reverse=True`).
3. Al recorrer la lista ordenada, usamos un contador. Por cada grupo de 3 elementos, sumamos al total los dos primeros y 'saltamos' el cobro del tercero.

```python
# Lista de precios simulada (7 productos)
precios = [2000, 4000, 1000, 2000, 4000, 3000, 2000]

# Ordenamos de mayor a menor para maximizar el descuento de los más caros
precios.sort(reverse=True)
print("Precios ordenados de mayor a menor:", precios)
```

```python
monto_pagar = 0
contador_descuento = 0

for i in range(len(precios)):
    contador_descuento += 1
    
    if contador_descuento == 3:
        # Llegamos al 3er producto del grupo, es GRATIS
        print(f"Producto de ${precios[i]} -> ¡GRATIS!")
        contador_descuento = 0 # Reiniciamos para el siguiente grupo
    else:
        # Pagamos el producto
        print(f"Producto de ${precios[i]} -> Pagado")
        monto_pagar += precios[i]

print("\nEl monto total a pagar optimizado es:", monto_pagar)
```

### Análisis y Estrategia de Resolución:
Debemos traducir fórmulas matemáticas estadísticas a código estructurado en Python.
1. Necesitamos calcular las sumatorias básicas: suma de X, suma de Y.
