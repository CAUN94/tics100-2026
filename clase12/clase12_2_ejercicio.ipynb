## 1. El Código Base

A continuación se presenta el código que ya construimos. Analízalo bien, porque tu trabajo será "desarmarlo" y meter sus diferentes partes dentro de funciones.

```python
import random

n = 8
bombas = 10
tablero = []

# 1. CREAR MATRIZ VACÍA
for i in range(n):
    fila = []
    for j in range(n):
        fila.append(0)
    tablero.append(fila)

# 2. COLOCAR BOMBAS AL AZAR
bombas_colocadas = 0
while bombas_colocadas < bombas:
    i_random = random.randint(0, n - 1)
    j_random = random.randint(0, n - 1)
    if tablero[i_random][j_random] != "#":
        tablero[i_random][j_random] = "#"
        bombas_colocadas += 1

# 3. CALCULAR VECINOS
for i in range(n):
    for j in range(n):
        if tablero[i][j] != "#":
            cant_bombas = 0
            for di in [-1, 0, 1]:
                for dj in [-1, 0, 1]:
                    if di != 0 or dj != 0:
                        if i + di >= 0 and i + di < n and j + dj >= 0 and j + dj < n:
                            if tablero[i + di][j + dj] == "#":
                                cant_bombas += 1
            tablero[i][j] = cant_bombas

# 4. MOSTRAR TABLERO
print("--- TABLERO RESUELTO ---")
for i in range(n):
    for j in range(n):
        print(tablero[i][j], end=" ")
    print()
```
