# Ejercicio: La Propagación del Gato (Autómata Celular Probabilístico)

Este ejercicio es una versión simplificada del famoso \"Juego de la Vida\". Aquí combinaremos todo lo que hemos aprendido: matrices, revisión de vecinos (con medidas de seguridad) y probabilidad con la librería `random`.

## Enunciado

Crea un simulador de propagación en una matriz de 7x7.

**Condiciones Iniciales:**
1. Crea un tablero lleno de ceros (`0`).
2. Coloca un único símbolo gato (`#`) exactamente en el centro de la matriz.

**Reglas de Interacción (Generaciones):**
El simulador debe avanzar turno a turno. En cada turno, el programa debe analizar la matriz:
* Si una celda ya es un `#`, se mantiene como `#`.
* Si una celda es un `0`, debes revisar sus 8 vecinos. Si **al menos uno** de sus vecinos es un `#`, existe riesgo de contagio.
* Si hay riesgo de contagio, se tira un dado al azar entre el 1 y el 3. Si el resultado es `1`, la celda se convierte en un `#`. Si no, sobrevive y se mantiene como `0` en este turno.

*Nota Clave:* Para evitar que el contagio se expanda por todo el tablero en un solo turno, los cambios deben guardarse en un **tablero nuevo**, y al final del turno, el tablero original se actualiza con el nuevo.

---
### Parte 1: El Tablero Inicial
Primero crearemos nuestra matriz de 7x7 y ubicaremos el punto de inicio en el centro.

```python
import random

n = 7
tablero = []

# Construcción del tablero de ceros
for i in range(n):
    fila = []
    for j in range(n):
        fila.append(0)
    tablero.append(fila)

# Posicionamos el gato inicial en el centro (índice 3 para un tablero de 7x7)
tablero[3][3] = "#"

print("--- GENERACIÓN 0 (INICIO) ---")
for i in range(n):
    for j in range(n):
        print(tablero[i][j], end=" ")
    print()
```

### Parte 2: La Lógica de Generación y Propagación

Ahora construiremos un ciclo que permita simular 5 turnos (o generaciones). 
En cada iteración crearemos un `tablero_nuevo` vacío. Revisaremos el `tablero` antiguo usando nuestra lógica de búsqueda de vecinos. Si se cumplen las condiciones probabilísticas, escribiremos un `#` en el `tablero_nuevo`.

```python
turnos_a_simular = 5

for turno in range(1, turnos_a_simular + 1):
    
    # 1. PREPARAMOS EL TABLERO DEL FUTURO (Vacío)
    tablero_nuevo = []
    for i in range(n):
        fila_vacia = []
        for j in range(n):
            fila_vacia.append(0)
        tablero_nuevo.append(fila_vacia)
        
    # 2. ANALIZAMOS EL TABLERO ACTUAL
    for i in range(n):
        for j in range(n):
            
            # Regla 1: Si ya es gato, sigue siendo gato en el futuro
            if tablero[i][j] == "#":
                tablero_nuevo[i][j] = "#"
                
            # Regla 2: Si es un 0, evaluamos su entorno
            else:
                riesgo_contagio = False
                
                # Revisamos los 8 vecinos
                for df in [-1, 0, 1]:
                    for dc in [-1, 0, 1]:
                        if df == 0 and dc == 0:
                            continue
                            
                        vi = i + df
                        vj = j + dc
                        
                        # Validamos que el vecino esté dentro del tablero
                        if vi >= 0 and vi < n and vj >= 0 and vj < n:
                            # Si detectamos un gato cerca, activamos la alerta
                            if tablero[vi][vj] == "#":
                                riesgo_contagio = True
                
                # Regla 3: Si hubo riesgo, tiramos los dados
                if riesgo_contagio == True:
                    azar = random.randint(1, 3)
                    if azar == 1:
                        tablero_nuevo[i][j] = "#" # Se contagia
                    else:
                        tablero_nuevo[i][j] = 0   # Se salva por ahora
                else:
                    tablero_nuevo[i][j] = 0 # Estaba a salvo, sigue siendo 0
                    
    # 3. ACTUALIZAMOS EL PRESENTE CON EL FUTURO
    # Reemplazamos nuestra matriz antigua por la nueva que acabamos de calcular
    tablero = tablero_nuevo
    
    # 4. MOSTRAMOS EL RESULTADO DEL TURNO
    print(f"\n--- GENERACIÓN {turno} ---")
    for i in range(n):
        for j in range(n):
            print(tablero[i][j], end=" ")
        print()
```

### Explicación del Fenómeno Visual

Si ejecutas el código varias veces, notarás que la mancha de `#` nunca crece de la misma manera. A veces se expande rápidamente hacia la izquierda, a veces crece como un cuadrado sólido, y otras veces forma patrones irregulares.
