### Generador con Rango Seguro

**Enunciado:**
Cree un programa que pida al usuario tres variables: `a`, `b` y la cantidad `n`. El código debe generar una lista con `n` valores aleatorios entre `a` y `b`. Sin embargo, el código tiene que detectar si el usuario se equivocó e ingresó el valor mayor en `a` y el menor en `b` (por ejemplo, entre 10 y 2), y en ese caso, debe invertirlos automáticamente para que la función de aleatoriedad no arroje error. Finalmente, muestre la lista por pantalla.

**Ejemplo de ejecución:**
```text
Ingrese límite A: 15
Ingrese límite B: 5
Ingrese cantidad N: 4
Se invirtieron los límites automáticamente a: 5 y 15
Lista generada: [6, 12, 5, 14]
```

````python
# --- RESPUESTA ---
import random

# a = int(input("Ingrese límite A: "))
# b = int(input("Ingrese límite B: "))
# n = int(input("Ingrese cantidad N: "))
a = 15   # Simulamos input del usuario
b = 5    # Simulamos input del usuario
n = 4    # Simulamos input del usuario

# Verificamos si están invertidos y los arreglamos
if a > b:
    temp = a
    a = b
    b = temp
    print(f"Se invirtieron los límites automáticamente a: {a} y {b}")

lista_azar = []
for i in range(n):
    lista_azar.append(random.randint(a, b))

print(f"Lista generada: {lista_azar}")
````

### Limpieza de Errores por Promedio

**Enunciado:**
Cree una lista de `n` (valor escrito directo en el código) valores aleatorios entre -4 y 10. Vamos a revisar todos los números creados y asumir que todos los valores negativos o ceros son "errores de la muestra". Modifique la lista para que esos errores sean reemplazados por el promedio exacto de los números válidos (es decir, el promedio excluyendo los errores).

**Ejemplo de ejecución:**
```text
Muestra original: [5, -2, 8, 0, 2]
El promedio de los datos válidos es: 5.0
Muestra corregida: [5, 5.0, 8, 5.0, 2]
```

````python
# --- RESPUESTA ---
import random

n = 5
muestra = []
for i in range(n):
    muestra.append(random.randint(-4, 10))
    
print(f"Muestra original: {muestra}")

# 1. Calcular el promedio solo de los números válidos (> 0)
suma_validos = 0
cant_validos = 0

for num in muestra:
    if num > 0:
        suma_validos += num
        cant_validos += 1

if cant_validos > 0:
    promedio = suma_validos / cant_validos
else:
    promedio = 0

print(f"El promedio de los datos válidos es: {promedio}")

# 2. Reemplazar los errores por el promedio calculado
for i in range(len(muestra)):
    if muestra[i] <= 0:
        muestra[i] = promedio

print(f"Muestra corregida: {muestra}")
````

### Posiciones de los Impares

**Enunciado:**
Haga un código que genere una lista de `n` números (el `n` está dado por el usuario) con números aleatorios entre 1 y 10. El programa debe recorrer la lista y devolver por pantalla exclusivamente la **posición (índice)** de todos los números que sean impares.

**Ejemplo de ejecución:**
```text
Lista generada: [4, 7, 2, 9, 8]
Número impar encontrado en la posición: 1
Número impar encontrado en la posición: 3
```

````python
# --- RESPUESTA ---
import random

# n = int(input("Ingrese la cantidad de números: "))
n = 5 # Simulamos el input

lista = []
for i in range(n):
    lista.append(random.randint(1, 10))

print(f"Lista generada: {lista}")

for i in range(len(lista)):
    # Evaluamos si el número de esa posición es impar
    if lista[i] % 2 != 0:
        print(f"Número impar encontrado en la posición: {i}")
````

### Buscador de Números Primos

**Enunciado:**
Haga un código que genere una lista de `n` números (el `n` está dado por el usuario) con números aleatorios entre 2 y 10. Por pantalla se debe devolver la posición de todos los números que son primos.

**Ejemplo de ejecución:**
```text
Lista generada: [1, 5, 8, 3, 9]
Primo encontrado en la posición: 1 (Número 5)
Primo encontrado en la posición: 3 (Número 3)
```

````python
# --- RESPUESTA ---
import random

# n = int(input("Ingrese la cantidad de números: "))
n = 5

lista = []
for i in range(n):
    lista.append(random.randint(1, 10))

print(f"Lista generada: {lista}")

for i in range(len(lista)):
    
    primo = True
    for j in range(2, lista[i]):
        if lista[i] % j == 0:
            primo = False

    if primo == True and lista[i] > 1:
        print(f"Número primo encontrado en la posición: {i}")
````

### Valles Topológicos

**Enunciado:**
Haga un código que genere una lista de `n` números aleatorios entre 1 y 10 (con `n` ingresado por el usuario), simulando un perfil topológico. El programa debe recorrer la lista y devolver la posición y el valor de todos los números que sean "valles". Un valle es cualquier posición interior donde los valores que están a su izquierda y a su derecha son **estrictamente mayores** al número del centro (ejemplo: en `[5, 1, 3]`, el 1 es un valle).

**Ejemplo de ejecución:**
```text
Topología: [8, 3, 7, 2, 5, 9]
Valle detectado en posición 1: Su valor es 3
Valle detectado en posición 3: Su valor es 2
```

````python
# --- RESPUESTA ---
import random

# n = int(input("Ingrese tamaño del terreno: "))
n = 6

terreno = []
for i in range(n):
    terreno.append(random.randint(1, 10))

print(f"Topología: {terreno}")

# Recorremos desde la posición 1 hasta la penúltima (n-2) para tener vecinos a ambos lados
for i in range(1, len(terreno) - 1):
    if terreno[i] < terreno[i - 1] and terreno[i] < terreno[i + 1]:
        print(f"Valle detectado en posición {i}: Su valor es {terreno[i]}")
````

### Rotación a la Izquierda

**Enunciado:**
Haga un código que genere una lista de `n` números aleatorios entre 1 y 5. Luego, modifique la lista de forma que todos los elementos se desplacen una posición hacia atrás (hacia la izquierda): el que está en `i` pasa a estar en `i - 1`. El elemento que estaba en la primera posición (la posición 0) debe pasar a la última posición de la lista.

**Ejemplo de ejecución:**
```text
Lista original: [3, 5, 1, 4, 2]
Lista desplazada: [5, 1, 4, 2, 3]
```

````python
# --- RESPUESTA ---
import random

# n = int(input("Ingrese tamaño de la lista: "))
n = 5
lista = []
for i in range(n):
    lista.append(random.randint(1, 5))

print(f"Lista original: {lista}")

# Guardamos el primer elemento antes de que sea sobreescrito
primer_elemento = lista[0]

# Desplazamos todos los elementos un índice hacia la izquierda
for i in range(1, len(lista)):
    lista[i - 1] = lista[i]

# Asignamos el que guardamos al final de la lista
lista[-1] = primer_elemento

print(f"Lista desplazada: {lista}")
````

### Buscador de Menú y Control de Stock

**Enunciado:**
Cree un código que tenga tres listas paralelas: artículos, precios y cantidad de stock. El programa debe imprimir por pantalla un menú numerado (1. Pan, 2. Cebolla, etc.). Luego, solicite al usuario que ingrese el número del producto que busca. 
Utilizando ese número para hallar el índice correcto, muestre el precio del artículo por pantalla. Si el stock de ese artículo es `0`, debe agregar un aviso indicando que no queda stock.

**Ejemplo de ejecución:**
```text
Menú de Artículos:
1) Pan
2) Cebolla
3) Huevos
Ingrese el número del artículo que busca: 2
Ha seleccionado: Cebolla
Precio: $800
¡Aviso! Actualmente no queda stock de este producto.
```

````python
# --- RESPUESTA ---
articulos = ["Pan", "Cebolla", "Huevos"]
precios = [1500, 800, 2500]
stock = [10, 0, 5]

print("Menú de Artículos:")
for i in range(len(articulos)):
    # Imprimimos i + 1 para que el menú empiece en 1 y no en 0
    print(f"{i + 1}) {articulos[i]}")

# seleccion = int(input("Ingrese el número del artículo que busca: "))
seleccion = 2 # Simulamos el input

# Convertimos la selección de vuelta a un índice de Python restando 1
indice = seleccion - 1

print(f"\nHa seleccionado: {articulos[indice]}")
print(f"Precio: ${precios[indice]}")

if stock[indice] == 0:
    print("¡Aviso! Actualmente no queda stock de este producto.")
````

### Bonificación por Asistencia

**Enunciado:**
Dadas tres listas paralelas: estudiantes, notas actuales y porcentajes de asistencia. Modifique directamente la lista de calificaciones aumentando un 10% la nota de aquellos estudiantes que tengan una asistencia del 80% o superior.

**Ejemplo de ejecución:**
```text
Notas originales: [4.0, 5.0, 3.5, 6.0]
El alumno Ana recibe bonificación. Sube a 4.4
El alumno Pedro recibe bonificación. Sube a 3.85
Notas finales: [4.4, 5.0, 3.85, 6.0]
```

````python
# --- RESPUESTA ---
estudiantes = ["Ana", "Luis", "Pedro", "Sofía"]
notas = [4.0, 5.0, 3.5, 6.0]
asistencia = [85, 60, 90, 75]

print(f"Notas originales: {notas}")

for i in range(len(estudiantes)):
    if asistencia[i] >= 80:
        # Aumentamos un 10% multiplicando por 1.10
        notas[i] = notas[i] * 1.10
        print(f"El alumno {estudiantes[i]} recibe bonificación. Sube a {notas[i]}")

print(f"Notas finales: {notas}")
````

### Explode / Implode: Arreglando Correos Institucionales

**Explicación de Conceptos:**
En lenguajes de programación antiguos, existían comandos llamados `explode` (para separar textos). En Python, estas herramientas son mucho más modernas y se conocen como:
* **`.split("separador")`**: Toma un texto y lo corta en pedazos cada vez que encuentra el separador, devolviendo una lista.

```
mail = "abc@alumos.uai.cl"

lista_mail = mail.split("@")
print(f"Lista generada: {lista_mail}") #["abc", "alumos.uai.cl"]
print(f"Usuario: {lista_mail[0]}") # "abc"
print(f"Dominio: {lista_mail[1]}") # "alumos.uai.cl

nuevo_dominio = "alumnos.uai.cl"
nuevo_mail = lista_mail[0] + "@" + nuevo_dominio
print(f"Nuevo mail con dominio corregido: {nuevo_mail}")
```

**Enunciado:**
Dada una lista de correos electrónicos institucionales, algunos de ellos vienen con un error de tipeo en el dominio (ej: `aluno.uai.cl`). Utilice `.split()` para separar cada correo por la arroba (`@`). Si detecta que la parte del dominio es igual a `"aluno.uai.cl"`, modifíquela a `"alumnos.uai.cl"`. Finalmente, utilice `.join()` para reconstruir el correo y actualizar la lista.

**Ejemplo de ejecución:**
```text
Correos con errores: ['juan@alumnos.uai.cl', 'maria@aluno.uai.cl', 'pedro@aluno.uai.cl']
Correos arreglados: ['juan@alumnos.uai.cl', 'maria@alumnos.uai.cl', 'pedro@alumnos.uai.cl']
```

````python
mail = "abc@alumos.uai.cl"

lista_mail = mail.split("@")
print(f"Lista generada: {lista_mail}") #["abc", "alumos.uai.cl"]
print(f"Usuario: {lista_mail[0]}") # "abc"
print(f"Dominio: {lista_mail[1]}") # "alumos.uai.cl

nuevo_dominio = "alumnos.uai.cl"
nuevo_mail = lista_mail[0] + "@" + nuevo_dominio
print(f"Nuevo mail con dominio corregido: {nuevo_mail}")
````

````python
# --- RESPUESTA ---
correos = ["juan@alumnos.uai.cl", "maria@aluno.uai.cl", "pedro@aluno.uai.cl"]
print(f"Correos con errores: {correos}")

for i in range(len(correos)):

    partes = correos[i].split("@")
    
    # Evaluamos la segunda parte (el dominio)
    if partes[1] == "aluno.uai.cl":
        partes[1] = "alumnos.uai.cl"
        
        correos[i] = partes[0] + "@" + partes[1]
        
print(f"Correos arreglados: {correos}")
````

### Búsqueda de Extremos Temporales

**Enunciado:**
Tiene dos listas paralelas: los nombres de competidores y el tiempo que demoraron en finalizar una carrera (en segundos). El ganador es quien tiene el **menor** tiempo, y el último es quien tiene el **mayor** tiempo. Recorra las listas por índice para encontrar quién ganó, e imprima su nombre y la diferencia exacta de segundos que le sacó al último lugar.

**Ejemplo de ejecución:**
```text
El ganador fue Mateo con 120 segundos.
Mateo le sacó una diferencia de 65 segundos al último lugar.
```

````python
# --- RESPUESTA ---
competidores = ["Lucas", "Mateo", "Simón", "Tomás"]
tiempos = [145, 120, 185, 130]

pos_menor = 0
pos_mayor = 0

for i in range(1, len(tiempos)):
    if tiempos[i] < tiempos[pos_menor]:
        pos_menor = i
        
    if tiempos[i] > tiempos[pos_mayor]:
        pos_mayor = i

diferencia = tiempos[pos_mayor] - tiempos[pos_menor]

print(f"El ganador fue {competidores[pos_menor]} con {tiempos[pos_menor]} segundos.")
print(f"{competidores[pos_menor]} le sacó una diferencia de {diferencia} segundos al último lugar.")
````
