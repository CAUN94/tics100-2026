## 1. Guerra de abejas y hormigas

Hay una guerra entre abejas y hormigas, y las abejas lanzarán "bombas de azúcar" en el territorio de las hormigas. Cuando una bomba de azúcar explota, el azúcar se esparce y se acumula en sus alrededores, formando cerros de azúcar.

Se le pide desarrollar un programa secuencial que simule este ataque siguiendo estos pasos:

1. **Ingreso de datos:** Solicite al usuario ingresar el tamaño del tablero y el número de bombas que serán lanzadas. (Asuma que el tamaño siempre será 3 o mayor).
2. **Creación del campo:** Construya una matriz cuadrada de ceros del tamaño indicado.
3. **Lanzamiento de bombas:** Utilice un ciclo para simular la caída. En cada repetición:
    * Genere coordenadas aleatorias (fila y columna) para el impacto. Las bombas **no pueden caer en los bordes** del tablero.
    * Aumente en 1 el valor de azúcar en la casilla de impacto.
    * Haga que el azúcar se esparza, incrementando en 1 el valor de sus **cuatro vecinos directos** (arriba, abajo, izquierda y derecha).
4. **Visualización:** Imprima en pantalla el estado final del tablero.

```python
import random

# Ingreso de datos (Comentado para ejecución automática)
# tamano = int(input("Ingrese el tamaño del tablero: "))
# num_bombas = int(input("Ingrese el número de bombas: "))
tamano = 4
num_bombas = 2

print(f"Tamaño del tablero: {tamano}")
print(f"Número de bombas: {num_bombas}\n")

# Creación de la matriz de ceros
tablero = []
for i in range(tamano):
    fila = []
    for j in range(tamano):
        fila.append(0)
    tablero.append(fila)

# Lanzamiento de bombas
for b in range(num_bombas):
    # Coordenadas aleatorias evitando los bordes (del 1 al tamano-2)
    f = random.randint(1, tamano - 2)
    c = random.randint(1, tamano - 2)
    
    # Impacto directo
    tablero[f][c] += 1
    
    # Expansión a vecinos directos
    tablero[f-1][c] += 1 # Arriba
    tablero[f+1][c] += 1 # Abajo
    tablero[f][c-1] += 1 # Izquierda
    tablero[f][c+1] += 1 # Derecha

print("Estado final del tablero:")
for i in range(tamano):
    print(tablero[i])
```

## 2. Cumbres Geográficas

Dada una matriz que representa la altitud de una zona geográfica, se le pide escribir un programa que calcule el número de cumbres existentes y determine cuál es la cumbre más alta de todas.

Una cumbre corresponde a un punto que es **estrictamente más alto que todos sus 8 vecinos** (arriba, abajo, lados y diagonales). Asuma que los bordes de la matriz no pueden ser cumbres, por lo que su programa solo debe evaluar las celdas interiores.

```python
tablero = [
    [42, 73, 51, 14, 43, 15, 35, 24, 20, 65],
    [65, 81, 89, 32, 74, 92, 50, 12, 94, 22],
    [ 2,  9, 70, 92, 61,  2, 17, 22, 65,  1],
    [65, 28, 57, 91, 67, 41,  8, 99, 99,  4],
    [16, 78, 57, 89, 50, 92, 24, 30, 92, 73],
    [14, 32, 74, 48, 40, 94, 52, 21,  8, 80]
]

cantidad_cumbres = 0
cumbre_mas_alta = 0

# Recorremos solo las celdas interiores
for i in range(1, len(tablero) - 1):
    for j in range(1, len(tablero[0]) - 1):
        actual = tablero[i][j]
        
        # Verificamos si es estrictamente mayor a sus 8 vecinos
        if (actual > tablero[i-1][j-1] and
            actual > tablero[i-1][j] and
            actual > tablero[i-1][j+1] and
            actual > tablero[i][j-1] and
            actual > tablero[i][j+1] and
            actual > tablero[i+1][j-1] and
            actual > tablero[i+1][j] and
            actual > tablero[i+1][j+1]):
            
            cantidad_cumbres += 1
            
            # Actualizamos la cumbre máxima si corresponde
            if actual > cumbre_mas_alta:
                cumbre_mas_alta = actual

print(f"Cantidad de cumbres encontradas: {cantidad_cumbres}")
print(f"La cumbre más alta tiene una altura de {cumbre_mas_alta}.")
```

## 3. Filtro de Ruido en Imágenes

Una imagen en escala de grises se representa como una matriz con valores entre 0 y 255. Se considera que una imagen tiene **ruido** cuando el valor de una celda es `0` y, al mismo tiempo, **todos sus 8 vecinos son distintos de `0`**.

Escriba un programa que reciba una imagen precargada e itere sobre sus celdas interiores para:
1. Identificar y contar cuántos píxeles con ruido existen.
2. Corregir el ruido reemplazando el valor `0` de la celda ruidosa por el **promedio (división entera `//`)** de sus 8 vecinos.
3. Imprimir el tablero corregido.

```python
imagenBlancoNegro = [
    [10, 73, 51, 14, 43, 15, 35, 24, 20, 65],
    [65, 81, 89, 00, 00, 92, 50, 12, 94, 22],
    [ 2,  9, 70, 10, 61,  2, 17, 22, 65,  1],
    [65, 28, 57, 00, 67, 41,  8, 99, 99,  4],
    [16, 00, 57, 89, 50, 92, 24, 30, 00, 73],
    [14, 32, 74, 48, 40, 94, 52, 21,  8, 80]
]

ruidos_encontrados = 0

for i in range(1, len(imagenBlancoNegro) - 1):
    for j in range(1, len(imagenBlancoNegro[0]) - 1):
        
        # Si la celda es 0, evaluamos sus 8 vecinos
        if imagenBlancoNegro[i][j] == 0:
            
            if (imagenBlancoNegro[i-1][j-1] != 0 and
                imagenBlancoNegro[i-1][j] != 0 and
                imagenBlancoNegro[i-1][j+1] != 0 and
                imagenBlancoNegro[i][j-1] != 0 and
                imagenBlancoNegro[i][j+1] != 0 and
                imagenBlancoNegro[i+1][j-1] != 0 and
                imagenBlancoNegro[i+1][j] != 0 and
                imagenBlancoNegro[i+1][j+1] != 0):
                
                # Es un píxel de ruido confirmado
                ruidos_encontrados += 1
                
                # Calculamos el promedio de los 8 vecinos
                suma_vecinos = (imagenBlancoNegro[i-1][j-1] + 
                                imagenBlancoNegro[i-1][j] + 
                                imagenBlancoNegro[i-1][j+1] +
                                imagenBlancoNegro[i][j-1] + 
                                imagenBlancoNegro[i][j+1] +
                                imagenBlancoNegro[i+1][j-1] + 
                                imagenBlancoNegro[i+1][j] + 
                                imagenBlancoNegro[i+1][j+1])
                
                promedio = suma_vecinos // 8
                
                # Corregimos el ruido
                imagenBlancoNegro[i][j] = promedio

print(f"Se encontraron y corrigieron {ruidos_encontrados} píxeles ruidosos.\n")
print("Imagen corregida:")
for i in range(len(imagenBlancoNegro)):
    for j in range(len(imagenBlancoNegro[0])):
        # Formateamos con tabulación para que la matriz se vea ordenada
        print(f"{imagenBlancoNegro[i][j]}\t", end="")
    print()
```

## 4. Detección de Objetos Sospechosos

Analice automáticamente la existencia de objetos sospechosos en una imagen de alta resolución. En la imagen, un objeto sospechoso aparece como **una cruz con valores `255` en sus cuatro extremos** (vecinos directos: arriba, abajo, izquierda y derecha).

1. Cree una matriz (lista anidada vacía) llamada `matrizObjetos`.
2. Cada vez que encuentre un objeto, agregue su posición (la del centro de la cruz) como una sub-lista `[fila, columna]` a esta matriz.
3. Imprima cuántos objetos se encontraron y la `matrizObjetos` resultante.

```python
imagenNorte = [
    [ 10, 255,  51,  14,  43,  15,  35,  24,  20,  65],
    [255,  81, 255,  00,  00,  92,  50,  12,  94,  22],
    [ 02, 255,  70,  10,  61,  02,  17, 255,  65,  01],
    [ 65,  28,  57,  00,  67,  41, 255,  99, 255,  04],
    [ 16,  00,  57,  89,  50,  92,  24, 255,  00,  73],
    [ 14,  32,  74,  48,  40,  94,  52,  21,   8,  80]
]

matrizObjetos = []

# Recorremos las celdas interiores buscando el patrón de cruz
for i in range(1, len(imagenNorte) - 1):
    for j in range(1, len(imagenNorte[0]) - 1):
        
        # Verificamos si los 4 vecinos directos valen 255
        if (imagenNorte[i-1][j] == 255 and  # Arriba
            imagenNorte[i+1][j] == 255 and  # Abajo
            imagenNorte[i][j-1] == 255 and  # Izquierda
            imagenNorte[i][j+1] == 255):    # Derecha
            
            # Objeto detectado, guardamos la coordenada [fila, columna]
            posicion = [i, j]
            matrizObjetos.append(posicion)

print(f"Se encontraron {len(matrizObjetos)} objetos sospechosos.")
print(f"Matriz de posiciones: {matrizObjetos}")
```
