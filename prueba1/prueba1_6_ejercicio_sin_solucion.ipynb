**Enunciado del Ejercicio:**

Se requiere desarrollar un programa que simule el proceso de monitoreo de temperatura en un sistema de refrigeración industrial. El sistema debe registrar la temperatura interna de un compartimento cada hora durante un periodo de 12 horas y tomar decisiones automáticas para la activación o desactivación del sistema de refrigeración según las condiciones del entorno.

El programa debe cumplir con las siguientes funciones:
1. Monitorear la temperatura del compartimento cada hora durante un período de 12 horas.
2. Si la temperatura supera los 30 C, el sistema de refrigeración debe activarse.
3. Si la temperatura está entre 20 C y 30 C, el sistema de refrigeración debe estar en modo de ahorro de energía.
4. Si la temperatura es inferior a 20 C, el sistema de refrigeración debe estar apagado.
5. El sistema debe generar un reporte al final del periodo de 12 horas, el cual debe incluir:
a. La cantidad de veces que se activó el sistema de refrigeración.
b. La cantidad de veces que el sistema estuvo en modo de ahorro de energía.
c. La cantidad de veces que el sistema estuvo apagado.

### Paso 1: ¿Como cuenta una computadora? (Los Contadores)
En este ejercicio nos piden entregar un reporte con la "cantidad de veces" que ocurrio un evento (activado, ahorro, apagado). 

Para que una computadora pueda contar, necesitamos crear variables que actuen como cajas vacias. Al principio, antes de medir nada, estas cajas valen cero. A estas variables se les llama **contadores**.

### Paso 2: Tomar la primera decision y sumar
Ahora debemos conectar la medicion de la temperatura con nuestro contador. Veamos solo el primer caso: "Si la temperatura supera los 30 C, el sistema debe activarse".

Para contar, la logica es: tomamos el valor actual de la variable, le sumamos 1, y lo volvemos a guardar en la misma variable (`variable = variable + 1`).

### Paso 3: Agregar las demas alternativas (elif y else)
El sistema no tiene un solo estado, tiene tres. Para que la computadora decida correctamente a que contador sumarle un punto, usamos un bloque completo de condicionales con `if`, `elif` y `else`.

Evaluemos una temperatura diferente (25.5 C) para ver como entra en la categoria de "Ahorro de energia".

### Paso 4: Automatizar las temperaturas con 'random'
En un sistema real, un sensor nos daria la temperatura. Como estamos simulando, podemos hacer que Python invente temperaturas por nosotros usando la libreria `random`.

Importaremos esta herramienta y usaremos su funcion `randint(inicio, fin)` para que nos entregue un numero entero aleatorio dentro de un rango que nosotros le digamos (por ejemplo, entre 10 y 45 grados).

### Paso 5: El comportamiento acumulativo a traves del tiempo
Antes de automatizar las 12 horas, veamos por que es necesario un ciclo. Si tuvieramos que escribir el codigo manualmente para 3 horas, usando nuestras temperaturas simuladas, tendriamos que repetir el bloque condicional del paso 3 una y otra vez.

Lo importante aqui es notar como las variables contadoras **acumulan memoria**. Si en la Hora 1 el ahorro suma 1, ese valor se guarda. Si en la Hora 2 ocurre lo mismo, toma el 1 guardado y lo convierte en 2.

### Paso 6: El ciclo for y por que usamos "hora + 1"
El monitoreo dura 12 horas, lo que significa que deberiamos evaluar 12 temperaturas. En lugar de escribir manualmente todos los bloques (como hicimos en el Paso 5), automatizamos esto usando un ciclo `for hora in range(12)`.

**¿Por que sumar 1?** La instruccion `range(12)` genera exactamente 12 vueltas, pero en programacion siempre comenzamos a contar desde el cero (generando numeros del 0 al 11). 

Como para una persona es extraño leer "Monitoreando la Hora 0", dentro de nuestro ciclo usaremos el truco de escribir `hora + 1` en los mensajes de texto (prints e inputs). Esto no altera el ciclo, solo mejora como se muestra en pantalla para que el humano vea visualmente "Hora 1" hasta "Hora 12".

### Paso 7: Solucion final uniendo conceptos
Ahora construiremos el sistema de monitoreo completo.

Definiremos los tres contadores en cero. Usaremos un ciclo `for` de 12 iteraciones para solicitar las temperaturas simuladas usando `random`. En cada vuelta, insertaremos nuestro bloque condicional e iremos sumando puntos a los contadores segun corresponda.

Al finalizar, imprimiremos el reporte.
