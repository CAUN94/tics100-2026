En este cuaderno construiremos, paso a paso, un sistema para registrar y evaluar las calificaciones de un estudiante. Iremos agregando complejidad de forma progresiva.

### Paso 1: Mensaje de bienvenida

```python
print("--- Bienvenido al Sistema de Registro Academico ---")
```

### Paso 2: Pedir nombre y mensaje de bienvenida
El sistema debe ser interactivo. Pide al usuario que ingrese el nombre del estudiante y luego muestra un saludo personalizado que incluya ese nombre.

**Hint:** Recuerda que la función `input("Mensaje: ")` se utiliza para solicitar texto al usuario y puedes guardarlo en una variable.

```python
nombre = input("Ingrese el nombre del estudiante: ")
print("Bienvenido al sistema, estudiante:", nombre)
```

### Paso 3: Pedir nota y mostrarla
Ahora agregaremos la calificación. Pide al usuario que ingrese la nota obtenida por el estudiante y muéstrala por pantalla junto a su nombre.

**Hint:** Como las notas son números con decimales, debes transformar el texto capturado. Utiliza `float(input("Mensaje: "))` para números reales o `int(input("Mensaje: "))` si fueran enteros.

```python
nombre = input("Ingrese el nombre del estudiante: ")
nota = float(input("Ingrese la nota obtenida: "))

print("Estudiante registrado:", nombre)
print("Calificacion ingresada:", nota)
```

### Paso 4: Pedir nota y decir si aprobo
Agreguemos inteligencia al sistema. Copia tu codigo anterior y agrega una validacion: si la nota es mayor o igual a 4.0, el sistema debe imprimir el mensaje "Estado: Aprobado".

**Hint:** Utiliza la estructura `if condicion:` recordando respetar la indentacion (espacio hacia la derecha) en la linea de abajo.

```python
nombre = input("Ingrese el nombre del estudiante: ")
nota = float(input("Ingrese la nota obtenida: "))

print("Estudiante:", nombre, "- Nota:", nota)

if nota >= 4.0:
    print("Estado: Aprobado")
```

### Paso 5: Decir si aprobo o reprobo con else
Actualmente, si el estudiante tiene menos de 4.0, el sistema no dice nada. Agrega la instruccion necesaria para capturar el caso contrario e imprimir "Estado: Reprobado".

**Hint:** La palabra reservada `else:` se utiliza al final de un bloque condicional para atrapar cualquier caso que no haya cumplido la condicion del `if`.

```python
nombre = input("Ingrese el nombre del estudiante: ")
nota = float(input("Ingrese la nota obtenida: "))

print("Estudiante:", nombre, "- Nota:", nota)

if nota >= 4.0:
    print("Estado: Aprobado")
else:
    print("Estado: Reprobado")
```

### Paso 6: Agregar examen recuperativo con elif
En nuestra universidad, los alumnos que obtienen entre un 3.6 y un 3.9 tienen una segunda oportunidad. Agrega una condicion intermedia para estos casos.

**Hint:** Utiliza `elif otra_condicion:` entre el `if` y el `else`. Python evaluara las condiciones de arriba hacia abajo.

```python
nombre = input("Ingrese el nombre del estudiante: ")
nota = float(input("Ingrese la nota obtenida: "))

print("Estudiante:", nombre, "- Nota:", nota)

if nota >= 4.0:
    print("Estado: Aprobado")
elif nota >= 3.6:
    print("Estado: Derecho a examen recuperativo")
else:
    print("Estado: Reprobado")
```

### Paso 7: Crear nueva condicion de excelencia
Premiemos a los mejores estudiantes. Modifica la estructura para que, si el estudiante tiene una nota mayor o igual a 6.0, el mensaje sea "Estado: Aprobado con Distincion Maxima".

**Hint:** Debes agregar un nuevo caso en la parte superior. Si pones la validacion del 6.0 debajo del 4.0, nunca se ejecutara porque Python se detendra al confirmar que es mayor a 4.0.

```python
nombre = input("Ingrese el nombre del estudiante: ")
nota = float(input("Ingrese la nota obtenida: "))

print("Estudiante:", nombre, "- Nota:", nota)

if nota >= 6.0:
    print("Estado: Aprobado con Distincion Maxima")
elif nota >= 4.0:
    print("Estado: Aprobado")
elif nota >= 3.6:
    print("Estado: Derecho a examen recuperativo")
else:
    print("Estado: Reprobado")
```

### Paso 8: Pasar de 1 a 3 notas promediadas
Un curso normal tiene mas de una nota. Modifica tu codigo para pedir 3 notas distintas al usuario, calcula un promedio simple matematico entre las 3, y luego pasa ese promedio a tu bloque condicional.

**Hint:** Debes crear variables como `nota1`, `nota2` y `nota3`, pedir los 3 inputs y calcular `promedio = (nota1 + nota2 + nota3) / 3` ANTES del bloque de los `if`.

```python
nombre = input("Ingrese el nombre del estudiante: ")

nota1 = float(input("Ingrese la nota 1: "))
nota2 = float(input("Ingrese la nota 2: "))
nota3 = float(input("Ingrese la nota 3: "))

promedio = (nota1 + nota2 + nota3) / 3
promedio = round(promedio, 1) # Redondeamos a un decimal

print("Estudiante:", nombre, "- Promedio Simple:", promedio)

if promedio >= 6.0:
    print("Estado: Aprobado con Distincion Maxima")
elif promedio >= 4.0:
    print("Estado: Aprobado")
elif promedio >= 3.6:
    print("Estado: Derecho a examen recuperativo")
else:
    print("Estado: Reprobado")
```

### Paso 9: Crear una lista de ponderaciones
En la universidad, las evaluaciones casi nunca valen lo mismo. Para manejar multiples datos de un mismo tipo, usaremos Listas. Crea una lista llamada `ponderaciones` que contenga tres valores decimales que sumen 1.0 (que representa el 100%).

```python
# Definimos el peso de la Nota 1, Nota 2 y Nota 3
ponderaciones = [0.30, 0.30, 0.40] 

print("Las ponderaciones del curso son:", ponderaciones)
```

### Paso 10: Calculo del nuevo promedio (ponderado)
Toma las tres notas que pedimos en el paso 8 y guardalas dentro de una lista llamada `notas` usando `notas = [nota1, nota2, nota3]`. Luego, calcula el promedio ponderado multiplicando cada nota por su respectiva ponderacion accediendo a los indices de las listas.

**Hint:** Recuerda que para acceder al primer elemento de una lista se utiliza la posicion 0. Ejemplo: `notas[0] * ponderaciones[0]`

```python
nota1 = float(input("Ingrese la nota 1: "))
nota2 = float(input("Ingrese la nota 2: "))
nota3 = float(input("Ingrese la nota 3: "))

notas = [nota1, nota2, nota3]
ponderaciones = [0.30, 0.30, 0.40]

# Multiplicamos cada posicion de la lista de notas por su correspondiente en ponderaciones
promedio_ponderado = (notas[0] * ponderaciones[0]) + (notas[1] * ponderaciones[1]) + (notas[2] * ponderaciones[2])
promedio_ponderado = round(promedio_ponderado, 1)

print("Tus notas son:", notas)
print("Tu promedio ponderado es:", promedio_ponderado)
```

### Paso 11: Mostrar toda la informacion por pantalla (Integracion final)
Junta todas las partes del codigo. Pide el nombre, pide las 3 notas, guárdalas en una lista, realiza el calculo del promedio ponderado y finalmente aplica el bloque de `if/elif/else` para imprimir el resumen academico completo del estudiante.

```python
print("--- INGRESO DE DATOS ---")
nombre = input("Ingrese el nombre del estudiante: ")
nota1 = float(input("Ingrese la nota 1: "))
nota2 = float(input("Ingrese la nota 2: "))
nota3 = float(input("Ingrese la nota 3: "))

notas = [nota1, nota2, nota3]
ponderaciones = [0.30, 0.30, 0.40]

promedio_ponderado = (notas[0] * ponderaciones[0]) + (notas[1] * ponderaciones[1]) + (notas[2] * ponderaciones[2])
promedio_ponderado = round(promedio_ponderado, 1)

print("\n--- RESUMEN ACADEMICO ---")
print("Estudiante:", nombre)
print("Calificaciones:", notas)
print("Promedio Final:", promedio_ponderado)

if promedio_ponderado >= 6.0:
    print("Estado: Aprobado con Distincion Maxima")
elif promedio_ponderado >= 4.0:
    print("Estado: Aprobado")
elif promedio_ponderado >= 3.6:
    print("Estado: Derecho a examen recuperativo")
else:
    print("Estado: Reprobado")
```
