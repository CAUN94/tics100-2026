### Pregunta 1: (1 punto)

Un condominio mantiene el registro de la cantidad de meses de deuda de gastos comunes de cada casa en una lista. El índice de la lista representa el número de la casa (0, 1, 2, 3, etc.).

Se le solicita escribir una función en Python llamada `casas_morosas(deudas)` que reciba esta lista de tamaño desconocido. Su función debe analizar el registro y **retornar una nueva lista** que contenga únicamente los números de las casas (es decir, los índices) que acumulen **estrictamente más de 3 meses de deuda**.

A continuación, se le entrega una lista `deudas_condominio` ya definida. El programa principal debe llamar a su función y, a partir de la lista retornada, mostrar el listado de casas a las que se les debe notificar de su morosidad.

**Ejemplo de ejecución:**
*(Si la lista de prueba es `[0, 4, 2, 3, 5, 1, 0, 6]`)*
```text
Las casas a las que se les debe cobrar son: [1, 4, 7]
```

````python
deudas_condominio = [0, 4, 2, 3, 5, 1, 0, 6, 2, 8, 4]

# Escriba su función casas_morosas aquí
def casas_morosas(deudas):
    lista_morosas = []
    
    for i in range(len(deudas)):
        if deudas[i] > 3:
            lista_morosas.append(i)
            
    return lista_morosas

# ==========================================
# Escriba su programa principal aquí

casas_a_cobrar = casas_morosas(deudas_condominio)
print("Las casas a las que se les debe cobrar son:", casas_a_cobrar)
````

### Pregunta 2: (2 puntos)

Un estacionamiento privado representa la ocupación de sus espacios mediante un plano en forma de matriz (lista de listas). Cada número en la matriz indica la cantidad de **minutos** que un vehículo lleva estacionado. Si la casilla contiene un `0`, significa que el espacio está desocupado.

La tarifa del estacionamiento se calcula mediante la siguiente ecuación:
- Los primeros 60 minutos (o fracción) tienen un costo base de `$5.000`.
- Pasados los 60 minutos, se cobran `$7` adicionales por cada minuto extra.

Se le solicita escribir una función llamada `calcular_recaudacion(plano)` que reciba la matriz del estacionamiento. Su función debe recorrer el plano utilizando ciclos anidados, calcular de forma independiente cuánto debe pagar cada auto según la regla tarifaria descrita, y **retornar el monto total de dinero** que el estacionamiento está ganando en ese momento.

A continuación se le entrega una matriz `plano_estacionamiento` ya definida. Llame a su función pasándole esta matriz y muestre la recaudación total por pantalla (sólo el número retornado, sin mensajes ni textos adicionales).

**Ejemplo de ejecución:**
*(Si un auto lleva 60 mins paga 5000; si lleva 120 mins paga 5000 + (60 * 7) = 5420. El total será la suma de todos)*
```text
15420
```

````python
plano_estacionamiento = [
    [0, 60,  0,  30],
    [0,  0, 120,  0],
    [45, 0,  0, 200]
]

# Escriba su función calcular_recaudacion aquí
def calcular_recaudacion(plano):
    total_recaudado = 0
    
    for i in range(len(plano)):
        for j in range(len(plano[i])):
            minutos = plano[i][j]
            
            if minutos > 0:
                if minutos <= 60:
                    total_recaudado = total_recaudado + 5000
                else:
                    minutos_extra = minutos - 60
                    total_recaudado = total_recaudado + 5000 + (minutos_extra * 7)
                    
    return total_recaudado

# ==========================================
# Programa principal (Pruebe su función aquí)

print(calcular_recaudacion(plano_estacionamiento))
````

### Pregunta 3: (3 puntos)

Una revista de pasatiempos lógicos le ha pedido programar un juego interactivo llamado "Cuadrado Mágico". El tablero es una cuadrícula de 3x3. El objetivo es rellenarla con los números del 1 al 9 (sin repetir), de tal forma que la suma de cada fila, cada columna y ambas diagonales sea exactamente **15**.

Para facilitarle el trabajo, sus colegas le han entregado tres funciones listas para usar:
- `mostrar_tablero(tablero)`: Imprime el tablero por pantalla. Los ceros (`0`) se muestran como puntos (`.`).
- `es_jugada_valida(tablero, fila, columna, numero)`: Retorna `True` si la coordenada está dentro de la cuadrícula de 3x3, la casilla está vacía (`0`), y el número ingresado (entre 1 y 9) no se ha usado previamente. Retorna `False` en caso contrario.
- `verificar_victoria(tablero)`: Retorna `True` si el tablero está lleno y todas sus líneas (filas, columnas y diagonales) suman exactamente 15. Retorna `False` en caso contrario.

Se le solicita que escriba una función llamada `jugar_cuadrado_magico(tablero)`. Esta función debe implementar un ciclo iterativo (`while`) que pregunte repetidamente al usuario la fila, la columna y el número que desea colocar. Por cada intento, utilice la función `es_jugada_valida`:
- Si retorna `True`, coloque el número en la matriz, imprima *"¡Número colocado con éxito!"* y muestre el tablero actualizado.
- Si retorna `False`, imprima *"Jugada inválida. Coordenada incorrecta, número inválido o ya utilizado."*

El ciclo debe terminar **únicamente** cuando el jugador llene por completo el tablero (es decir, cuando ya no queden ceros). Al finalizar el ciclo, llame a la función `verificar_victoria(tablero)`. 
Si ésta retorna `True`, imprima *"¡Felicidades, logró el Cuadrado Mágico!"*. 
Si retorna `False`, imprima *"Tablero lleno, pero las sumas no dan 15. Fin del juego."*

**Ejemplos de ejecución:**
```text
--- CUADRADO MÁGICO ---
. . .
. . .
. . .
-----------------------
Ingrese fila: 1
Ingrese columna: 1
Ingrese número: 5
¡Número colocado con éxito!

--- CUADRADO MÁGICO ---
. . .
. 5 .
. . .
-----------------------
... (luego de llenar las 9 casillas)
¡Felicidades, logró el Cuadrado Mágico!
```

````python
# ==========================================
# FUNCIONES PRE-PROGRAMADAS (NO MODIFICAR)
# ==========================================
def mostrar_tablero(tablero):
    print("\n--- CUADRADO MÁGICO ---")
    for i in range(len(tablero)):
        for j in range(len(tablero[i])):
            if tablero[i][j] == 0:
                print(".", end=" ")
            else:
                print(tablero[i][j], end=" ")
        print()
    print("-----------------------\n")

def es_jugada_valida(tablero, fila, col, num):
    # Verificamos que las coordenadas existan y que el número esté entre 1 y 9
    if 0 <= fila < 3 and 0 <= col < 3 and 1 <= num <= 9:
        if tablero[fila][col] == 0:
            # Verificar que el número no se haya usado en todo el tablero
            for i in range(3):
                for j in range(3):
                    if tablero[i][j] == num:
                        return False
            return True
    return False

def verificar_victoria(tablero):
    # Comprobar filas y columnas
    for i in range(3):
        if sum(tablero[i]) != 15:
            return False
        if tablero[0][i] + tablero[1][i] + tablero[2][i] != 15:
            return False
    # Comprobar diagonales
    if tablero[0][0] + tablero[1][1] + tablero[2][2] != 15:
        return False
    if tablero[0][2] + tablero[1][1] + tablero[2][0] != 15:
        return False
    return True

tablero_vacio = [
    [0, 0, 0],
    [0, 0, 0],
    [0, 0, 0]
]

# ==========================================
# Escriba su función jugar_cuadrado_magico aquí
def jugar_cuadrado_magico(tablero):
    # Sabemos que un tablero 3x3 tiene 9 casillas en total
    casillas_vacias = 9
    
    while casillas_vacias > 0:
        fila_jugada = int(input("Ingrese fila: "))
        columna_jugada = int(input("Ingrese columna: "))
        numero_jugado = int(input("Ingrese número: "))
        
        # Evaluamos usando la función que nos dieron
        if es_jugada_valida(tablero, fila_jugada, columna_jugada, numero_jugado):
            tablero[fila_jugada][columna_jugada] = numero_jugado
            casillas_vacias -= 1  # Restamos una casilla vacía
            print("¡Número colocado con éxito!")
            mostrar_tablero(tablero)
        else:
            print("Jugada inválida. Coordenada incorrecta, número inválido o ya utilizado.")

    # Cuando casillas_vacias llega a 0, el while termina y verificamos el tablero completo
    if verificar_victoria(tablero):
        print("¡Felicidades, logró el Cuadrado Mágico!")
    else:
        print("Tablero lleno, pero las sumas no dan 15. Fin del juego.")

# Llamada para iniciar el juego (Descomentar para probar)
# mostrar_tablero(tablero_vacio)
# jugar_cuadrado_magico(tablero_vacio)
````
