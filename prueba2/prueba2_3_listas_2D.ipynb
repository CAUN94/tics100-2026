# Guía: Listas de dos dimensiones (matrices)

Esta guía es material de apoyo para trabajar con listas de dos dimensiones (matrices). Una matriz es una tabla de valores organizada en filas y columnas, y se recorre con dos índices: `i` para la fila y `j` para la columna.

**Importante:** La matriz es una lista de listas de Python, se construye con `.append()` y se recorre con ciclos `for` anidados.

## 1. Qué es una matriz y cómo se accede

Una matriz `n x m` tiene `n` filas y `m` columnas. Cada celda se ubica con dos índices: `matriz[i][j] = fila i, columna j`. Ambos parten en 0.

```text
         j=0  j=1  j=2  j=3
       +----+----+----+----+
   i=0 | 42 | 73 | 51 | 14 |
       +----+----+----+----+
   i=1 | 65 | 81 | 89 | 32 |
       +----+----+----+----+
   i=2 | 02 | 09 | 70 | 92 |
```

* `matriz[1][2] = 89` (fila 1, columna 2)
* `matriz[2][3] = 92` (fila 2, columna 3)

* `matriz[i]` es una fila completa (una lista 1D).
* `matriz[i][j]` es una celda explícita.
* `len(matriz)` = número de filas.
* `len(matriz[0])` = número de columnas (largo de la primera fila).

## 2. Crear una matriz

### 2.1 Escrita directamente (literal)
Para ejercicios donde la matriz ya viene dada, es una lista de listas: cada `[...]` interno es una fila.

````python
matriz_ejemplo = [
    [42, 73, 51, 14],
    [65, 81, 89, 32],
    [ 2,  9, 70, 92]
]

print(f"Elemento en la fila 1, columna 2: {matriz_ejemplo[1][2]}")
````

### 2.2 Crear y llenar de tamaño n x m con append
Se construye fila por fila: por cada fila se crea una lista vacía, se llena con `append`, y luego esa fila se agrega a la matriz principal.

````python
n = 3 # int(input("Filas: "))
m = 4 # int(input("Columnas: "))
matriz = [] # matriz vacía

for i in range(n):
    fila = [] # ¡una fila NUEVA y vacía en cada vuelta!
    
    for j in range(m):
        # valor = int(input(f"Valor [{i}][{j}]: "))
        valor = (i + 1) * (j + 1) # Simulando el ingreso de datos
        fila.append(valor) # agrega la celda a la fila actual
        
    matriz.append(fila) # agrega la fila completa a la matriz

print(f"Matriz generada: {matriz}")
````

**¡CUIDADO!** La línea `fila = []` debe ir dentro del `for i` (una fila nueva por cada vuelta). Si la pones antes del ciclo, todas las filas serían la misma lista y se pisarían entre ellas.

## 3. Recorrer e imprimir la matriz
El patrón base: `i` recorre filas, `j` recorre columnas. Por cada fila `i` se recorren todas sus columnas `j`.

````python
print(f"--- MATRIZ EN FORMATO TABLA ---")
for i in range(len(matriz)): # cada fila
    for j in range(len(matriz[0])): # cada columna de esa fila
        print(f"{matriz[i][j]}", end=" ") # \t = tabulación, alinea columnas
    print() # salto de línea al terminar la fila
````

## 4. Cálculos sobre la matriz

### 4.1 Suma de todos los elementos

````python
suma = 0
for i in range(len(matriz)):
    for j in range(len(matriz[0])):
        suma = suma + matriz[i][j]
        
print(f"Suma total: {suma}")
````

### 4.2 Máximo y su posición

````python
maximo = matriz[0][0]
filaMax = 0
colMax = 0

for i in range(len(matriz)):
    for j in range(len(matriz[0])):
        if matriz[i][j] > maximo:
            maximo = matriz[i][j]
            filaMax = i
            colMax = j
            
print(f"Máximo: {maximo} en fila {filaMax} columna {colMax}")
````

### 4.3 Suma por fila
`matriz[i]` es una fila completa; se suma con un `for` interno:

````python
for i in range(len(matriz)):
    sumaFila = 0
    for j in range(len(matriz[0])):
        sumaFila = sumaFila + matriz[i][j]
    print(f"Fila {i} suma: {sumaFila}")
````

## 5. Los 8 vecinos de una celda

Muchos ejercicios piden mirar los 8 vecinos de una celda (arriba, abajo, lados y diagonales):

```text
[i-1][j-1] | [i-1][j] | [i-1][j+1]
 [i][j-1]  |  [i][j]  |  [i][j+1]
[i+1][j-1] | [i+1][j] | [i+1][j+1]
```
(el centro `[i][j]` NO es vecino de sí mismo)

### 5.1 El problema del borde
Una celda en el borde no tiene los 8 vecinos: mirar `matriz[i-1][j]` cuando `i=0` se sale de la matriz (`i-1 = -1`). La solución más simple cuando se exige tener 8 vecinos: **recorrer solo el interior**, dejando fuera la primera y última fila/columna.

```python
for i in range(1, len(matriz)-1):       # filas interiores: de 1 a n-2
    for j in range(1, len(matriz[0])-1):  # columnas interiores: de 1 a m-2
        # aquí [i][j] SIEMPRE tiene sus 8 vecinos
```
`range(1, n-1)` salta el índice 0 (primer borde) y el `n-1` (último borde).

## 6. Ejemplo completo 1: contar cumbres

Dada una matriz con alturas, una cumbre es una celda estrictamente mayor que sus 8 vecinos.
Cuente cuántas cumbres hay e imprima la altura de la cumbre más alta. Todo en el programa principal.

````python
tablero = [
    [42, 73, 51, 14, 43, 15, 35, 24, 20, 65],
    [65, 81, 89, 32, 74, 92, 50, 12, 94, 22],
    [ 2,  9, 70, 92, 61,  2, 17, 22, 65,  1],
    [65, 28, 57, 91, 67, 41,  8, 99, 99,  4],
    [16, 78, 57, 89, 50, 92, 24, 30, 92, 73],
    [14, 32, 74, 48, 40, 94, 52, 21,  8, 80]
]

numCumbres = 0
cumbreMasAlta = 0

# solo el interior: las cumbres necesitan sus 8 vecinos
for i in range(1, len(tablero)-1):
    for j in range(1, len(tablero[0])-1):
        actual = tablero[i][j]
        
        # ¿es mayor que TODOS sus 8 vecinos?
        esCumbre = (actual > tablero[i-1][j-1] and
                    actual > tablero[i-1][j] and
                    actual > tablero[i-1][j+1] and
                    actual > tablero[i][j-1] and
                    actual > tablero[i][j+1] and
                    actual > tablero[i+1][j-1] and
                    actual > tablero[i+1][j] and
                    actual > tablero[i+1][j+1])
        
        if esCumbre:
            numCumbres = numCumbres + 1
            if actual > cumbreMasAlta:
                cumbreMasAlta = actual

print(f"Número de cumbres: {numCumbres}")
print(f"La cumbre más alta tiene una altura de {cumbreMasAlta}")
````

Los 99 de la fila 3 no son cumbres: son vecinos entre sí, así que ninguno es *estrictamente mayor* que el otro.

### 6.1 Truco: comparar con el vecino más grande (versión corta)
En vez de escribir 8 condiciones, se puede buscar el mayor de los 8 vecinos y comparar una sola vez:

````python
numCumbres = 0

for i in range(1, len(tablero)-1):
    for j in range(1, len(tablero[0])-1):
        
        # mayor vecino, recorriendo el bloque 3x3 y saltando el centro
        mayorVecino = 0
        for di in range(-1, 2):      # di = -1, 0, 1
            for dj in range(-1, 2):  # dj = -1, 0, 1
                
                if di == 0 and dj == 0:
                    continue # saltar la celda central
                    
                if tablero[i+di][j+dj] > mayorVecino:
                    mayorVecino = tablero[i+di][j+dj]
                    
        if tablero[i][j] > mayorVecino:
            numCumbres = numCumbres + 1

print(f"Número de cumbres calculado con matriz 3x3: {numCumbres}")
````

`di` y `dj` son los desplazamientos: `(i+di, j+dj)` recorre las 9 celdas del bloque 3x3; el `continue` salta el centro.

## 7. Ejemplo completo 2: etiquetar puntos calientes

Dada una matriz de clics. Un punto caliente anómalo es una celda (no en el borde) cuyo valor es al menos 5 veces el promedio de sus 8 vecinos. Cuente cuántos hay y reemplácelos por `999`.

Se hace en dos fases: primero detectar (guardando las posiciones en dos listas paralelas, como en la guía 1D), después etiquetar. Así, marcar un punto con 999 no altera el promedio de los vecinos que aún faltan revisar.

````python
matriz = [
    [1,  2, 1,  3,  2],
    [2, 60, 2,  1,  4],
    [1,  2, 3, 50, 21],
    [3,  1, 2,  4,  1]
]

# FASE 1: detectar. Guardo fila y columna de cada punto caliente.
filasHot = []
colsHot = []

for i in range(1, len(matriz)-1):
    for j in range(1, len(matriz[0])-1):
        
        # suma de los 8 vecinos
        sumaVecinos = (matriz[i-1][j-1] + matriz[i-1][j] + matriz[i-1][j+1] +
                       matriz[i][j-1]                    + matriz[i][j+1] +
                       matriz[i+1][j-1] + matriz[i+1][j] + matriz[i+1][j+1])
        
        promedioVecinos = sumaVecinos / 8
        
        if matriz[i][j] >= 5 * promedioVecinos:
            filasHot.append(i)
            colsHot.append(j)

contador = len(filasHot)

# FASE 2: etiquetar. Recorro las posiciones detectadas y pongo 999.
for k in range(contador):
    fila = filasHot[k]
    col = colsHot[k]
    matriz[fila][col] = 999

print(f"Puntos calientes encontrados: {contador}")
print(f"Matriz etiquetada:")
for i in range(len(matriz)):
    for j in range(len(matriz[0])):
        print(f"{matriz[i][j]}\t", end="")
    print()
````

**Dos detalles:**
* En la suma de vecinos se omite a propósito `matriz[i][j]` (el centro), por eso son 8 términos y no 9.
* `filasHot` y `colsHot` son dos listas paralelas (la posición `k` de una va con la de la otra), igual que en la guía de listas 1D.

## 8. Errores frecuentes

* **Crear `fila = []` fuera del `for i`** — todas las filas terminan siendo la misma lista. La fila nueva se crea dentro del ciclo de filas.
* **`IndexError` en los bordes** — si necesitas los 8 vecinos, recorre el interior con `range(1, n-1)` y `range(1, m-1)`.
* **Confundir filas con columnas** — `matriz[i][j]`: el primer índice es la fila, el segundo la columna. `len(matriz)` son filas, `len(matriz[0])` columnas.
* **Usar el largo equivocado** — la cantidad de columnas es `len(matriz[0])`, no `len(matriz)`.
* **Comparar con `>=` cuando se pedía `>` (o al revés)** — en "cumbres" es estrictamente mayor (`>`); los empates (los 99) no cuentan.
* **Reiniciar mal los acumuladores** — la `sumaFila` debe volver a `0` dentro del `for` de filas, antes de recorrer sus columnas.

### Recomendaciones

* Dibuja la matriz como una grilla con los índices `i` (filas) y `j` (columnas) numerados desde 0.
* Para los ejercicios de vecinos, dibuja el bloque `3x3` alrededor de `[i][j]` y verifica con una celda interior antes de programar.
* Prueba siempre con una matriz pequeña (`3x3` o `4x5`) y rutea a mano la primera celda interior.
