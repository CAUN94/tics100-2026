# Guía: Listas de una dimensión y uso de índice

Esta guía es material de apoyo para trabajar con listas unidimensionales (1D). El objetivo central es trabajar con una o dos listas en paralelo y calcular algo recorriéndolas por índice.


## 1. Qué es una lista 1D

Una lista de una dimensión es una fila de casilleros numerados, cada uno con un valor. Los casilleros se numeran (su índice) desde 0:

```text
índice:    0    1    2    3    4
        +----+----+----+----+----+
valor:  | 23 | 41 | 19 | 35 | 28 |
        +----+----+----+----+----+
```

````python
lista = [23, 41, 19, 35, 28]
````

### 1.1 Crear una lista vacía y llenarla con append

Se parte con una lista vacía `[]` y se van agregando datos con `.append()` dentro de un `for`:

````python
n = 3 # int(input("Cuántos datos: "))
lista = [] # lista vacía: []

for i in range(n):
    # dato = int(input(f"Ingrese dato {i+1}: "))
    dato = (i + 1) * 10 # Simulando ingreso de datos: 10, 20, 30...
    lista.append(dato)  # agrega el dato al final de la lista

print(f"Lista resultante: {lista}")
````

Cada `.append(dato)` agrega un valor al final. Después de `n` vueltas, la lista tiene `n` elementos.

### 1.2 Leer y recorrer por índice

Una vez llena, se accede a cada elemento por su índice con `lista[i]`:

````python
for i in range(n):
    print(f"Posición {i} tiene el valor {lista[i]}")
````

* `lista[i]` lee el casillero `i`.
* Primer elemento: `lista[0]`.
* Último elemento: `lista[n-1]` (¡no `lista[n]`, eso da error!).
* `len(lista)` entrega el tamaño de la lista (sirve si no guardaste `n`).

**"Usar índice" es justamente esto: acceder a cada dato con `lista[i]`**. Es la base de todos los cálculos que siguen y de combinar dos listas en paralelo.

## 2. Cálculos básicos (con índice)

Todo se calcula recorriendo la lista con `for i in range(...)` y acumulando en una variable.

### 2.1 Suma y promedio

````python
lista_ejemplo = [10, 20, 30]
suma = 0

# Recuerda que len(lista_ejemplo) nos da el número de elementos en la lista, que es 3 en este caso

for i in range(len(lista_ejemplo)):
    suma = suma + lista_ejemplo[i] # acumula
    
promedio = suma / len(lista_ejemplo)

print(f"Suma: {suma} | Promedio: {promedio}")
````

La variable `suma` parte en 0 y crece en cada vuelta. El `promedio` es la suma dividida por la cantidad de datos (`len(lista)`).

### 2.2 Mínimo y máximo (el patrón clave)

Se parte suponiendo que el primer elemento es el mínimo (y el máximo), y se compara con el resto:

````python
lista_valores = [23, 41, 19, 35]

minimo = lista_valores[0] # supongo que el primero es el menor
maximo = lista_valores[0] # y también el mayor

for i in range(0, len(lista_valores)): 
    
    if lista_valores[i] < minimo:
        minimo = lista_valores[i] # encontré uno más chico
        
    if lista_valores[i] > maximo:
        maximo = lista_valores[i] # encontré uno más grande

print(f"Mínimo: {minimo} | Máximo: {maximo}")
````

Tabla de ruteo para `lista = [23, 41, 19, 35]`:

| i | `lista[i]` | ¿< minimo? | `minimo` | ¿> maximo? | `maximo` |
|---|---|---|---|---|---|
| 0 | 23 | no | 23 | no | 23 |
| 1 | 41 | no | 23 | sí | 41 |
| 2 | 19 | sí | 19 | no | 41 |
| 3 | 35 | no | 19 | no | 41 |

Resultado: `minimo = 19`, `maximo = 41`.

**Error típico:** partir con `minimo = 0`. Si todos los datos son positivos, el mínimo "calculado" quedaría en 0, que no existe en la lista. Por eso se parte con `lista[0]`.

## 3. Dos listas en paralelo

**La idea central:** dos listas del mismo tamaño, donde la posición `i` de una corresponde a la posición `i` de la otra. Se recorren con el mismo índice `i`.

```text
índice:         0      1      2      3
precios:    [ 1500,  800,  2300,  1200 ]
cantidades: [    2,    5,     1,     3 ]
                     ^ ^
          precio[1]=800 va con cantidad[1]=5
```

### 3.1 Crear y llenar dos listas paralelas
Las dos se llenan en el mismo `for`, así quedan alineadas por posición:

````python
n = 2 # int(input("Cuántos productos: "))
precios = []
cantidades = []

for i in range(n):
    # precios.append(int(input(f"Precio del producto {i+1}: ")))
    # cantidades.append(int(input(f"Cantidad del producto {i+1}: ")))
    precios.append((i+1) * 1000) # Simulando ingreso: 1000, 2000...
    cantidades.append(i + 2)     # Simulando ingreso: 2, 3...

print(f"Precios: {precios}")
print(f"Cantidades: {cantidades}")
````

### 3.2 Combinar por índice (calcular algo)
Como ambas comparten el índice, en cada vuelta se usan `precios[i]` y `cantidades[i]` juntos:

````python
total = 0
for i in range(len(precios)):
    total = total + precios[i] * cantidades[i] # precio por cantidad de ESE producto
    
print(f"Total a pagar: ${total}")
````

### 3.3 Ejemplo: promedio ponderado (notas y créditos)
Dos listas: la nota de cada ramo y sus créditos. El promedio ponderado usa ambas por índice:

````python
notas = [5.5, 6.0, 4.0]
creditos = [4, 6, 2]

sumaPond = 0
sumaCreditos = 0

for i in range(len(notas)):
    sumaPond = sumaPond + notas[i] * creditos[i]
    sumaCreditos = sumaCreditos + creditos[i]

promedio = sumaPond / sumaCreditos
print(f"Promedio ponderado: {promedio}")
````

### 3.4 Ejemplo: buscar el "mejor" según otra lista
Encontrar el producto más caro y mostrar de paso cuántas unidades hay de ese producto (dato que vive en la otra lista, misma posición):

````python
precios = [1500, 800, 2300, 1200]
cantidades = [2, 5, 1, 3]

posMax = 0 
for i in range(1, len(precios)):
    if precios[i] > precios[posMax]:
        posMax = i # posición del precio más alto
        # guardo la POSICIÓN, no solo el valor

print(f"Producto más caro: posición {posMax}")
print(f"Precio: ${precios[posMax]} | Cantidad disponible: {cantidades[posMax]}")
````

Guardar la posición (`posMax`) es la técnica clave: con un solo índice accedes al dato en las dos listas.

## 4. Ejemplo completo

Una empresa quiere analizar a sus clientes. De cada cliente conoce su edad y su gasto mensual. Pida la edad y el gasto de `numClientes` clientes, guárdelos en dos listas, y muestre: el promedio de edad, la edad mínima y máxima, y el gasto total de todos los clientes.

````python
# 1) Tamaño y creación de las dos listas vacías
numClientes = 3 # int(input("Número de clientes: "))
edades = [25, 40, 19] # Simulando ingresos
gastos = [50000, 120000, 15000] # Simulando ingresos

# (Se omite el paso 2 de llenado con input() para que el código se pueda ejecutar)

# 3) Cálculos a mano por índice (sin numpy)
sumaEdad = 0
minEdad = edades[0]
maxEdad = edades[0]
gastoTotal = 0

for i in range(numClientes):
    sumaEdad = sumaEdad + edades[i]
    gastoTotal = gastoTotal + gastos[i]
    
    if edades[i] < minEdad:
        minEdad = edades[i]
        
    if edades[i] > maxEdad:
        maxEdad = edades[i]

promedioEdad = sumaEdad / numClientes

# 4) Mostrar resultados
print(f"Promedio de edad: {promedioEdad}")
print(f"Edad mínima: {minEdad}")
print(f"Edad máxima: {maxEdad}")
print(f"Gasto total de la empresa: ${gastoTotal}")
````

**Nota:** se llena con `append` y después se recorre por índice para calcular. Todo el cálculo cabe en un mismo `for`; también sería correcto usar un `for` por cálculo.

## 5. Errores frecuentes

* `lista[n]` — el último índice válido es `n-1` (o `len(lista)-1`). `lista[n]` produce `IndexError: list index out of range`.
* **Partir minimo/maximo en 0.** Inícialos con `lista[0]`, no con 0.
* **Recorrer dos listas con índices distintos.** Si son paralelas, usa el mismo `i` para ambas: `lista1[i]` y `lista2[i]`.
* **Llenar las dos listas en for separados con cantidades distintas** — quedan desalineadas. Llénalas juntas o cuida que tengan el mismo tamaño.
* **Olvidar inicializar el acumulador** (`suma = 0`) antes del ciclo.
* **Dividir el promedio dentro del for.** El promedio se calcula después de terminar de sumar, una sola vez.
