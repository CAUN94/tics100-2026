### Ejercicio 1: Obtención de índices usando `range(len())`

**Concepto:** Cuando usamos un simple `for elemento in lista`, podemos *ver* los datos, pero no podemos modificarlos en su lugar de origen. Si queremos alterar la lista original, necesitamos saber el número de la casilla exacta (el índice `i`). Para eso combinamos `range()` (que genera números) y `len()` (que nos dice cuántos elementos hay).

**Enunciado:** 
Tienes una lista de precios base. Usa un ciclo `for` con `range(len(lista))` para recorrer la lista entera. Multiplica cada precio por 2 y guarda el resultado exactamente en la misma posición de la lista original. Muestra la lista antes y después.

```python
precios = [100, 250, 50, 300]
print("Precios originales:", precios)

# len(precios) es 4. 
# range(4) genera los números: 0, 1, 2, 3. Los indices de la lista 'precios'.
for i in range(len(precios)):
    # Tomamos el valor en la casilla 'i', lo multiplicamos y lo guardamos de vuelta en 'i'
    precios[i] = precios[i] * 2

print("Precios actualizados:", precios)
```

### Ejercicio 2: Mostrando una Matriz por Pantalla

**Concepto:** Una matriz ya creada en memoria se ve fea si la imprimimos de golpe con un `print(matriz)`. Para verla como una cuadrícula, necesitamos un ciclo que recorra las filas (hacia abajo) y otro ciclo adentro que recorra las columnas (hacia el lado).

**Enunciado:**
Dada una matriz de 3x3 predefinida, utiliza un doble ciclo `for` para mostrarla en pantalla. Recuerda usar `end=" "` para que los números queden al lado, y un `print()` vacío para hacer el salto de línea al terminar la fila.

```python
matriz_numeros = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print("--- MATRIZ EN CUADRÍCULA ---")

# Ciclo exterior 'i' para las FILAS
# Len(matriz_numeros) es 3, así que range(3) genera los números: 0, 1, 2. Los índices de las filas.
for i in range(len(matriz_numeros)):
    
    # Ciclo interior 'j' para las COLUMNAS
    # Len(matriz_numeros[i]) es 3, así que range(3) genera los números: 0, 1, 2. Los índices de las columnas.
    # Veamos que se usa el índice 'i' para acceder a la fila correcta.
    for j in range(len(matriz_numeros[i])):
        # end=" " evita que salte hacia abajo de inmediato
        print(matriz_numeros[i][j], end=" ")
        
    # Este print vacío pertenece al ciclo exterior. Actúa como la tecla "Enter"
    print()
```

### Ejercicio 3: Creación de una Matriz Aleatoria

**Concepto:** Para construir una matriz desde cero, necesitamos crear una lista general vacía (`[]`). Luego, con un doble ciclo `for`, creamos listas "temporales" (que serán las filas), las rellenamos con datos y finalmente metemos esa fila completa dentro de la lista general.

**Enunciado:**
Usando la librería `random`, crea una matriz de 4x4. El ciclo exterior creará las filas y el interior generará números al azar entre 0 y 9. Al terminar de construirla, usa otro doble `for` para mostrarla por pantalla.

```python
import random

# Creamos una matriz vacía y definimos su dimensión
matriz_azar = []
dimension = 4
# La dimensión es 4, así que la matriz tendrá 4 filas y 4 columnas. En total, 16 números al azar.

# --- 1. CREACIÓN DE LA MATRIZ ---

for i in range(dimension):
    fila_temporal = []  # Nace una fila vacía
    
    for j in range(dimension):
        numero = random.randint(0, 9)
        # numero = int(input("Número para la fila " + str(i) + ": "))  # Para llenar la fila a mano
        fila_temporal.append(numero)  # Llenamos la fila hacia el lado
        
    matriz_azar.append(fila_temporal) # Metemos la fila terminada a la matriz grande


# --- 2. VISUALIZACIÓN DE LA MATRIZ ---
print("--- MATRIZ CREADA AL AZAR ---")
for i in range(dimension):
    for j in range(dimension):
        print(matriz_azar[i][j], end=" ")
    print()
```

### Ejercicio 4: ¿Cómo funciona la Revisión de Vecinos?

**Concepto:** Para revisar qué hay alrededor de una coordenada, no escribimos 8 condiciones a mano. Usamos "desplazamientos". Iteramos una lista `[-1, 0, 1]` para las filas y columnas. Al sumar esos desplazamientos a nuestra coordenada central, ¡obtenemos las coordenadas de los vecinos mágicamente!

**Enunciado:**
Tenemos una matriz pequeña de 3x3 y nuestro objetivo está en el centro (Fila 1, Columna 1). Usa un doble `for` iterando sobre `[-1, 0, 1]` para imprimir el valor de todos los vecinos. Incluye la regla para NO leer el centro (nosotros mismos) y la regla para no salirte de los bordes.

```python
mapa = [
    ["A", "B", "C"],
    ["D", "X", "E"], # La "X" está en el centro: mapa[1][1]
    ["F", "G", "H"]
]

fila_centro = 1
col_centro = 1
limite = len(mapa)

print("Nuestros vecinos alrededor de la 'X' son:\n")

# df = desplazamiento en fila | dc = desplazamiento en columna
for df in [-1, 0, 1]:
    for dc in [-1, 0, 1]:
        
        # REGLA 1: Si no me muevo (0,0), soy yo mismo. Ignorar.
        if df == 0 and dc == 0:
            continue
            
        # Sumamos el desplazamiento a nuestro centro para hallar al vecino
        vecino_fila = fila_centro + df
        vecino_col = col_centro + dc
        
        # REGLA 2: Límite de seguridad. Que las coordenadas existan y no sean negativas.
        if vecino_fila >= 0 and vecino_fila < limite and vecino_col >= 0 and vecino_col < limite:
            
            # Extraemos el valor usando nuestras coordenadas vecinas ya calculadas y seguras
            valor = mapa[vecino_fila][vecino_col]
            print(f"En la coordenada [{vecino_fila}][{vecino_col}] está el vecino: {valor}")
```
