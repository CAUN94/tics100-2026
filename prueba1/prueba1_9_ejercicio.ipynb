**Enunciado del Ejercicio:**

La Conjetura de Collatz (o problema 3x+1) es un famoso problema matemático no resuelto. Las reglas para generar la secuencia son muy simples:
1. Se comienza con cualquier número entero positivo.
2. Si el número es par, se divide entre 2.
3. Si el número es impar, se multiplica por 3 y se le suma 1.
4. El proceso se repite con el nuevo número obtenido hasta que el resultado final sea 1.

Escriba un programa que solicite al usuario un número entero positivo inicial. El programa debe generar y mostrar toda la secuencia de Collatz paso a paso hasta llegar al número 1, y finalmente debe imprimir cuántos pasos (operaciones) le tomó a la computadora llegar a ese resultado.

### Paso 1: La logica de las reglas (Par o Impar)
El núcleo de este ejercicio es saber qué fórmula matemática aplicar. Para esto, la computadora necesita saber si el número actual es par o impar.

Como vimos en ejercicios anteriores, usamos el operador módulo (`%`). Si al dividir un número entre 2 el resto es cero (`numero % 2 == 0`), entonces es par. Si no, obligatoriamente es impar.

```python
numero_prueba = 10

print("Evaluando el numero:", numero_prueba)

if numero_prueba % 2 == 0:
    print("El numero es PAR. La regla dice que debemos dividirlo entre 2.")
else:
    print("El numero es IMPAR. La regla dice que debemos multiplicarlo por 3 y sumar 1.")
```

### Paso 2: El concepto mas importante, la ACTUALIZACION del numero
En matemáticas, si tenemos `x = 10` y luego `x = 5`, son dos variables distintas. En programación, la variable es una sola "caja de memoria" que **se sobrescribe a sí misma**.

Para que la secuencia de Collatz avance, debemos tomar el número, aplicarle la fórmula, y **guardar el resultado dentro de la misma variable**. 
La instrucción `numero = numero // 2` significa: *"Toma el valor actual, divídelo en 2, y destruye el valor viejo para guardar este nuevo resultado"*.

*(Nota: Usamos doble barra `//` para que la división sea exacta y no nos entregue números con decimales como 5.0)*.

Veamos cómo la variable transmuta su valor ejecutando la lógica manualmente un par de veces:

```python
# Iniciamos con el numero 6
numero = 6
print("Valor INICIAL de la variable:", numero)

# --- PRIMER PASO MANUAL ---
if numero % 2 == 0:
    print("\n-> Como", numero, "es par, aplicamos: numero = numero // 2")
    numero = numero // 2  # Aqui ocurre la sobrescritura
else:
    numero = (numero * 3) + 1
    
print("Valor ACTUALIZADO de la variable:", numero)

# --- SEGUNDO PASO MANUAL ---
if numero % 2 == 0:
    numero = numero // 2
else:
    print("\n-> Como", numero, "es impar, aplicamos: numero = (numero * 3) + 1")
    numero = (numero * 3) + 1  # Aqui ocurre la sobrescritura

print("Valor ACTUALIZADO de la variable:", numero)

# --- TERCER PASO MANUAL ---
if numero % 2 == 0:
    print("\n-> Como", numero, "es par, aplicamos: numero = numero // 2")
    numero = numero // 2  # Aqui ocurre la sobrescritura
else:
    numero = (numero * 3) + 1

print("Valor ACTUALIZADO de la variable:", numero)
```

### Paso 3: Usar 'while' para automatizar la actualizacion
Como pudiste ver, tuvimos que escribir el bloque `if/else` tres veces a mano, y el número apenas bajó de 6 a 5. Hay números iniciales (como el 27) que toman más de 100 pasos en llegar al 1. Como la cantidad de pasos es **totalmente impredecible**, no podemos usar un ciclo `for`.

Estamos obligados a usar un ciclo `while`. La condición es muy simple, tal como dice la regla matemática: el proceso se repite hasta llegar al 1. En lenguaje de programación, esto se traduce a: **repetir MIENTRAS el número sea distinto de 1 (`while numero != 1:`)**.

El ciclo `while` se encargará de atrapar nuestro bloque `if/else` del Paso 2 y repetirlo a la velocidad de la luz, sobrescribiendo la variable una y otra vez hasta que el valor atrapado en la memoria sea exactamente `1`.

```python
numero = 6
print("Iniciando secuencia automatica para el numero:", numero)

# El ciclo encierra la logica de actualizacion
while numero != 1:
    if numero % 2 == 0:
        numero = numero // 2
    else:
        numero = (numero * 3) + 1
        
    print("Nuevo valor:", numero)

print("El ciclo ha terminado porque la variable finalmente vale 1.")
```

### Paso 4: Solucion final uniendo conceptos
Ahora armaremos el programa final.

Pediremos el número inicial al usuario. Además, el enunciado nos pide contar cuántos pasos tomó el proceso. Para esto, crearemos un contador `pasos = 0` antes del ciclo.

Dentro del ciclo `while`, agregaremos la instrucción `pasos = pasos + 1`. Así, cada vez que la computadora actualice el número (ya sea dividiéndolo o multiplicándolo), registrará que ha dado un paso más.

```python
print("--- SIMULADOR DE LA CONJETURA DE COLLATZ ---")

# 1. Solicitamos el numero de inicio y creamos el contador
numero = int(input("Ingrese un numero entero positivo: "))
pasos = 0

print("\nSecuencia generada:")
print(numero) # Imprimimos el numero inicial antes de empezar a transformarlo

# 2. Iniciamos el ciclo indefinido hasta llegar al 1
while numero != 1:
    
    # 3. Logica de actualizacion de la variable
    if numero % 2 == 0:
        numero = numero // 2
    else:
        numero = (numero * 3) + 1
        
    # 4. Contamos que se realizo una operacion y mostramos el nuevo valor
    pasos = pasos + 1
    print(numero)

# 5. Reporte final fuera del ciclo
print("\n--- FIN DEL PROCESO ---")
print("La secuencia ha llegado al numero 1.")
print("Total de pasos necesarios:", pasos)
```
