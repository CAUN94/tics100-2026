```python
import random

tablero = []
n = 8

for i in range(n):
  fila = []
  for j in range(n):
    fila.append(0)
  tablero.append(fila)


for i in range(len(tablero)):
  for j in range(len(tablero[0])):
    print(tablero[i][j],end=" ")
  print()

print("-"*10)
print("BOMBAS")
print("-"*10)

bombas = 22

while bombas > 0:
  i_random = random.randint(0,len(tablero)-1)
  j_random = random.randint(0,len(tablero[0])-1)

  if tablero[i_random][j_random] != "#":
    tablero[i_random][j_random] = "#"
    bombas = bombas - 1

for i in range(len(tablero)):
  for j in range(len(tablero[0])):
    print(tablero[i][j],end=" ")
  print()

print("-"*10)
print("BOMBAS Y VALORES")
print("-"*10)

for i in range(len(tablero)):
  for j in range(len(tablero[0])):

    cant_bombas = 0

    for di in [-1,0,1]:
      for dj in [-1,0,1]:

        if di != 0 or dj != 0:

          if i + di >= 0 and i + di < len(tablero) and j + dj >= 0 and j + dj < len(tablero[j]):

            if tablero[i + di][j + dj] == "#":
              cant_bombas = cant_bombas + 1
    if tablero[i][j] == 0:
      tablero[i][j] = cant_bombas


for i in range(len(tablero)):
  for j in range(len(tablero[0])):
    print(tablero[i][j],end=" ")
  print()
```
