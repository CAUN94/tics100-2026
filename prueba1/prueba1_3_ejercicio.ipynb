**Enunciado del Ejercicio:**

Escriba un programa que solicite al usuario un número entero positivo y determine si dicho número es primo o no. 

Recuerde que, en matemáticas, un número es primo si es mayor a 1 y tiene exactamente dos divisores: el 1 y él mismo.

### Paso 1: Entender la divisibilidad matematicamente
Para que la computadora sepa si un número divide a otro de forma exacta, utilizamos el operador módulo (`%`). Este operador realiza una división y nos entrega solamente el resto. Si el resto es 0, significa que la división fue exacta y, por lo tanto, es un divisor válido.

Probemos el concepto viendo qué ocurre cuando dividimos el 5 entre 1 y entre 2.

````python
numero = 5

resto_con_uno = numero % 1
print("El resto de 5 dividido en 1 es:", resto_con_uno)

resto_con_dos = numero % 2
print("El resto de 5 dividido en 2 es:", resto_con_dos)
print("Como sobra 1, el 2 no es un divisor exacto de 5.")
````

### Paso 2: Contar los divisores manualmente
Un número es primo si tiene exactamente 2 divisores en total. Hagamos el cálculo de forma totalmente manual para el número 5.

Vamos a crear una variable que actúe como contador, partiendo en cero. Luego verificaremos la división exacta uno por uno, desde el 1 hasta el 5, usando condicionales `if`. Si la división es exacta, le sumamos 1 a nuestro contador.

````python
numero = 5
cantidad_divisores = 0

# Evaluamos manualmente los 5 posibles divisores
if numero % 1 == 0:
    cantidad_divisores = cantidad_divisores + 1

if numero % 2 == 0:
    cantidad_divisores = cantidad_divisores + 1

if numero % 3 == 0:
    cantidad_divisores = cantidad_divisores + 1

if numero % 4 == 0:
    cantidad_divisores = cantidad_divisores + 1

if numero % 5 == 0:
    cantidad_divisores = cantidad_divisores + 1

# Ahora evaluamos la condicion final para ser primo
print("El numero 5 tiene", cantidad_divisores, "divisores en total.")

if cantidad_divisores == 2:
    print("Por lo tanto, 5 es un numero primo.")
else:
    print("Por lo tanto, 5 NO es un numero primo.")
````

### Paso 3: Entender el ciclo for y range()
La logica anterior funciona perfecto para el numero 5. Pero, ¿que pasaria si nos piden evaluar si el numero 1047 es primo? Seria imposible escribir 1047 bloques de `if` de forma manual.

Para esto existe el ciclo `for` en combinacion con la funcion `range()`. Esta herramienta nos permite repetir una instruccion de forma automatica e ir cambiando un valor secuencialmente.

**Hint importante:** La funcion `range(inicio, fin)` genera numeros comenzando desde el valor de inicio, pero se detiene UN NUMERO ANTES del valor de fin. Es decir, si queremos contar del 1 al 5, nuestro rango debe ser `range(1, 6)`.

````python
print("Generando la secuencia de forma automatica:")

# La variable 'i' tomara un valor distinto en cada vuelta del ciclo
for i in range(1, 6):
    print("Vuelta actual, el valor de 'i' es:", i)
````

### Paso 4: Como se une el ciclo 'for' con el calculo
Antes de pasar a la solucion final, es vital entender el "puente" entre contar manualmente y usar el ciclo.

En el **Paso 2**, hicimos esto a mano para el numero 5:
```python
if numero % 1 == 0:
    cantidad_divisores = cantidad_divisores + 1
if numero % 2 == 0:
    cantidad_divisores = cantidad_divisores + 1
# ... y asi hasta el 5
```

Si observas bien, **la unica parte del codigo que cambia es el divisor (1, 2, 3, 4, 5)**.

Como vimos en el **Paso 3**, la variable `i` de un ciclo `for` va tomando exactamente esos valores uno por uno. Entonces, podemos fusionar ambas ideas. Reemplazamos los numeros manuales por la letra `i` y ponemos ese unico `if` dentro del ciclo:

```python
for i in range(1, numero + 1):
    if numero % i == 0:
        cantidad_divisores = cantidad_divisores + 1
```

El ciclo `for` se encargara de repetir ese `if` cambiando la `i` en cada vuelta, evaluando todos los divisores posibles desde el 1 hasta el numero ingresado (usamos `numero + 1` en el `range` para asegurarnos de que el ciclo incluya al propio numero en la division).

### Paso 5: Solucion final uniendo todo
Ahora si, resolvamos el ejercicio completo armando el rompecabezas.

Le pediremos un numero al usuario, crearemos nuestro contador de divisores en cero y dejaremos que el ciclo `for` haga toda la magia de validacion y suma que vimos en el paso anterior. 

Finalmente, evaluaremos si el contador de divisores es exactamente igual a 2 para declarar si es primo o no.

````python
# 1. Solicitamos el numero al usuario (debe ser un numero entero)
numero = int(input("Ingrese un numero entero positivo: "))

# 2. Creamos nuestra variable contadora en cero antes de que empiece el ciclo
cantidad_divisores = 0

# 3. Iniciamos el ciclo desde el 1 hasta el numero ingresado (por eso sumamos 1 al rango)
for i in range(1, numero + 1):
    
    # Evaluamos si 'i' es un divisor exacto del 'numero'
    if numero % i == 0:
        # Si la condicion es verdadera, aumentamos nuestro contador en uno
        cantidad_divisores = cantidad_divisores + 1

# 4. Mostramos los resultados obtenidos una vez terminado todo el ciclo for
print("\n--- RESULTADOS ---")
print("El numero", numero, "tiene un total de", cantidad_divisores, "divisores exactos.")

if cantidad_divisores == 2:
    print("Respuesta Final: El numero", numero, "ES PRIMO.")
else:
    print("Respuesta Final: El numero", numero, "NO ES PRIMO.")
````
