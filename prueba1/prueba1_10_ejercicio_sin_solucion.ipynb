**Enunciado del Ejercicio:**

En matemáticas existen varias series infinitas, y una de las más importantes es la serie de Leibniz para aproximar el valor de pi. Esta serie es representada de la siguiente manera:

pi = 4 * (1 - 1/3 + 1/5 - 1/7 + 1/9 ...)

Cree un programa que calcule el valor de pi utilizando la serie de Leibniz. Para ello, debe pedir al usuario el número de iteraciones con el que se calculará la aproximación de pi y luego calcular el valor de pi con ese número de iteraciones.

Por ejemplo, si se ingresa el valor de 5, la aproximación será calculada usando solo los primeros cinco términos, es decir pi aproximado a 4 * (1 - 1/3 + 1/5 - 1/7 + 1/9) lo cual resulta en 3.33968 aproximadamente.

### Paso 1: Descubrir el patron matematico y traducirlo a variables
Al mirar la formula `(1 - 1/3 + 1/5 - 1/7 + 1/9)`, podemos notar que hay dos cosas que cambian constantemente en cada nuevo termino:

1. **El denominador (el numero de abajo):** Comienza en 1 (porque 1/1 es 1), luego pasa a 3, luego a 5, luego a 7. Es decir, en cada paso **se le suma 2** al valor anterior.
2. **El signo:** Alterna entre positivo y negativo (+, -, +, -). El gran truco en programacion para hacer que algo alterne es tomar una variable y **multiplicarla por -1**. (Ej: 1 * -1 = -1. Luego -1 * -1 = 1).

Crearemos variables para representar estas partes moviles y ver como se actualizan.

### Paso 2: Ejecutar el patron manualmente y sumar
La formula de Leibniz nos pide ir sumando o restando estos terminos uno tras otro. 

Para lograr esto, usaremos la logica de acumulacion (como un contador). Tendremos una variable `suma_total = 0.0`. En cada paso calcularemos el `termino` matematico con las variables de nuestro paso anterior, se lo sumaremos al total, y luego prepararemos las variables para el siguiente paso.

### Paso 3: Entender la utilidad del ciclo for
Como pudiste ver en el Paso 2, escribimos exactamente las mismas 4 lineas de codigo tres veces seguidas para calcular solo 3 terminos.

Si el usuario quisiera calcular 5 terminos (como dice el enunciado), tendriamos que copiar y pegar ese bloque de codigo 5 veces. Si quisiera 1000 iteraciones (para que el valor de pi sea mas exacto), seria imposible hacerlo a mano.

Es por esto que esa seccion manual repetitiva la convertiremos a un ciclo `for`. 
La forma correcta de solucionar esto es **solicitar un numero al usuario** al principio (para saber la cantidad exacta de iteraciones) y luego **usar esa variable dentro de un `range()`**. De esta forma, la computadora copiara y ejecutara internamente nuestro bloque matematico la cantidad exacta de veces que el usuario solicito.

### Paso 4: Solucion final uniendo conceptos
Ahora armaremos el programa definitivo para resolver la Serie de Leibniz.

1. Solicitaremos al usuario el numero de iteraciones.
2. Crearemos nuestras 3 variables de inicio (`suma_total`, `denominador`, `signo`).
3. Crearemos el ciclo `for` en base a las iteraciones del usuario.
4. Dentro del ciclo, pondremos nuestra logica de calcular el termino, sumarlo, y actualizar el denominador y el signo para la siguiente vuelta.
5. Al terminar el ciclo `for`, **no debemos olvidar multiplicar toda la suma por 4**, tal como indica la formula principal: `pi = 4 * (Suma_de_terminos)`.
