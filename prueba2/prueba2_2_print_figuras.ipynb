# Guía: Ciclos for anidados para imprimir figuras

Esta guía explica paso a paso cómo funcionan dos ciclos for anidados cuando se usan para dibujar figuras con caracteres (asteriscos, letras, triángulos, etc.). Es material de apoyo para resolver ejercicios de "Ciclos anidados".

La idea central es una sola:

**El ciclo de afuera (`i`) controla las FILAS. El ciclo de adentro (`j`) controla las COLUMNAS de cada fila.**

Si dominas esa frase, dominas el 90% de los ejercicios de figuras.

## 1. Mecánica de los dos ciclos

### 1.1 El esqueleto básico

Toda figura rectangular se construye con esta estructura:

````python
numFilas = 3
numColumnas = 4

for i in range(numFilas):           # ciclo EXTERNO -> recorre las filas (eje vertical)
    for j in range(numColumnas):    # ciclo INTERNO -> recorre las columnas (eje horizontal)
        print("*", end="")          # imprime UN caracter, SIN salto de linea
    print()                         # al terminar la fila, SALTA a la linea siguiente
````

Dos detalles que siempre causan errores:

* `print("*", end="")` — el `end=""` evita el salto de línea automático. Así los asteriscos quedan uno al lado del otro (en la misma fila).
* `print()` (vacío, fuera del ciclo interno pero dentro del externo) — produce el salto de línea que cierra la fila y empieza la siguiente.

### 1.2 Tabla de ruteo (cómo "ejecutar a mano")

Para `numFilas = 3` y `numColumnas = 4`, el código de arriba imprime un rectángulo de 3 x 4:
```text
****
****
****
```
La tabla de ruteo muestra qué hace cada variable en cada paso. Es la herramienta principal para entender el código:

| i | j recorre | Qué se imprime | Línea resultante |
|---|---|---|---|
| 0 | 0, 1, 2, 3 | `* * * *` | `****` |
| 1 | 0, 1, 2, 3 | `* * * *` | `****` |
| 2 | 0, 1, 2, 3 | `* * * *` | `****` |

Observa: por cada valor de `i` (externo), el ciclo `j` (interno) hace todo su recorrido completo. El interno es más rápido: gira entero antes de que el externo avance un paso.

### 1.3 Recordatorio: cómo funciona range

`range` es el motor de los ciclos. Sus tres formas:

```python
range(5)          # 0, 1, 2, 3, 4     -> empieza en 0, NO incluye el 5
range(2, 6)       # 2, 3, 4, 5        -> desde 2 hasta 6 (sin incluir 6)
range(0, 10, 2)   # 0, 2, 4, 6, 8     -> con paso de 2
```

**Regla de oro:** `range(n)` genera `n` números, de 0 a `n-1`. El valor final nunca se incluye. Este es el error más común: usar `range(1, n)` cuando se quería `range(0, n)` o `range(n)`, lo que produce una fila o columna de menos.

```python
range(1, 5) # 1, 2, 3, 4 -> SOLO 4 numeros, falta uno!
range(0, 5) # 0, 1, 2, 3, 4 -> 5 numeros, correcto
```

### 1.4 Figura cuadrada vs. rectangular

**Cuadrado:** ambos ciclos usan el mismo límite.

````python
n = 5 # int(input("Tamaño: "))

for i in range(n):
    for j in range(n):
        print("*", end="")
    print()
````

**Rectángulo:** límites distintos para filas y columnas.

````python
filas = 3    # int(input("Filas: "))
columnas = 6 # int(input("Columnas: "))

for i in range(filas):
    for j in range(columnas):
        print("*", end="")
    print()
````

## 2. Imprimir filas y columnas específicas

Aquí está el verdadero truco de las figuras: no siempre imprimimos un `*`. A veces imprimimos `*` y a veces un espacio `" "`, dependiendo de en qué fila y columna estamos. Para decidirlo usamos un `if` dentro del ciclo interno.

La pregunta clave que debes hacerte en cada celda es:

*"¿Estoy en una posición que debe llevar asterisco, o en una que debe llevar espacio?"*

### 2.1 Identificar bordes con los índices

Para una figura de `n x n`, los índices `i` (fila) y `j` (columna) van de 0 a `n-1`. Eso da nombres exactos a cada borde:

| Posición | Condición |
|---|---|
| Primera fila | `i == 0` |
| Última fila | `i == n-1` |
| Primera columna | `j == 0` |
| Última columna | `j == n-1` |
| Fila del medio | `i == n//2` |
| Columna del medio | `j == n//2` |

`//` es la división entera. Para `n = 7`, `n//2 = 3`, que es justo el índice central (hay 3 a cada lado). El medio "exacto" solo existe cuando `n` es impar.

### 2.2 Ejemplo: solo la primera y la última fila
Para imprimir un asterisco únicamente en la primera y última fila (y espacios en el resto):

````python
n = 5 # int(input("Tamaño: "))

for i in range(n):
    for j in range(n):
        if i == 0 or i == n-1: # primera O ultima fila
            print("*", end="")
        else:
            print(" ", end="")
    print()
````

### 2.3 Ejemplo: marco / borde completo (la "caja")
Un asterisco en cualquier borde (primera/última fila o primera/última columna), espacio en el interior:

````python
n = 5 # int(input("Tamaño: "))

for i in range(n):
    for j in range(n):
        if i == 0 or i == n-1 or j == 0 or j == n-1:
            print("*", end="")
        else:
            print(" ", end="")
    print()
````

### 2.4 Ejemplo: la letra C
La C es: primera fila completa + última fila completa + primera columna completa. No lleva la última columna en el medio. Combina lo anterior quitando la condición `j == n-1`:

````python
filas = 5    # int(input("Ingrese número de filas: "))
columnas = 7 # int(input("Ingrese número de columnas: "))

for i in range(filas):
    for j in range(columnas):
        if i == 0 or i == filas-1 or j == 0:
            print("*", end="")
        else:
            print(" ", end="")
    print()
````

### 2.5 Ejemplo: la columna y la fila del medio (una cruz +)

````python
n = 5 # int(input("Tamaño (impar): "))
medio = n // 2

for i in range(n):
    for j in range(n):
        if i == medio or j == medio:
            print("*", end="")
        else:
            print(" ", end="")
    print()
````

### 2.6 Resumen de condiciones

| Quiero pintar… | Condición en el if |
|---|---|
| Primera fila | `i == 0` |
| Última fila | `i == n-1` |
| Primera columna | `j == 0` |
| Última columna | `j == n-1` |
| Fila del medio | `i == n//2` |
| Columna del medio | `j == n//2` |
| Marco (caja) | `i==0 or i==n-1 or j==0 or j==n-1` |
| Letra C | `i==0 or i==n-1 or j==0` |
| Cruz + | `i==n//2 or j==n//2` |

## 3. Imprimir diagonales

Las diagonales son el paso más importante para resolver figuras como la X, el triángulo o el rombo. El secreto está en comparar `i` con `j` (fila con columna).

### 3.1 La diagonal principal: `i == j`
La diagonal que baja de arriba-izquierda hacia abajo-derecha cumple que el número de fila es igual al número de columna: `(0,0)`, `(1,1)`, `(2,2)`, ...

````python
n = 5 # int(input("Tamaño: "))

for i in range(n):
    for j in range(n):
        if i == j: # diagonal principal (\)
            print("*", end="")
        else:
            print(" ", end="")
    print()
````

### 3.2 La diagonal secundaria: `i + j == n-1`
La diagonal que baja de arriba-derecha hacia abajo-izquierda. En ella, la suma de fila y columna siempre da `n-1`: `(0,4)`, `(1,3)`, `(2,2)`, `(3,1)`, `(4,0)` para `n = 5`.

````python
n = 5 # int(input("Tamaño: "))

for i in range(n):
    for j in range(n):
        if i + j == n-1: # diagonal secundaria (/)
            print("*", end="")
        else:
            print(" ", end="")
    print()
````

### 3.3 Por qué funciona (tabla de ruteo de la diagonal secundaria)

| i | n-1-i = j con * | Línea (n=5) |
|---|---|---|
| 0 | 4 | `    *` |
| 1 | 3 | `   * ` |
| 2 | 2 | `  * ` |
| 3 | 1 | ` * ` |
| 4 | 0 | `* ` |

En cada fila el asterisco se corre una columna a la izquierda, porque `j = n-1-i` disminuye al aumentar `i`. Esa resta es la que produce la inclinación.

### 3.4 Las dos diagonales juntas = la X
Una X es simplemente la diagonal principal más la secundaria. Se combinan las dos condiciones con `or`:

````python
n = 7 # int(input("Ingrese el tamaño: "))

for i in range(n):
    for j in range(n):
        if i == j or i + j == n-1:
            print("*", end="")
        else:
            print(" ", end="")
    print()
````

Las dos diagonales se cruzan justo en el centro (`i == j` y `i + j == n-1` ocurren a la vez cuando `n` es impar), por eso el medio tiene un solo asterisco.

### 3.5 Tip: diagonales con desigualdades (triángulos rellenos)
Cambiando `==` por `<` o `>` se rellena un lado completo en vez de una línea:

````python
# Triángulo inferior izquierdo: pinta donde j <= i
n = 5 # int(input("Tamaño: "))

for i in range(n):
    for j in range(n):
        if j <= i:
            print("*", end="")
        else:
            print(" ", end="")
    print()
````

| Quiero… | Condición |
|---|---|
| Diagonal principal `\` | `i == j` |
| Diagonal secundaria `/` | `i + j == n-1` |
| Las dos (X) | `i == j or i + j == n-1` |
| Triángulo bajo la diagonal | `j <= i` |
| Triángulo sobre la diagonal | `j >= i` |

## 4. Estrategia para resolver cualquier figura

1. **Cuenta las filas y columnas.** ¿Es cuadrada (`n x n`) o rectangular (filas x columnas)? Eso define los dos `range`.
2. **Dibuja la figura en papel** con una grilla numerada (filas `i` y columnas `j` desde 0). Anota las coordenadas de cada `*`.
3. **Busca el patrón** entre `i` y `j` de las celdas con asterisco: ¿son iguales (`==`)? ¿suman una constante? ¿es un borde (`0` o `n-1`)?
4. **Traduce el patrón** a un `if` dentro del ciclo interno: `*` si se cumple, `" "` si no.
5. **Rutea a mano** con un valor pequeño (`n = 3` o `n = 5`) y compara con la figura pedida.

## 5. Errores frecuentes

* `range(1, n)` en vez de `range(n)` — falta una fila o columna. Empieza siempre en 0 salvo que tengas una razón.
* **Olvidar `end=""`** en el print del `*` — cada asterisco cae en su propia línea (figura “vertical” rota).
* **Poner el `print()` vacío mal ubicado** — debe ir dentro del ciclo externo (`i`) y fuera del interno (`j`). Si lo pones dentro del interno, saltas de línea en cada columna.
* **Confundir `n` con `n-1`** en los bordes — el último índice válido es `n-1`, no `n`.
* **Usar `=` en vez de `==` en el if** — `=` asigna, `==` compara. En Python `if i = 0:` da error de sintaxis.

### Recomendaciones

* Resuelve primero en papel con la grilla numerada antes de tocar el teclado. La mayoría de los errores se ven a mano.
* Haz la tabla de ruteo de cada figura con un `n` pequeño, anotando el output como lo vimos.
* Prueba tu código con valores pares e impares de `n`: muchas figuras (X, cruz, rombo) solo quedan perfectas con `n` impar.
