## 1. El Kit de Herramientas Básico para Matrices

Vamos a construir dos funciones esenciales que usaremos repetidamente a lo largo de esta guía:
1. Una función que construye y **retorna** una matriz llena de números aleatorios.
2. Una función que no retorna nada, sino que **imprime** en pantalla (formato cuadrícula) cualquier matriz que le entreguemos.

````python
import random

# 1. FÁBRICA DE MATRICES
def crear_matriz_aleatoria(filas, columnas):
    matriz = []
    for i in range(filas):
        fila_temporal = []
        for j in range(columnas):
            # Generamos un valor aleatorio entre 0 y 9
            valor = random.randint(0, 9)
            fila_temporal.append(valor)
        matriz.append(fila_temporal)
    return matriz

# 2. IMPRESORA DE MATRICES
def mostrar_matriz(matriz):
    # Recibimos una matriz y la imprimimos con un doble for
    for i in range(len(matriz)):
        for j in range(len(matriz[i])):
            print(matriz[i][j], end=" ")
        print()

# ==========================================
# EJEMPLO DE USO
# ==========================================
print("Creando matriz de 3x4...")
mi_nueva_matriz = crear_matriz_aleatoria(3, 4)

print("Mostrando la matriz generada:")
mostrar_matriz(mi_nueva_matriz)
````

## Ejercicio 1: Transponer una Matriz

**Enunciado:**
Crea una función llamada `transponer_matriz(matriz)` que tome una matriz de `n * m` y la **trasponga**. 
Transponer significa que las filas se convierten en columnas y las columnas en filas.

**Ejemplo esperado:**
```text
Original:
1 2 3 
4 5 6 

Transpuesta:
1 4 
2 5 
3 6 
```

````python
# --- RESPUESTA ---
def transponer_matriz(matriz):
    filas_original = len(matriz)
    columnas_original = len(matriz[0])
    
    matriz_t = []
    
    # El nuevo exterior iterará sobre las columnas originales
    for j in range(columnas_original):
        nueva_fila = []
        # El interior iterará sobre las filas originales
        for i in range(filas_original):
            nueva_fila.append(matriz[i][j])
        matriz_t.append(nueva_fila)
        
    return matriz_t

# ==========================================
# PRUEBAS (Modifica o prueba tu código aquí)
# ==========================================
matriz_prueba = [
    [1, 2, 3], 
    [4, 5, 6]
]
print("Original:")
mostrar_matriz(matriz_prueba)

print("\nTranspuesta:")
resultado = transponer_matriz(matriz_prueba)
mostrar_matriz(resultado)
````

## Ejercicio 2: Analizador Estadístico de Matrices

**Enunciado:**
Crea una función `analizar_matriz(matriz)` que reciba una matriz con puros números. La función debe recorrerla para encontrar el **menor valor**, el **mayor valor** y calcular el **promedio** de todos los valores. Finalmente, la función debe imprimir esta información ordenada (esta función no necesita usar `return`, solo `print`).

````python
# --- RESPUESTA ---
def analizar_matriz(matriz):
    menor = matriz[0][0]
    mayor = matriz[0][0]
    suma = 0
    cantidad_datos = 0
    
    for i in range(len(matriz)):
        for j in range(len(matriz[i])):
            valor = matriz[i][j]
            
            if valor < menor:
                menor = valor
            if valor > mayor:
                mayor = valor
                
            suma += valor
            cantidad_datos += 1
            
    promedio = suma / cantidad_datos
    
    print("--- ANÁLISIS DE LA MATRIZ ---")
    print(f"Menor valor: {menor}")
    print(f"Mayor valor: {mayor}")
    print(f"Promedio: {promedio}")

# ==========================================
# PRUEBAS (Modifica o prueba tu código aquí)
# ==========================================
matriz_datos = [
    [10, 5, 2], 
    [8, 20, 15]
]
analizar_matriz(matriz_datos)
````

## Ejercicio 3: Limpiador de Tipos de Datos

**Enunciado:**
Imagina que descargaste datos de internet y la matriz llegó sucia: tiene números enteros, pero también números guardados como texto (ej. `"5"`) y números con decimales (ej. `3.8`).

Crea la función `limpiar_a_enteros(matriz)` que recorra la matriz, tome cada valor, lo convierta forzosamente a un número entero usando `int()`, y devuelva la **misma matriz** (ya modificada) mediante `return`.

````python
# --- RESPUESTA ---
def limpiar_a_enteros(matriz):
    for i in range(len(matriz)):
        for j in range(len(matriz[i])):
            # Sobreescribimos el valor convirtiéndolo a entero
            matriz[i][j] = int(matriz[i][j])
    
    return matriz

# ==========================================
# PRUEBAS (Modifica o prueba tu código aquí)
# ==========================================
matriz_sucia = [
    [1, "2", 3.5], 
    ["4", 5.9, "6"]
]
print("Matriz Sucia:", matriz_sucia)

matriz_limpia = limpiar_a_enteros(matriz_sucia)
print("Matriz Limpia:", matriz_limpia)
````

## Ejercicio 4: Buscador de Ocurrencias

**Enunciado:**
Crea una función llamada `contar_ocurrencias(matriz, objetivo)`. Se le entrega una matriz y un número objetivo. La función debe recorrer toda la matriz, contar cuántas veces aparece ese número exacto y **retornar** el conteo final.

````python
# --- RESPUESTA ---
def contar_ocurrencias(matriz, objetivo):
    contador = 0
    for i in range(len(matriz)):
        for j in range(len(matriz[i])):
            if matriz[i][j] == objetivo:
                contador += 1
    return contador

# ==========================================
# PRUEBAS (Modifica o prueba tu código aquí)
# ==========================================
tablero = [
    [5, 2, 5], 
    [1, 5, 8], 
    [5, 5, 3]
]
veces = contar_ocurrencias(tablero, 5)
print(f"El número 5 aparece {veces} veces en el tablero.")
````

## Ejercicio 5: El Creador de Buscaminas

**Enunciado:**
Crearemos dos funciones para un juego de Buscaminas:

1. `crear_buscaminas(n, m, cantidad_bombas)`: Genera una matriz de `n` filas por `m` columnas llena de ceros (`0`). Luego, usando `random`, coloca la `cantidad_bombas` indicadas usando el símbolo `'*'`. Finalmente **retorna** la matriz, no es necario ver los valores numericos de los vecinos.
2. `mostrar_con_coordenadas(matriz)`: Imprime la matriz. Pero además, **si encuentra una bomba**, debe imprimir un mensaje de alerta indicando la coordenada exacta `[i][j]` donde la vio.

````python
# --- RESPUESTA ---
import random

def crear_buscaminas(n, m, cantidad_bombas):
    tablero = []
    for i in range(n):
        fila = []
        for j in range(m):
            fila.append(0)
        tablero.append(fila)
        
    bombas_colocadas = 0
    while bombas_colocadas < cantidad_bombas:
        i_azar = random.randint(0, n - 1)
        j_azar = random.randint(0, m - 1)
        
        if tablero[i_azar][j_azar] != '*':
            tablero[i_azar][j_azar] = '*'
            bombas_colocadas += 1
            
    return tablero

def mostrar_con_coordenadas(matriz):
    print("--- MAPA BUSCAMINAS ---")
    for i in range(len(matriz)):
        for j in range(len(matriz[i])):
            print(matriz[i][j], end=" ")
        print()
        
    print("\n--- UBICACIÓN DE LAS BOMBAS ---")
    for i in range(len(matriz)):
        for j in range(len(matriz[i])):
            if matriz[i][j] == '*':
                print(f"¡Peligro! Hay una bomba en la coordenada [{i}][{j}]")

# ==========================================
# PRUEBAS (Modifica o prueba tu código aquí)
# ==========================================
mi_juego = crear_buscaminas(4, 5, 3)
mostrar_con_coordenadas(mi_juego)
````

## Ejercicio 7: Producto Punto y Producto Cruz (Álgebra de Vectores)

**Explicación Matemática:**
En el álgebra lineal, los "vectores" pueden representarse computacionalmente como listas de números. Existen dos multiplicaciones fundamentales cuando tenemos dos listas del mismo tamaño:

* **Producto Punto (Escalar):** Multiplica cada elemento de una lista con el elemento en la misma posición de la otra lista, y luego suma todos esos resultados parciales. El resultado final es un **único número**.
* **Producto Cruz (Vectorial):** Se usa típicamente en 3 dimensiones (listas de tamaño 3). Genera un **nuevo vector (nueva lista)** que es perpendicular a los dos originales. Su fórmula matemática cruzada para dos vectores $A$ y $B$ es:
  * Posición X: `(A[1] * B[2]) - (A[2] * B[1])`
  * Posición Y: `(A[2] * B[0]) - (A[0] * B[2])`
  * Posición Z: `(A[0] * B[1]) - (A[1] * B[0])`

**Enunciado:**
Crea dos funciones:
1. `producto_punto(v1, v2)`: Recibe dos listas del mismo tamaño, recorre sus índices para multiplicar las posiciones iguales, acumula la suma de esas multiplicaciones y **retorna** la suma total.
2. `producto_cruz(v1, v2)`: Recibe dos listas de exactamente 3 elementos, aplica la fórmula matemática cruzada para calcular las componentes X, Y, Z, y **retorna** una lista nueva con los 3 resultados, asuma vectores de 3 dimensiones.

````python
# --- RESPUESTA ---
def producto_punto(v1, v2):
    suma_total = 0
    # Como ambas listas tienen el mismo tamaño, usamos el len de cualquiera de ellas
    for i in range(len(v1)):
        suma_total += (v1[i] * v2[i])
        
    return suma_total

def producto_cruz(v1, v2):
    # Asumimos estrictamente vectores de 3 dimensiones (índices 0, 1 y 2)
    x = (v1[1] * v2[2]) - (v1[2] * v2[1])
    y = (v1[2] * v2[0]) - (v1[0] * v2[2])
    z = (v1[0] * v2[1]) - (v1[1] * v2[0])
    
    # Creamos la nueva lista (el nuevo vector)
    vector_resultante = [x, y, z]
    return vector_resultante

# ==========================================
# PRUEBAS (Modifica o prueba tu código aquí)
# ==========================================
vector_A = [1, 2, 3]
vector_B = [4, 5, 6]

resultado_punto = producto_punto(vector_A, vector_B)
print(f"El Producto Punto entre {vector_A} y {vector_B} es: {resultado_punto}")

resultado_cruz = producto_cruz(vector_A, vector_B)
print(f"El Producto Cruz entre {vector_A} y {vector_B} es: {resultado_cruz}")
````

## Ejemplos Clásicos de Uso de Funciones y Matrices

Para finalizar, aquí tienes algunos ejemplos comunes de cómo las funciones nos salvan la vida al empaquetar lógicas muy largas que usamos todo el tiempo.

````python
# Ejemplo Clásico 1: Sumar dos matrices matemáticamente
# (Se asume que ambas tienen el mismo tamaño)
def sumar_matrices(matriz_A, matriz_B):
    matriz_resultado = []
    for i in range(len(matriz_A)):
        fila_nueva = []
        for j in range(len(matriz_A[i])):
            # Sumamos la misma coordenada de ambas matrices
            suma = matriz_A[i][j] + matriz_B[i][j]
            fila_nueva.append(suma)
        matriz_resultado.append(fila_nueva)
    return matriz_resultado

m1 = [
    [1, 2], 
    [3, 4]
]
m2 = [
    [10, 20], 
    [30, 40]
]
print("Suma de Matrices:")
mostrar_matriz(sumar_matrices(m1, m2))
print("-" * 20)


# Ejemplo Clásico 2: Extraer una columna específica
def extraer_columna(matriz, indice_columna):
    columna_extraida = []
    for i in range(len(matriz)):
        # Guardamos el dato manteniendo la columna fija y bajando de fila en fila
        columna_extraida.append(matriz[i][indice_columna])
    return columna_extraida

datos = [
    [1, 9, 3], 
    [4, 9, 6], 
    [7, 9, 9]
]
print("Extrayendo la columna del medio (índice 1):")
print(extraer_columna(datos, 1))
````
