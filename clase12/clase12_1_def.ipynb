# Creación y Uso de Funciones (`def`)

Hasta ahora hemos escrito programas que se ejecutan de arriba hacia abajo una sola vez. Pero, ¿qué pasa si tenemos un bloque de código muy útil (como dibujar un tablero o calcular un promedio) y queremos usarlo muchas veces en distintas partes de nuestro programa sin tener que copiar y pegar el código completo?

Aquí entran las **funciones**. Una función es como una pequeña "fábrica" o "mini-programa" dentro de nuestro código. Le damos un nombre, guardamos instrucciones adentro, y luego la podemos "llamar" o invocar cuantas veces queramos.

### La Estructura Básica
Para crear una función usamos la palabra reservada `def` (de definir). Idealmente, todas las funciones se deben escribir y definir **al principio de tu archivo de código**, para que cuando el programa principal empiece a correr, ya conozca que estas "fábricas" existen.

La estructura mínima es:
```python
def NOMBRE():
    # Aquí va el código indentado (con sangría)
    # que queremos que se ejecute
```

### Función Sin Argumentos

La forma más simple de una función es aquella que siempre hace exactamente lo mismo. No necesita que le entreguemos información externa (no tiene "argumentos" dentro del paréntesis).

````python
# DEFINICIÓN DE LA FUNCIÓN (La fábrica)
def saludar_simple():
    print("¡Hola! Bienvenido al sistema.")

# PROGRAMA PRINCIPAL
# Aquí "llamamos" a la función para que trabaje
print("Iniciando programa...")
saludar_simple()
saludar_simple() # Podemos llamarla varias veces
print("Fin del programa.")
````

### Función Con Un Argumento

Podemos hacer que nuestra fábrica sea más inteligente pasándole datos o "materiales" para trabajar. A estos datos los llamamos **argumentos** o **parámetros**, y van dentro de los paréntesis.

Crearemos una función que reciba un número, y que imprima nuestro saludo esa cantidad exacta de veces.

````python
# La función ahora exige que le entreguemos una variable 'n'
def saludar_multiplicado(n):
    for i in range(n):
        print("¡Hola! Bienvenido al sistema.")

# PROGRAMA PRINCIPAL
print("Llamando a la función con el número 3:")
saludar_multiplicado(3)

print("\nLlamando a la función con el número 1:")
saludar_multiplicado(1)
````

### Función Con Múltiples Argumentos

Podemos entregar tantos argumentos como necesitemos separándolos con comas. 
Crearemos una función que reciba dos valores: una cantidad de pisos (`n`) y un símbolo (`caracter`). Con estos datos dibujará una pirámide a medida.

````python
def hacer_piramide(pisos, caracter):
    print(f"\n--- Pirámide de {pisos} pisos usando '{caracter}' ---")
    for i in range(1, pisos + 1):
        for j in range(i):
            print(caracter, end=" ")
        print()

# PROGRAMA PRINCIPAL
hacer_piramide(4, "#")
hacer_piramide(3, "+")
````

## El poder de `return` (Devolver un valor)

Hasta ahora, nuestras funciones solo han usado `print()`. Hacen un trabajo y lo muestran en la pantalla. Pero en programación profesional, rara vez queremos que una función imprima cosas. Lo que queremos es que haga un cálculo y **nos devuelva el resultado** para poder seguir trabajando con él.

Para esto usamos la palabra mágica `return`. Cuando una función ejecuta un `return`, expulsa ese dato hacia el programa principal, pero NO lo imprime. Nosotros debemos "atrapar" ese dato en una variable para poder usarlo después.

### Función con `return` básico

Haremos una función sin argumentos ni prints. Simplemente devolverá un texto.

````python
def obtener_saludo():
    return "hola mundo"

# PROGRAMA PRINCIPAL
# Si solo la llamo así: obtener_saludo() -> No se verá nada en pantalla.

# Debemos ATRAPAR lo que la función nos lanza en una variable:
mensaje_secreto = obtener_saludo()

# Ahora podemos usar esa variable
print("El mensaje que nos entregó la fábrica fue:", mensaje_secreto)
````

### Función con `return` usando parámetros

Haremos una función con un argumento (un número). La función no imprimirá nada; su único trabajo es construir una cadena (un texto) llena de gatos (`#`) de ese largo, y devolverla.

````python
def generador_de_gatos(cantidad):
    texto_gatos = ""
    for i in range(cantidad):
        texto_gatos += "#"
    
    # Una vez terminado el ciclo, entregamos el texto completo
    return texto_gatos

# PROGRAMA PRINCIPAL
# Podemos atrapar el valor y luego imprimirlo
linea_pequena = generador_de_gatos(3)
print(linea_pequena)

# O podemos imprimir directamente lo que la función nos devuelve
print(generador_de_gatos(10))
````

### Función con `return` analizando listas

Las funciones son geniales para analizar datos. Haremos una función que reciba dos cosas: una lista de números y un número específico (el objetivo). La función contará y nos **devolverá** cuántas veces apareció ese número en la lista.

````python
def contar_apariciones(lista_evaluar, numero_buscado):
    contador = 0
    for num in lista_evaluar:
        if num == numero_buscado:
            contador += 1
            
    return contador

# PROGRAMA PRINCIPAL
mis_notas = [4, 5, 4, 7, 4, 6, 5]

cantidad_cuatros = contar_apariciones(mis_notas, 4)
print(f"El número 4 apareció {cantidad_cuatros} veces en el registro.")
````
