## Enunciado del Ejercicio: Ordenamiento de 3 Números
Escribe un programa en Python que solicite al usuario 3 números y los muestre por pantalla ordenados de mayor a menor.

### Análisis y Estrategia de Resolución:
Este es un clásico ejercicio de algoritmia. Aunque Python posee funciones integradas como `sort()` o `max()`, el objetivo pedagógico aquí es desarrollar tu lógica de programación usando **operadores relacionales** (`>`, `<`) y **operadores lógicos** (`and`).

El problema lo dividiremos en las siguientes partes:
1. **Captura de datos:** Necesitamos tres variables independientes en memoria para guardar los números.
2. **Determinación del Mayor (Lógica principal):** Para saber si un número es el mayor de todos, debe ser mayor que el segundo **Y** mayor que el tercero al mismo tiempo. Evaluaremos cada número como candidato a ser el mayor.
3. **Determinación del Medio y Menor (Anidación):** Una vez que sabemos quién es el mayor, solo nos queda comparar los dos números restantes para ver quién ocupa el segundo y tercer lugar.

### Resolución Paso a Paso (Bloques de Código):

#### Paso 1: Inicialización y Captura de Datos
Primero, preparamos nuestro entorno solicitando al usuario los tres valores numéricos. Usaremos `int()` asumiendo que trabajaremos con números enteros.

```python
# Solicitamos los tres números al usuario y los convertimos a enteros
print("Ingresa número 1:")
N1 = int(input())

print("Ingresa número 2:")
N2 = int(input())

print("Ingresa número 3:")
N3 = int(input())

# Creamos tres variables vacías (o en 0) para guardar los resultados ordenados
mayor1 = 0 # Guardará el número más grande
mayor2 = 0 # Guardará el número del medio
mayor3 = 0 # Guardará el número más pequeño
```

#### Paso 2: Evaluar si el Primer Número (N1) es el Mayor
Aquí aplicamos el operador lógico `and`. Evaluamos el escenario donde `N1` gana el primer lugar.

```python
# Verificamos si N1 es estrictamente mayor que N2 y que N3
if (N1 > N2 and N1 > N3):
    # Si la condición se cumple, N1 es el más grande de todos
    mayor1 = N1
    
    # Ahora solo queda comparar N2 y N3 para asignar el segundo y tercer lugar
    if (N2 > N3):
        mayor2 = N2
        mayor3 = N3
    else:
        # Si N2 no es mayor que N3, entonces N3 va al medio y N2 al final
        mayor2 = N3
        mayor3 = N2
```

#### Paso 3: Evaluar si el Segundo (N2) o Tercer Número (N3) son el Mayor
Continuamos con la misma lógica secuencial para los otros dos candidatos.

```python
# Verificamos el escenario donde N2 es el más grande
if (N2 > N1 and N2 > N3):
    mayor1 = N2
    # Comparamos los restantes: N1 y N3
    if (N1 > N3):
        mayor2 = N1
        mayor3 = N3
    else:
        mayor2 = N3
        mayor3 = N1

# Finalmente, verificamos el escenario donde N3 es el más grande
if (N3 > N1 and N3 > N2):
    mayor1 = N3
    # Comparamos los restantes: N1 y N2
    if (N1 > N2):
        mayor2 = N1
        mayor3 = N2
    else:
        mayor2 = N2
        mayor3 = N1
```

#### Paso 4: Salida de Resultados
Una vez que las variables `mayor1`, `mayor2` y `mayor3` han sido asignadas correctamente, mostramos el resultado.

```python
# Imprimimos las variables en el orden establecido
print("El orden de mayor a menor es:", mayor1, mayor2, mayor3)
```

#### Código Completo

## Enunciado del Ejercicio: Ecuación Cuadrática
Escribe el código de un algoritmo que solicita los valores $a$, $b$ y $c$ de una ecuación cuadrática de la forma:
$$y = ax^2 + bx + c$$

y luego despliegue por pantalla los siguientes mensajes:
* "No hay soluciones reales" (si el discriminante es negativo)
* "Hay una solución real y es igual a " y desplegar el valor (cuando el discriminante es igual a 0)
* "Hay dos soluciones reales iguales a " y desplegar sus valores (cuando el discriminante es positivo).

Recuerda que las soluciones se calculan con la siguiente fórmula:
$$x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}$$

### Análisis y Estrategia de Resolución:
Este problema combina matemáticas puras con control de flujo. La clave del problema no es la fórmula completa de inmediato, sino el **discriminante** ($b^2 - 4ac$), el cual está dentro de la raíz cuadrada.
1. **Variables de Entrada:** Requerimos los coeficientes $a$, $b$ y $c$.
2. **Cálculo Previo (Optimizador):** Calcularemos el discriminante primero y lo guardaremos en una variable. Esto hace el código más limpio y eficiente.
3. **Árbol de Decisión:**
   * Si `discriminante < 0`: Mostramos mensaje de error (no se puede calcular raíz par de un negativo en los reales).
   * Si `discriminante == 0`: Calculamos una sola raíz, ya que la parte $\pm 0$ no altera el resultado.
   * Si `discriminante > 0`: Calculamos dos raíces, una usando `+` y otra usando `-`.
4. **Operación Matemática:** Para la raíz cuadrada en Python, podemos elevar a la potencia de $0.5$ (`** 0.5`).

### Resolución Paso a Paso (Bloques de Código):

#### Paso 1: Captura de Coeficientes y Cálculo del Discriminante
Solicitamos los datos y calculamos la pieza fundamental que determinará qué "camino" tomará nuestro programa.

```python
print('Ingrese los valores a, b y c de una ecuación cuadrática')

# Solicitamos los coeficientes convirtiéndolos a flotantes para mayor precisión
print('Ingrese a:')
a = float(input())

print('Ingrese b:')
b = float(input())

print('Ingrese c:')
c = float(input())

# Calculamos el discriminante: b al cuadrado menos 4 por a por c
disc = (b ** 2) - (4 * a * c)
```

#### Paso 2: Evaluación del Discriminante y Cálculo de Raíces
Usaremos estructuras condicionales anidadas (un `if` dentro del `else`) para abarcar los 3 escenarios posibles.

```python
# Escenario 1: El discriminante es negativo (Raíz imaginaria)
if disc < 0:
    print('No hay soluciones reales')

# Si no es negativo, puede ser cero o positivo
else:
    # Escenario 2: El discriminante es exactamente cero
    if disc == 0:
        # Calculamos la única solución posible. La raíz de 0 es 0, así que la omitimos.
        x1 = (-b) / (2 * a)
        print('Hay una solución real y esta es: x1 =', x1)
        
    # Escenario 3: El discriminante es positivo (Por descarte, entra al último else)
    else:
        # Calculamos la primera raíz usando el símbolo de suma (+)
        # Nota: (disc ** 0.5) equivale a calcular la raíz cuadrada del discriminante
        x1 = (-b + (disc ** 0.5)) / (2 * a)
        
        # Calculamos la segunda raíz usando el símbolo de resta (-)
        x2 = (-b - (disc ** 0.5)) / (2 * a)
        
        # Mostramos ambas soluciones
        print('Hay dos soluciones reales. Estas son: x1 =', x1, 'y x2 =', x2)
```

Ahora incluimos el código completo

```python
#### Código Completo

print('Ingrese los valores a, b y c de una ecuación cuadrática')

# Solicitamos los coeficientes convirtiéndolos a flotantes para mayor precisión
print('Ingrese a:')
a = float(input())

print('Ingrese b:')
b = float(input())

print('Ingrese c:')
c = float(input())

# Calculamos el discriminante: b al cuadrado menos 4 por a por c
disc = (b ** 2) - (4 * a * c)

# Escenario 1: El discriminante es negativo (Raíz imaginaria)
if disc < 0:
    print('No hay soluciones reales')

# Si no es negativo, puede ser cero o positivo
else:
    # Escenario 2: El discriminante es exactamente cero
    if disc == 0:
        # Calculamos la única solución posible. La raíz de 0 es 0, así que la omitimos.
        x1 = (-b) / (2 * a)
        print('Hay una solución real y esta es: x1 =', x1)
        
    # Escenario 3: El discriminante es positivo (Por descarte, entra al último else)
    else:
        # Calculamos la primera raíz usando el símbolo de suma (+)
        # Nota: (disc ** 0.5) equivale a calcular la raíz cuadrada del discriminante
        x1 = (-b + (disc ** 0.5)) / (2 * a)
        
        # Calculamos la segunda raíz usando el símbolo de resta (-)
        x2 = (-b - (disc ** 0.5)) / (2 * a)
        
        # Mostramos ambas soluciones
        print('Hay dos soluciones reales. Estas son: x1 =', x1, 'y x2 =', x2)
```
