# Estructuras de Control Iterativas (Ciclos Básicos)

Escribe un algoritmo que resuelva el siguiente problema: mostrar por pantalla los primeros 10 números pares. Resuelve este problema de 2 maneras distintas: usando ciclo `while` y ciclo `for`.

### Análisis y Estrategia de Resolución:
Para resolver este problema, necesitamos generar una secuencia numérica. Sabemos que los números pares positivos comienzan en 2 y avanzan de 2 en 2 (2, 4, 6, 8...).
Como el problema nos pide exactamente los **primeros 10**, necesitamos llevar el control de cuántas iteraciones llevamos.
* **Estrategia con `while`:** Necesitaremos dos variables. Una que actúe como "contador" para saber cuándo detenernos (al llegar a 10) y otra que almacene el número par a imprimir.
* **Estrategia con `for`:** Podemos aprovechar la función `range()`, que es ideal para cuando sabemos exactamente cuántas iteraciones queremos (10). Multiplicando el índice de la iteración por 2, obtendremos el número par sin necesidad de variables adicionales.

### Resolución Paso a Paso (Bloques de Código):

**Solución 1: Usando la estructura `while`**

Primero, inicializamos nuestras variables de estado antes de entrar al ciclo.

```python
# Inicializamos el contador en 1, ya que vamos a contar desde el 1er hasta el 10mo número.
contador = 1

# Inicializamos nuestra variable 'par' con el primer número par que queremos mostrar.
par = 2
```

A continuación, construimos el ciclo. La condición debe evaluar si aún no hemos alcanzado nuestro objetivo de 10 números.

```python
# El ciclo se repetirá MIENTRAS el contador sea menor o igual a 10.
while (contador <= 10):
    
    # Imprimimos el número par actual en la consola.
    print(par)
    
    # Incrementamos el contador. ¡Esta línea es vital para evitar un ciclo infinito!
    contador = contador + 1
    
    # Preparamos el siguiente número par sumándole 2 al actual.
    par = par + 2
```

**Solución 2: Usando la estructura `for`**

Veamos cómo el ciclo `for` nos permite hacer lo mismo de manera más compacta.

```python
# La función range(1, 11, 1) genera números del 1 al 10. 
# El 11 es el límite superior (no inclusivo) y el último 1 es el paso o incremento.
for contador in range(1, 11, 1):
    
    # En lugar de sumar 2 en cada paso, usamos la matemática: 
    # Si iteración=1 -> par=2. Si iteración=2 -> par=4.
    print(contador * 2)
```

## Ciclos con Límites Dinámicos

Escribe un algoritmo que resuelva el siguiente problema: mostrar por pantalla los primeros N números pares. N es un valor ingresado por el usuario. Resuelve este problema usando ciclo `while` y ciclo `for`.

### Análisis y Estrategia de Resolución:
Este ejercicio es una generalización del anterior. La única diferencia es que el límite del ciclo ya no es un número "duro" (hardcoded) como 10, sino una variable `N` provista por el usuario mediante `input()`.
Debemos capturar ese valor, transformarlo a entero (ya que `input` devuelve texto) y usarlo como nuestro nuevo límite. El enésimo número par siempre será `2 * N`, lo que nos sirve como límite matemático.

### Resolución Paso a Paso (Bloques de Código):

**Solución 1: Usando la estructura `while`**

Primero, solicitamos la información al usuario.

```python
# Capturamos la entrada del usuario y la convertimos explícitamente a un número entero (int)
N = int(input("¿Cuántos números pares quieres desplegar? "))

# Esta vez, usaremos 'contador' tanto para iterar como para imprimir el valor.
# Empezamos en 2, que es el primer par.
contador = 2
```

Ahora iteramos hasta llegar al límite matemático de los números pares requeridos.

```python
# Sabemos que el N-ésimo número par es N * 2. 
# Por ejemplo, si N=4, el último par a imprimir es 8.
while (contador <= 2 * N):
    print(contador)
    
    # Saltamos de 2 en 2 en cada iteración
    contador = contador + 2
```

**Solución 2: Usando la estructura `for`**

Nuevamente, el ciclo `for` junto con `range()` optimiza nuestro código.

```python
# Asumiendo que ya capturamos N en la celda anterior:
# range(inicio, fin, paso)
# Empezamos en 2. Terminamos en 2*N + 1 (el +1 es porque el límite final en range no es inclusivo).
# Avanzamos de a 2 pasos.
for contador in range(2, 2 * N + 1, 2):
    print(contador)
```

## Ciclos con Ramificaciones Condicionales (`if/else`)

Escribe un algoritmo que resuelva el siguiente problema: desplegar por pantalla todos los números enteros que existen entre los números A y B (ambos inclusive) que ingresa el usuario. Nota que si A es menor que B, se despliegan los números en orden creciente. Pero si A es mayor que B, los números se despliegan en orden decreciente.

### Análisis y Estrategia de Resolución:
Aquí introducimos la toma de decisiones (`if/else`) combinada con ciclos.
1.  **Lectura de datos:** Pedimos los valores A (N1) y B (N2).
2.  **Evaluación Lógica:** Debemos comparar A y B.
    * Si A <= B: Estamos en un escenario creciente. Usaremos un ciclo que sume +1 en cada paso.
    * Si A > B: Estamos en un escenario decreciente. Usaremos un ciclo que reste -1 en cada paso.
Usar un ciclo `for` es ideal aquí porque conocemos exactamente los límites de inicio y fin.

### Resolución Paso a Paso (Bloques de Código):

Primero, leemos las variables.

```python
# Solicitamos los límites al usuario
N1 = int(input("Ingrese el primer número (A): "))
N2 = int(input("Ingrese el segundo número (B): "))
```

A continuación, creamos la estructura condicional que bifurcará nuestro programa eligiendo el ciclo adecuado.

```python
# Caso 1: Orden creciente (o iguales)
if (N1 <= N2):
    # Rango desde N1 hasta N2. Usamos N2 + 1 para incluir N2. El paso es 1 (creciente).
    for contador in range(N1, N2 + 1, 1):
        print(contador)

# Caso 2: Orden decreciente
else:
    # Rango desde N1 bajando hasta N2. Usamos N2 - 1 para incluir N2 al ir bajando. 
    # El paso es -1, crucial para que el ciclo cuente hacia atrás.
    for contador in range(N1, N2 - 1, -1):
        print(contador)
```

# Título de la Materia: Control de Flujo con Variables de Estado (Banderas / Intentos)

Escribe un algoritmo en Python que emule la solicitud de clave en una cajero automático: si el usuario equivoca la clave 3 veces consecutivas, el cajero muestra el mensaje "Tarjeta retenida". Si el usuario logra ingresar la clave válida en uno de los 3 primeros intentos, entonces aparece el mensaje "Bienvenid@ al servicio". Supón que la clave válida se encuentra almacenada en una variable.

### Análisis y Estrategia de Resolución:
Este es un clásico problema de "máquina de estados" simple. 
1.  Tenemos un límite de seguridad: 3 intentos. Esto nos indica que usaremos un `while` que evalúe cuántos intentos llevamos.
2.  Tenemos una condición de "éxito prematuro": Si el usuario ingresa la clave correcta al primer o segundo intento, el ciclo debe romperse y no pedir la clave más veces.
3.  Al salir del ciclo, debemos saber *por qué* salimos: ¿porque se acabaron los intentos o porque la clave fue correcta? Evaluaremos esto con un `if` final.

### Resolución Paso a Paso (Bloques de Código):

Configuramos el estado inicial de nuestro cajero virtual.

```python
# Guardamos la clave correcta (hardcoded según el enunciado)
claveValida = 711

# Inicializamos nuestro contador de intentos en 0
numeroIntentos = 0
```

Implementamos la lógica de autenticación cíclica.

```python
# Permitimos la entrada mientras el usuario no haya consumido sus 3 intentos
while (numeroIntentos < 3):
    
    # Solicitamos la clave en cada intento
    claveIngresada = int(input("Ingresa tu clave: "))
    
    # Incrementamos inmediatamente el contador de intentos consumidos
    numeroIntentos = numeroIntentos + 1
    
    # Si la clave es correcta, aplicamos un "truco" para salir del ciclo:
    # Forzamos que la condición del while se vuelva falsa asignando el máximo de intentos.
    # (Nota: Otra forma más moderna en Python sería usar la palabra reservada 'break').
    if (claveIngresada == claveValida):
        numeroIntentos = 3 
```

Validación final tras finalizar el ciclo.

```python
# El ciclo terminó. Ahora verificamos si la última clave ingresada fue la correcta
if (claveIngresada == claveValida):
    print("Bienvenid@ al servicio")
else:
    # Si llegó aquí y la clave no es válida, significa que agotó los 3 intentos fallando.
    print("Tarjeta retenida")
```


## Ciclos con Variables Acumuladoras y Condición de Centinela

Escribe un algoritmo en Python que solicite números enteros al usuario y vaya calculando la suma de todos ellos. La solicitud de números al usuario termina cuando se ingresa un -1, en cuyo caso se muestra por pantalla el resultado de la suma. Está permitido que el usuario ingrese como primer valor un -1, en cuyo caso el algoritmo termina inmediatamente.

### Análisis y Estrategia de Resolución:
Aquí nos enfrentamos a un ciclo donde **no sabemos** cuántas veces iteraremos. El límite lo define el usuario ingresando un "valor centinela" (el `-1`).
Para esto, el `while` es obligatorio. Además, introducimos el concepto de **variable acumuladora**: una variable que empieza en 0 y a la cual le iremos sumando valores iteración tras iteración (`suma = suma + valor`).

El orden de lectura es clave:
1. Leer el primer número antes del ciclo.
2. Comprobar si no es el centinela.
3. Sumarlo al acumulador.
4. Volver a leer un número al *final* del bloque del ciclo.

### Resolución Paso a Paso (Bloques de Código):

Preparamos nuestro acumulador y la primera lectura (lectura de inicialización).

```python
# 'suma' es nuestro acumulador. Debe empezar en 0 (el neutro aditivo).
suma = 0

# Hacemos la primera lectura ANTES del ciclo para poder evaluar la condición.
numIngresado = int(input("Ingrese un número entero o -1 para terminar: "))
```

Procesamos los datos mientras no se detecte el centinela.

```python
# El ciclo iterará indefinidamente mientras el usuario NO ingrese un -1
while (numIngresado != -1):
    
    # Sumamos el número válido ingresado al acumulador total
    suma = suma + numIngresado
    
    # Volvemos a pedir un número. ¡Esto es vital! 
    # Si no volvemos a leer un dato, el ciclo será infinito evaluando el mismo número siempre.
    numIngresado = int(input("Ingrese un número entero o -1 para terminar: "))
```

Por último, mostramos los resultados del proceso.

```python
# Esta línea solo se ejecutará cuando el ciclo while termine (es decir, cuando numIngresado sea -1)
print("La suma total de números es: ", suma)
```
