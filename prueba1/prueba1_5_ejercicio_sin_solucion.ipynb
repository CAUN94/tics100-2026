**Enunciado del Ejercicio:**

En un laboratorio de física, se está realizando un experimento para medir la capacidad térmica del agua. Para ello, se calienta una muestra de agua y se registra la temperatura en diferentes momentos.

Escriba un programa que solicite al usuario la cantidad de mediciones a registrar y luego reciba cada una de las temperaturas. Finalmente, el programa debe mostrar el promedio de las temperaturas, la temperatura máxima y la temperatura mínima registrada durante el experimento.

### Paso 1: Entender el calculo del promedio matematicamente
Antes de procesar muchas temperaturas usando ciclos, veamos como funciona la logica base si tuvieramos solo 3 temperaturas fijas.

Para calcular el promedio, simplemente sumamos todas las temperaturas obtenidas y dividimos ese gran total por la cantidad de mediciones que hicimos.

### Paso 2: El concepto de encontrar el mayor y menor en una lista
Para que una computadora encuentre el numero mayor en un grupo de opciones, no puede verlos todos a la vez como nosotros. Debe ir evaluando uno por uno.

Si tenemos las temperaturas guardadas en una lista de 3 opciones, el truco logico esta en **asumir que el primer valor (la posicion 0) es el mas grande y el mas pequeño a la vez** y guardarlo en una variable.

Luego, lo comparamos con el siguiente numero usando un condicional `if`. Si el numero nuevo es mas grande, actualizamos nuestra variable. Si no lo es, lo ignoramos.

### Paso 3: Evitar el trabajo manual usando for y range()
Si el experimento tuviera 100 mediciones, no podriamos escribir 100 bloques `if` a mano. Para solucionar esto, usamos un ciclo `for` con `range()`.

Esta herramienta nos permite repetir una instruccion de pedir datos la cantidad de veces que queramos.

### Paso 4: El uso de 'i' para detectar la primera iteracion
Al automatizar la recoleccion de datos, surge un problema logico: ¿Como asignamos el primer maximo y minimo si el usuario ingresa los datos uno a uno y ya no tenemos una lista con la posicion 0?

La solucion esta en la variable `i` de nuestro ciclo `for`. Recuerda que `i` siempre comienza valiendo `0` en la primera vuelta.

Por lo tanto, utilizamos la condicion `if i == 0:` para decirle a la computadora: *"Esta es la primera temperatura que ingresa el usuario. Como es la unica que tenemos, guardala automaticamente como la maxima y minima inicial sin comparar con nada"*.

En todas las vueltas siguientes (`else`), la computadora sabra que ya tiene un valor guardado y recien ahi aplicara las comparaciones que vimos en el Paso 2.

### Paso 5: Solucion final uniendo conceptos
Ahora armaremos el sistema definitivo. 

Usaremos el ciclo `for` para pedir las temperaturas una por una. Dentro de ese mismo ciclo, iremos sumando cada temperatura a un acumulador (para calcular el promedio final) y usaremos la logica del `if i == 0:` para manejar el maximo y minimo paso a paso.
