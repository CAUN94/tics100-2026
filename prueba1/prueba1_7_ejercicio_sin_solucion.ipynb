**Enunciado del Ejercicio:**

Hoy en día, la gran mayoría de los dispositivos electrónicos usa baterías de litio. Estudios han mostrado que mantener el nivel de carga de este tipo de baterías en un rango entre el 20% y el 80% alarga su vida útil. Su empresa que crea dispositivos electrónicos tiene un listado en papel con la información de cada cliente.

Específicamente, cada papel corresponde a un dispositivo y cada línea corresponde a la carga de la batería por cada minuto que el dispositivo estuvo encendido durante la semana.

Cree un programa donde una persona ingrese a mano cada uno de los valores mostrados y reporte el porcentaje de tiempo que un dispositivo ha pasado (1) en rango óptimo de carga, (2) sobre el rango óptimo y (3) por debajo del rango óptimo en la última semana.

Para el ingreso de datos usted deberá considerar dos cosas:
1. Los valores ingresados corresponden a porcentajes, por lo cual la persona ingresará números enteros entre 0 y 100. Caso contrario el ingreso no es válido, no debe ser considerado, y deberá mostrar un mensaje por pantalla indicando el error.
2. El ingreso de datos finaliza cuando el valor ingresado sea igual a 0.

### Introduccion a la logica del ejercicio
Antes de programar, debemos notar una diferencia fundamental con ejercicios anteriores: **no sabemos cuántos datos ingresará el usuario**. 

El ingreso podría terminar al tercer número o al número mil, dependiendo de cuándo la persona escriba un `0`. Por esta razón, más adelante estaremos obligados a usar un ciclo `while` (mientras), ya que nos permite repetir acciones para una cantidad **indefinida** de valores. 

Pero primero, construyamos las piezas internas paso a paso.

### Paso 1: Clasificar el dato y contarlo
Necesitamos contadores en cero para saber cuántos minutos pasó la batería en cada estado. Crearemos tres contadores.

Luego, utilizaremos un bloque condicional `if / elif / else` para evaluar una carga estática y sumarle 1 al contador que corresponda.

### Paso 2: Como calcular un porcentaje relativo
El enunciado nos pide reportar el "porcentaje de tiempo". Para calcular qué porcentaje representa un contador respecto al total de datos ingresados, usamos la fórmula matemática clásica: `(Parte / Total) * 100`.

Para esto, necesitamos un cuarto contador que lleve el registro del total de minutos válidos que se han ingresado.

### Paso 3: Validaciones controladas con if y else
El sistema tiene una regla estricta: no acepta números menores a 0 ni mayores a 100. 

Para manejar esto limpiamente, usamos un bloque `if` que detecte el error y muestre el mensaje. Todo lo que sea un dato correcto (es decir, el cálculo de contadores) lo pondremos dentro del bloque `else`. Así, si hay error, se ignora el conteo.

### Paso 4: El ciclo while controlado por una desigualdad
Como explicamos al inicio, no sabemos cuándo se ingresará el `0` que termina el programa. 

La forma más ordenada y segura de hacer esto es pedir el dato una vez ANTES de iniciar el ciclo. Luego, la condición del `while` será una simple desigualdad: `while numero != 0:`. Al final del bloque del ciclo, volvemos a pedir el dato para la siguiente vuelta.

### Paso 5: Solucion final uniendo conceptos
Ahora armaremos el sistema completo.

1. Crearemos los cuatro contadores.
2. Pediremos la carga por primera vez.
3. Usaremos el ciclo `while carga != 0:` para procesar datos indefinidamente.
4. Dentro del ciclo, usaremos el esquema `if/else` del Paso 3 para separar los errores de los datos válidos.
5. Al final del bloque del ciclo, volveremos a pedir la `carga`.
6. Al terminar el `while`, calcularemos los porcentajes y mostraremos el reporte.

### Desafio Extra: La memoria del pasado
Existe una restricción lógica adicional que no hemos aplicado: *"La batería no puede subir de porcentaje, por lo cual si el valor ingresado es exclusivamente mayor que el último valor ingresado, no debe ser considerado, y deberá mostrar un mensaje de error".*

Para resolver esto **sin** usar comandos para saltar código de forma abrupta, simplemente agregamos un `elif` a nuestra cadena inicial de validaciones de error. 

Solo si el número sobrevive a los `if / elif` de errores, caerá en el `else` donde lo procesaremos como un dato válido. Además, debemos recordar actualizar la memoria (`carga_anterior = carga`) dentro de ese mismo bloque `else`.
