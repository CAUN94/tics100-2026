### Ejercicio 1: Clasificación de Números Pares e Impares

**Enunciado:** Crea un programa donde el usuario entregue números de forma indeterminada. Si el número es par, se guarda en una lista de pares, y si es impar, en una de impares. El programa se detiene si el usuario ingresa un `0` o un número menor. Al finalizar, muestra ambas listas por pantalla.

El número final no se incluye en las listas, ya que es el que indica el fin del ingreso de datos



```python
pares = []
impares = []
numero = 1  # Inicializamos con un valor positivo para entrar en el bucle

while numero > 0:
    numero = int(input("Ingrese un número (0 o negativo para terminar): "))
        
    # Clasificación
    if numero % 2 == 0 and numero > 0:
        pares.append(numero)
    elif numero % 2 != 0 and numero > 0:
        impares.append(numero)

print("Lista de números pares:", pares)
print("Lista de números impares:", impares)
```

### Ejercicio 2: Buscador de Divisores Comunes

**Enunciado:** Pide al usuario 3 números. El primer número será la cantidad de números a revisar (el límite, por ejemplo, los números del 1 al 20). Los otros dos números serán los divisores. Debes encontrar y mostrar todos los números dentro del rango que sean divisibles EXACTAMENTE por los dos divisores entregados.

**Hint:** Usa `for i in range(1, limite + 1)` y luego evalúa con el operador módulo `%`.

```python
limite = int(input("Ingrese la cantidad de números a revisar (ej: 20): "))
div1 = int(input("Ingrese el primer divisor: "))
div2 = int(input("Ingrese el segundo divisor: "))

divisibles_comunes = []

for i in range(1, limite + 1):
    # Verificamos si es divisible por AMBOS números (el resto debe ser 0)
    if i % div1 == 0 and i % div2 == 0:
        divisibles_comunes.append(i)

print(f"\nNúmeros entre 1 y {limite} que son divisibles por {div1} y {div2}:")
print(divisibles_comunes)
```

### Ejercicio 3: Actualización Masiva de una Lista

**Enunciado:** Tienes una lista predefinida con 8 valores. Pídele al usuario que ingrese un número multiplicador. Utilizando `for i in range(len(lista))`, pasa por todos los números de la lista original y actualiza su valor al resultado de multiplicarlo por el número del usuario. Imprime la lista actualizada.

**Hint:** Modifica la lista directamente usando su índice `lista[i] = lista[i] * multiplicador`.

```python
# Lista predefinida con 8 valores
valores = [2, 5, 8, 10, 12, 15, 20, 25]

print("Lista original:", valores)

multiplicador = int(input("\nIngrese un número para multiplicar la lista: "))

# Actualización modificando la posición exacta de memoria
for i in range(len(valores)):
    valores[i] = valores[i] * multiplicador
    
print("\nLista actualizada:", valores)
```

### Ejercicio 4: Reporte de Inventario y Ventas (Listas Paralelas)

**Enunciado:** Tienes 3 listas asociadas: nombres de productos, cantidades disponibles y el precio por unidad. Usando un ciclo `for`, muestra por pantalla el nombre del producto, la cantidad y el valor total (precio * cantidad). 
Si un producto no tiene stock (cantidad es igual a 0), ocúltalo del reporte de ventas y guarda su nombre en una lista de productos sin stock. Muéstralos al final.

**Hint:** Usa `range(len(nombres))` como un índice maestro para acceder a la misma posición en las 3 listas al mismo tiempo.

```python
nombres = ["Teclado", "Mouse", "Monitor", "Audífonos", "Webcam"]
cantidades = [15, 0, 5, 0, 10]
precios = [25000, 15000, 120000, 30000, 45000]

sin_stock = []

print("--- REPORTE DE PRODUCTOS DISPONIBLES ---")

for i in range(len(nombres)):
    # Evaluamos si hay stock en la posición actual
    if cantidades[i] > 0:
        valor_total = cantidades[i] * precios[i]
        print(f"Producto: {nombres[i]} | Stock: {cantidades[i]} | Valor Total: ${valor_total}")
    else:
        # Si no hay stock, guardamos el nombre para el reporte final
        sin_stock.append(nombres[i])

print("\n--- ALERTA: PRODUCTOS SIN EXISTENCIA ---")
for producto in sin_stock:
    print("-", producto)
```
