## Desafío 1: El Motor del Sudoku

A continuación, tienes un bloque de código que incluye un tablero de Sudoku incompleto y dos funciones ya programadas:

1. `mostrar_sudoku(tablero)`: Imprime el tablero de forma amigable.
2. `es_jugada_valida(tablero, fila, columna, numero)`: Recibe el tablero, la fila, la columna y el número que intentas colocar. Retorna `True` si la jugada es válida (no se repite en la fila, columna ni en el bloque de 3x3) y `False` si la jugada es ilegal.

**Tu Misión:**
En la celda que dice "TU CÓDIGO AQUÍ", debes programar una función llamada `jugar_sudoku(tablero)`. Esta función debe tener un ciclo `while` interactivo que le pregunte al usuario la fila, columna y el número a jugar.

Debes usar la función `es_jugada_valida` para evaluar el movimiento. Si es válido, colócalo en el tablero y muéstralo. Si no, imprime un mensaje de error. (Asume que el juego no termina por ahora, solo crea el ciclo para ir colocando números).

```python
# ==========================================
# FUNCIONES PRE-PROGRAMADAS (¡No tocar!)
# ==========================================
def mostrar_sudoku(tablero):
    print("\n--- TABLERO SUDOKU ---")
    for i in range(9):
        if i % 3 == 0 and i != 0:
            print("-" * 21)
        for j in range(9):
            if j % 3 == 0 and j != 0:
                print("|", end=" ")
            valor = tablero[i][j]
            print(valor if valor != 0 else ".", end=" ")
        print()
    print("----------------------\n")

def es_jugada_valida(tablero, fila, col, num):
    # Revisar fila y columna
    for i in range(9):
        if tablero[fila][i] == num or tablero[i][col] == num:
            return False
    # Revisar bloque 3x3
    inicio_fila = (fila // 3) * 3
    inicio_col = (col // 3) * 3
    for i in range(3):
        for j in range(3):
            if tablero[inicio_fila + i][inicio_col + j] == num:
                return False
    return True

tablero_base = [
    [5, 3, 0, 0, 7, 0, 0, 0, 0],
    [6, 0, 0, 1, 9, 5, 0, 0, 0],
    [0, 9, 8, 0, 0, 0, 0, 6, 0],
    [8, 0, 0, 0, 6, 0, 0, 0, 3],
    [4, 0, 0, 8, 0, 3, 0, 0, 1],
    [7, 0, 0, 0, 2, 0, 0, 0, 6],
    [0, 6, 0, 0, 0, 0, 2, 8, 0],
    [0, 0, 0, 4, 1, 9, 0, 0, 5],
    [0, 0, 0, 0, 8, 0, 0, 7, 9]
]

# ==========================================
# TU CÓDIGO AQUÍ
# ==========================================

def jugar_sudoku(tablero):
    mostrar_sudoku(tablero)
    while True:
        i = int(input("Ingresa la fila (0-8): "))
        j = int(input("Ingresa la columna (0-8): "))
        num = int(input("Ingresa el número (1-9): "))
        if es_jugada_valida(tablero, i, j, num):
            tablero[i][j] = num
            mostrar_sudoku(tablero)
        else:
            print("¡Jugada no válida! Intenta de nuevo.")
    

# Descomenta la siguiente línea para probar tu código:
jugar_sudoku(tablero_base)
```

## Desafío 2: Camuflando la Sopa de Letras

En este ejercicio cuentas con dos funciones prefabricadas:

1. `crear_tablero_oculto(palabra, n)`: Le entregas una palabra y un valor `n`. Automáticamente crea una matriz de `(largo_de_la_palabra + n)` de tamaño, e inserta la palabra oculta horizontalmente en una posición al azar. El resto del tablero se llena con espacios vacíos `"-"`.
2. `mostrar_sopa(tablero)`: Imprime la matriz por pantalla.


Debes crear una función llamada `rellenar_sopa(tablero, palabra)`. Esta función debe recibir la matriz que tiene los espacios `"-"` y la misma palabra que se ocultó.
La función debe recorrer la matriz buscando todos los `"-"`. Al encontrarlos, debe reemplazarlos por una **letra al azar escogida de la misma palabra original**, para lograr el camuflaje perfecto.

```python
import random

# ==========================================
# FUNCIONES PRE-PROGRAMADAS (¡No tocar!)
# ==========================================
def crear_tablero_oculto(palabra, n):
    tamano = len(palabra) + n
    tablero = []
    # Crear matriz vacía con guiones
    for i in range(tamano):
        fila = []
        for j in range(tamano):
            fila.append("-")
        tablero.append(fila)
        
    # Insertar la palabra al azar
    fila_azar = random.randint(0, tamano - 1)
    col_azar = random.randint(0, n) # Hasta n para que la palabra no se salga
    
    for k in range(len(palabra)):
        tablero[fila_azar][col_azar + k] = palabra[k]
        
    return tablero

def mostrar_sopa(tablero):
    print("\n--- SOPA DE LETRAS ---")
    for i in range(len(tablero)):
        for j in range(len(tablero[i])):
            print(tablero[i][j], end=" ")
        print()
    print("----------------------\n")

# ==========================================
# TU CÓDIGO AQUÍ
# ==========================================

def rellenar_sopa(tablero, palabra):
    
    for i in range(len(tablero)):
        for j in range(len(tablero[i])):
            pos_rand = random.randint(0, len(palabra) - 1)
            letra = palabra[pos_rand]
            if tablero[i][j] == "-":
                tablero[i][j] = letra

    
# PRUEBAS DEL SISTEMA
palabra_secreta = "PYTHON"
dificultad = 4

# 1. Usamos la caja negra para crear el tablero base
mi_sopa = crear_tablero_oculto(palabra_secreta, dificultad)
mostrar_sopa(mi_sopa)

# 2. Aquí llamaremos a tu función para que la rellene
rellenar_sopa(mi_sopa, palabra_secreta)
mostrar_sopa(mi_sopa)
```
