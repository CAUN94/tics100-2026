## El ciclo `for` leyendo una lista directamente

El ciclo `for` es excelente para trabajar con colecciones de datos. Nos permite recorrer los elementos de una lista uno por uno de manera directa. En cada iteración (vuelta) del ciclo, una variable temporal que nosotros nombramos toma el valor del elemento actual para que podamos evaluarlo o usarlo.

### Ejemplo: Calculando un promedio selectivo
Supongamos que queremos calcular el promedio de un curso, pero **solamente** sumando las notas que están aprobadas (iguales o mayores a 4.0). Mezclaremos el ciclo `for` con un condicional `if` para filtrar las notas.

**Ejemplo de ejecución esperada:**
```text
Procesando nota: 3.2
Procesando nota: 5.5 -> ¡Aprobada! Sumando...
Procesando nota: 6.0 -> ¡Aprobada! Sumando...
Procesando nota: 2.8
Procesando nota: 4.5 -> ¡Aprobada! Sumando...

El promedio de las 3 notas aprobadas es: 5.333333333333333
```

````python
notas_curso = [3.2, 5.5, 6.0, 2.8, 4.5]
suma_aprobadas = 0
cantidad_aprobadas = 0

# Recorremos la lista directamente
for nota in notas_curso:
    if nota >= 4.0:
        print("Procesando nota:", nota, "-> ¡Aprobada! Sumando...")
        suma_aprobadas = suma_aprobadas + nota
        cantidad_aprobadas = cantidad_aprobadas + 1
    else:
        print("Procesando nota:", nota)

# Calculamos el promedio al salir del ciclo
promedio = suma_aprobadas / cantidad_aprobadas

print()
print("El promedio de las", cantidad_aprobadas, "notas aprobadas es:", promedio)
````

## El ciclo `for` usando `range()`

A veces no tenemos una lista guardada en la memoria, sino que necesitamos que una acción se repita un número específico de veces o necesitamos generar una secuencia numérica matemática. Para eso combinamos el `for` con la función `range(inicio, fin, paso)`.

### Ejemplo: Generador de Tablas de Multiplicar
Crearemos un programa que solicite al usuario un número por teclado y luego genere automáticamente la tabla de multiplicar de ese número (del 1 al 10), mostrando la operación matemática en pantalla paso a paso.

**Ejemplo de ejecución esperada:**
```text
Ingrese el número para generar su tabla: 7
7 x 1 = 7
7 x 2 = 14
...
7 x 10 = 70
```

````python
numero_elegido = int(input("Ingrese el número para generar su tabla: "))

# Usamos range del 1 al 11 (porque el número final es excluyente)
for multiplicador in range(1, 11):
    resultado = numero_elegido * multiplicador
    print(numero_elegido, "x", multiplicador, "=", resultado)
````

## El ciclo `for` usando `range(len(lista))`

Recorrer una lista directamente (como en el primer bloque) es útil para *leer* información. Sin embargo, si necesitamos **modificar** la lista original, necesitamos saber exactamente en qué "casilla" o índice (0, 1, 2...) estamos parados. 

Para lograr esto combinamos tres cosas: `for`, la función `range()` y la función `len()`.

### Ejemplo: Actualizando precios de una tienda
Tenemos una lista de precios en una tienda. Usaremos el índice `i` para evaluar la casilla de cada precio. Si el precio original es mayor a 3000, le aplicaremos un descuento de 1000 pesos y lo reemplazaremos en la lista original.

**Ejemplo de ejecución esperada:**
```text
Revisando el índice 0 con precio 1000
Revisando el índice 1 con precio 2500
Revisando el índice 2 con precio 5000 -> Aplicando descuento...
Revisando el índice 3 con precio 10000 -> Aplicando descuento...
Lista original actualizada: [1000, 2500, 4000, 9000]
```

````python
precios = [1000, 2500, 5000, 10000]

# len(precios) es 4, por ende range(4) generará los números 0, 1, 2, 3.
# Esos números coinciden exactamente con los índices de la lista.
for i in range(len(precios)):
    if precios[i] > 3000:
        print("Revisando el índice", i, "con precio", precios[i], "-> Aplicando descuento...")
        precios[i] = precios[i] - 1000
    else:
        print("Revisando el índice", i, "con precio", precios[i])

print("Lista original actualizada:", precios)
````

## El ciclo `while` (La importancia de la condición y la actualización)

El ciclo `while` repite un bloque de código **mientras** una condición sea evaluada como verdadera. Todo ciclo `while` bien programado y seguro necesita estrictamente dos cosas:

1. **Una condición inicial** definida antes del ciclo (para que pueda arrancar).
2. **Una actualización matemática o lógica** dentro del bloque de código. Si la variable nunca cambia, la condición nunca se volverá falsa y caeremos en un temido *Bucle Infinito* (un programa que se cuelga para siempre).

En este curso no utilizaremos atajos para forzar la salida (como la palabra reservada `break`). Todo se controlará a través de la lógica de la condición.

### Ejemplo: Simulador de Login con Límite de Intentos
Haremos un sistema que pida una contraseña, pero solo nos dará 3 oportunidades.

**Ejemplo de ejecución esperada:**
```text
Ingrese contraseña: hola
Contraseña incorrecta. Intentos gastados: 1
Ingrese contraseña: secreto
Acceso concedido al sistema.
```

````python
# 1. Condición inicial antes de que arranque el ciclo
intentos_gastados = 0

while intentos_gastados < 3:
    clave = input("Ingrese contraseña: ")
    
    if clave == "secreto":
        print("Acceso concedido al sistema.")
        # 2. ACTUALIZACIÓN: Alteramos la variable matemática para forzar 
        # que la condición (intentos_gastados < 3) se vuelva Falsa y el ciclo termine.
        intentos_gastados = 3 
    else:
        # 2. ACTUALIZACIÓN: Sumamos 1 al contador. Eventualmente llegará a 3 y el ciclo terminará.
        intentos_gastados += 1
        print("Contraseña incorrecta. Intentos gastados:", intentos_gastados)
````

## Aplicaciones: Contadores

Un contador es un patrón de programación muy común y útil. Es simplemente una variable numérica que **inicializamos en 0** antes de un ciclo. Cada vez que el programa evalúa una condición dentro de las repeticiones y resulta verdadera, le sumamos 1 al contador (`contador = contador + 1` o `contador += 1`).

### Ejemplo: Contando letras en un texto
Le pediremos al usuario que ingrese una frase secreta. Utilizando un `for` (recordemos que los textos o *strings* se comportan exactamente como listas de letras), contaremos cuántas vocales "a" (mayúsculas o minúsculas) existen.

**Ejemplo de ejecución esperada:**
```text
Ingrese una frase: La programacion es apasionante
La letra 'a' o 'A' aparece 4 veces en el texto.
```

````python
frase = input("Ingrese una frase: ")

contador_a = 0  # Condición inicial del contador

# La variable 'letra' tomará el valor de cada caracter de la frase, uno por uno.
for letra in frase:
    if letra == "a" or letra == "A":
        contador_a += 1  # Actualización sumando de 1 en 1

print("La letra 'a' o 'A' aparece", contador_a, "veces en el texto.")
````

## Aplicaciones: Acumuladores / Sumadores

A diferencia de un contador (que suma rígidamente de 1 en 1), un acumulador guarda **totales dinámicos**. Inicia en `0` (para sumas o restas), y en cada iteración se le suma un valor que puede ser distinto en cada vuelta (`acumulador = acumulador + nuevo_valor`).

### Ejemplo: Suma de los Cuadrados Perfectos
Pediremos un número `N` al usuario y usaremos un acumulador para sumar $1^2 + 2^2 + 3^2 ...$ hasta llegar a $N^2$.

**Ejemplo de ejecución esperada:**
```text
Ingrese el número límite: 4
Sumando el cuadrado de 1 (es decir 1). Acumulado: 1
Sumando el cuadrado de 2 (es decir 4). Acumulado: 5
Sumando el cuadrado de 3 (es decir 9). Acumulado: 14
Sumando el cuadrado de 4 (es decir 16). Acumulado: 30
El total acumulado de los cuadrados es: 30
```

````python
limite = int(input("Ingrese el número límite: "))
acumulador_cuadrados = 0

# Usamos limite + 1 para que el range llegue efectivamente hasta el número ingresado
for i in range(1, limite + 1):
    cuadrado = i ** 2
    acumulador_cuadrados = acumulador_cuadrados + cuadrado
    print("Sumando el cuadrado de", i, "(es decir", cuadrado, "). Acumulado:", acumulador_cuadrados)

print("El total acumulado de los cuadrados es:", acumulador_cuadrados)
````

## Aplicaciones: Variables Bandera (Flags)

Una "bandera" (o flag en inglés) es una variable lógica que asume un estado (generalmente `False` o `True`). Se utiliza para dejar un registro en la memoria de que un evento especial ocurrió dentro de un ciclo masivo.

Es perfecta para usarse como parte de la condición de un `while` y detener una búsqueda o proceso cuando encontramos lo que queríamos, sin tener que recurrir a comandos bruscos que rompan el código.

### Ejemplo: Búsqueda en Bitácora de Seguridad
Evaluaremos una lista de palabras buscando el término "intruso". Usaremos un `while` con dos condiciones entrelazadas: que el índice no supere el tamaño de la lista **Y** que nuestra bandera `alerta` siga siendo `False`.

**Ejemplo de ejecución esperada:**
```text
Revisando registro...
Revisando registro...
¡ALERTA! Intruso detectado en la posición 2. Deteniendo búsqueda.
```

````python
bitacora = ["usuario1", "usuario2", "intruso", "usuario3", "usuario4"]

# 1. Condiciones iniciales
indice = 0
alerta_activada = False

# 2. El while exige que ambas cosas se cumplan para seguir dando vueltas
while indice < len(bitacora) and alerta_activada == False:
    
    if bitacora[indice] == "intruso":
        # Si lo encontramos, "levantamos la bandera".
        # Esto hará que el while se detenga en la siguiente evaluación.
        alerta_activada = True
        print("¡ALERTA! Intruso detectado en la posición", indice, ". Deteniendo búsqueda.")
    else:
        print("Revisando registro...")
        # Si todo está normal, simplemente avanzamos de casilla sumando 1 al índice
        indice += 1

# Una vez fuera del ciclo, podemos tomar decisiones basadas en cómo quedó la bandera
if alerta_activada == False:
    print("Revisión completada sin problemas. Sistema seguro.")
````
