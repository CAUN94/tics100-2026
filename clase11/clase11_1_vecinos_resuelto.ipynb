# Guía: Dos Formas de Revisar Vecinos en una Matriz

Cuando necesitamos saber qué hay alrededor de una coordenada en una matriz (sus 8 vecinos), existen dos formas de programarlo:
1. **El Método Manual (Fuerza Bruta):** Escribiendo una combinación de múltiples `if` para cada dirección.
2. **El Método Matemático (Elegante):** Usando un doble `for` con desplazamientos.

En esta guía compararemos ambos métodos para que entiendas la lógica detrás de cada uno y descubras por qué uno es preferido sobre el otro en el desarrollo profesional.

---
## El Tablero de Pruebas

Primero, definiremos una matriz de 4x4 y estableceremos nuestro "Punto Central". Nos pararemos en la Fila 1, Columna 1 (donde está el número `21`).

```python
matriz = [
    [10, 11, 12, 13],
    [20, 21, 22, 23],  # Nuestro objetivo es el 21
    [30, 31, 32, 33],
    [40, 41, 42, 43]
]

mi_fila = 1
mi_col = 1
limite_filas = len(matriz)
limite_cols = len(matriz[0])

print(f"Estamos parados en el número: {matriz[mi_fila][mi_col]}")
```

---
## Método 1: La Fuerza Bruta (Combinación de `if`)

En este método, evaluaremos cada uno de los 8 vecinos de forma manual.

Para cada dirección, debemos calcular su coordenada y escribir un `if` que verifique que esa coordenada específica no se salga de los límites del tablero (para evitar un `IndexError`). 

Observa la cantidad de código necesario:

```python
matriz = [
    [10, 11, 12, 13],
    [20, 21, 22, 23],  # Nuestro objetivo es el 21
    [30, 31, 32, 33],
    [40, 41, 42, 43]
]

mi_fila = 1
mi_col = 1
limite_filas = len(matriz)
limite_cols = len(matriz[0])

print(f"Estamos parados en el número: {matriz[mi_fila][mi_col]}")


print("--- REVISIÓN MANUAL CON IF ---")

# 1. Arriba (Fila - 1, Misma Columna)
if mi_fila - 1 >= 0:
    print(f"Arriba: {matriz[mi_fila - 1][mi_col]}")

# 2. Abajo (Fila + 1, Misma Columna)
if mi_fila + 1 < limite_filas:
    print(f"Abajo: {matriz[mi_fila + 1][mi_col]}")

# 3. Izquierda (Misma Fila, Columna - 1)
if mi_col - 1 >= 0:
    print(f"Izquierda: {matriz[mi_fila][mi_col - 1]}")

# 4. Derecha (Misma Fila, Columna + 1)
if mi_col + 1 < limite_cols:
    print(f"Derecha: {matriz[mi_fila][mi_col + 1]}")

# 5. Diagonal Arriba-Izquierda (Fila - 1, Columna - 1)
if mi_fila - 1 >= 0 and mi_col - 1 >= 0:
    print(f"Diagonal Sup-Izq: {matriz[mi_fila - 1][mi_col - 1]}")

# 6. Diagonal Arriba-Derecha (Fila - 1, Columna + 1)
if mi_fila - 1 >= 0 and mi_col + 1 < limite_cols:
    print(f"Diagonal Sup-Der: {matriz[mi_fila - 1][mi_col + 1]}")

# 7. Diagonal Abajo-Izquierda (Fila + 1, Columna - 1)
if mi_fila + 1 < limite_filas and mi_col - 1 >= 0:
    print(f"Diagonal Inf-Izq: {matriz[mi_fila + 1][mi_col - 1]}")

# 8. Diagonal Abajo-Derecha (Fila + 1, Columna + 1)
if mi_fila + 1 < limite_filas and mi_col + 1 < limite_cols:
    print(f"Diagonal Inf-Der: {matriz[mi_fila + 1][mi_col + 1]}")
```

### Análisis del Método 1
Aunque funciona perfectamente, escribirlo tiene graves problemas:
- **Es tedioso:** Toma muchas líneas de código.
- **Propenso a errores (Typos):** Al copiar, pegar y modificar tantos signos `+`, `-`, `<`, `>=`, es extremadamente fácil equivocarse en un número o signo y arruinar el programa sin darte cuenta.
- **Difícil de modificar:** Si las reglas cambian, debes actualizar 8 bloques de código distintos.

---
## Método 2: La Solución Matemática (Doble `for`)

En lugar de escribir las 8 direcciones a mano, podemos usar la combinatoria matemática. 
Si te fijas en el código anterior, lo único que hicimos fue sumar o restar `1` o `0` a nuestras coordenadas.

Podemos automatizar esto creando un doble ciclo `for` que itere sobre la lista `[-1, 0, 1]`. El computador generará las 9 combinaciones posibles automáticamente. Solo debemos decirle que ignore el `0, 0` (que somos nosotros mismos) y aplicamos un único `if` global para proteger los límites.

```python
matriz = [
    [10, 11, 12, 13],
    [20, 21, 22, 23],  # Nuestro objetivo es el 21
    [30, 31, 32, 33],
    [40, 41, 42, 43]
]

mi_fila = 1
mi_col = 1
limite_filas = len(matriz)
limite_cols = len(matriz[0])

print(f"Estamos parados en el número: {matriz[mi_fila][mi_col]}")


print("--- REVISIÓN MATEMÁTICA CON FOR ---")

```

---
## Escalando la Solución: Revisando Toda la Matriz

Ya sabemos cómo usar la combinatoria para revisar los vecinos de una sola coordenada. Pero, ¿qué pasa si necesitamos analizar el tablero completo? En lugar de cambiar nuestras coordenadas de inicio manualmente una y otra vez, podemos automatizar el viaje.

Para lograrlo, simplemente envolvemos nuestra estrategia matemática (el "radar" de vecinos) dentro de **otro doble ciclo `for` principal**. Estos dos nuevos ciclos exteriores se encargarán de recorrer la matriz entera, fila por fila y columna por columna, deteniéndose en cada número.

Al hacer esto, combinamos ambas herramientas y terminamos usando **cuatro ciclos `for` anidados**: los dos primeros actúan como un viajero que avanza por cada casilla del tablero, y los dos ciclos interiores (la lista `[-1, 0, 1]`) se encienden en cada parada para revisar los alrededores inmediatos de esa coordenada específica. ¡Así conseguimos analizar todos los vecinos de todos los valores de la matriz de forma automática!

```python
matriz = [
    [10, 11, 12, 13],
    [20, 21, 22, 23],
    [30, 31, 32, 33],
    [40, 41, 42, 43]
]

limite_filas = len(matriz)
limite_cols = len(matriz[0])

print("--- REVISIÓN DE VECINOS PARA TODA LA MATRIZ ---\n")

```
