## P1) Análisis de Resultados (Listas 1D)

**Enunciado:**
Tienes tres listas paralelas que registran la campaña del UAI FC. La primera lista contiene los nombres de los equipos rivales, la segunda los goles que anotó el UAI FC, y la tercera los goles que encajó (recibió). 

Debes usar un ciclo `for` y acceder mediante el índice (`i`) a las tres listas simultáneamente para determinar el resultado de cada partido. Al final, debes mostrar por pantalla qué partidos se ganaron, empataron o perdieron, y calcular el total de puntos obtenidos sabiendo que: Victoria = 3 pts, Empate = 1 pt, Derrota = 0 pts.

**Ejemplo de ejecución:**
```text
UAI Vs Equipo A: Victoria
UAI Vs Equipo B: Derrota
UAI Vs Equipo C: Empate
UAI Vs Equipo D: Victoria

Puntaje total: 7
```

````python
lista_nombres = ["Equipo A", "Equipo B", "Equipo C", "Equipo D"]
goles_anotados = [2, 0, 1, 3]
goles_encajados = [1, 2, 1, 0]

puntos_totales = 0

for i in range(len(lista_nombres)):
    if goles_anotados[i] > goles_encajados[i]:
        resultado = "Victoria"
        puntos_totales += 3
    elif goles_anotados[i] == goles_encajados[i]:
        resultado = "Empate"
        puntos_totales += 1
    else:
        resultado = "Derrota"
        
    print(f"UAI Vs {lista_nombres[i]}: {resultado}")

print(f"\nPuntaje total: {puntos_totales}")
````

## P2) La Cruz del Mapa (Impresión de Símbolos en Matriz)

**Enunciado:**
Haga un código donde se le pida al usuario que ingrese 4 valores: una dimensión numérica `n`, un número de fila, un número de columna y un signo especial (por ejemplo, `+`).

El código debe crear una matriz bidimensional de puros gatos (`#`) de tamaño `n x n`. Sin embargo, durante la creación de la matriz, todas las casillas que pertenezcan a la **fila elegida** o a la **columna elegida** deben reemplazarse por el signo que entregó el usuario. Al finalizar, muestra la matriz por pantalla.

**Ejemplo de ejecución:**
```text
Ingrese dimensión del tablero: 5
Ingrese la fila: 2
Ingrese la columna: 2
Ingrese el signo: +

# # + # # 
# # + # # 
+ + + + + 
# # + # # 
# # + # # 
```

````python
n = int(input("Ingrese dimensión del tablero: "))
fila_elegida = int(input("Ingrese la fila: "))
columna_elegida = int(input("Ingrese la columna: "))
signo = input("Ingrese el signo: ")

tablero = []

# Construcción de la matriz
for i in range(n):
    fila_temporal = []
    for j in range(n):
        # Si estamos en la fila elegida o en la columna elegida, ponemos el signo
        if i == fila_elegida or j == columna_elegida:
            fila_temporal.append(signo)
        else:
            fila_temporal.append("#")
    tablero.append(fila_temporal)

print() # Salto de línea estético

# Mostrar la matriz por pantalla
for i in range(n):
    for j in range(n):
        print(tablero[i][j], end=" ")
    print()
````

## P3) Generador de Buscaminas (Revisión de Vecinos)

**Enunciado:**
Haga un programa para generar un tablero del clásico juego Buscaminas. Pide al usuario la dimensión `n` del tablero.
1. Usa un doble `for` para llenar la matriz. En cada casilla genera un número aleatorio entre el `1` y `n`. Si el número es igual a `1`, guarda una mina (`*`). En cualquier otro caso, guarda un espacio seguro (`#`). Imprime este primer tablero por pantalla.
2. Usa un nuevo doble `for` para recorrer el tablero ya creado. Cada vez que encuentres un `#`, revisa a sus 8 vecinos. Cuenta cuántas bombas hay a su alrededor y reemplaza el `#` por ese número.
3. Imprime el tablero final ya resuelto.

**Ejemplo de ejecución:**
```text
Ingrese dimensión del tablero: 4

--- TABLERO CON MINAS ---
# * # # 
# # # # 
* # # * 
# # # # 

--- TABLERO RESUELTO ---
1 * 1 0 
2 2 2 1 
* 1 1 * 
1 1 1 1 
```

````python
import random

n = int(input("Ingrese dimensión del tablero: "))
tablero = []

# 1. CREACIÓN DEL TABLERO CON MINAS
for i in range(n):
    fila = []
    for j in range(n):
        azar = random.randint(1, n)
        if azar == 1:
            fila.append("*")
        else:
            fila.append("#")
    tablero.append(fila)

print("\n--- TABLERO CON MINAS ---")
for i in range(n):
    for j in range(n):
        print(tablero[i][j], end=" ")
    print()

# 2. REVISIÓN DE VECINOS Y SUSTITUCIÓN DE LOS '#'
for i in range(n):
    for j in range(n):
        # Solo evaluamos si es una casilla segura
        if tablero[i][j] == "#":
            bombas_cercanas = 0
            
            # Ciclo de revisión de vecinos (3x3)
            for df in [-1, 0, 1]:
                for dc in [-1, 0, 1]:
                    if df == 0 and dc == 0:
                        continue
                    
                    vecino_i = i + df
                    vecino_j = j + dc
                    
                    # Verificación de límites del tablero
                    if vecino_i >= 0 and vecino_i < n and vecino_j >= 0 and vecino_j < n:
                        # Si el vecino válido es una mina, sumamos 1 al contador
                        if tablero[vecino_i][vecino_j] == "*":
                            bombas_cercanas += 1
            
            # Reemplazamos el '#' por la cantidad encontrada (convertida a texto para mantener uniformidad)
            tablero[i][j] = str(bombas_cercanas)

print("\n--- TABLERO RESUELTO ---")
for i in range(n):
    for j in range(n):
        print(tablero[i][j], end=" ")
    print()
````
