## 1. Enunciado y Reglas del Juego

El flujo del programa debe seguir las siguientes mecánicas:

1. **Dos Tableros:** El juego necesita generar internamente un `tablero_logico` (que tiene las bombas y los números calculados, pero está escondido) y un `tablero_visible` (lleno de letras `'X'`, que es el que se le imprime al usuario en cada turno).
2. **El Ciclo de Turnos:** Utilizando un ciclo `while`, el juego mostrará el tablero visible y le preguntará repetidamente al jugador qué coordenada (fila y columna) desea descubrir.
3. **Pisar una Mina:** Si en la coordenada elegida del `tablero_logico` hay una mina (`#`), el jugador pierde inmediatamente. El ciclo se rompe, se imprime *"¡BOOM! Has pisado una mina"* y se muestra el `tablero_logico` completo para revelar dónde estaban todas las bombas.
4. **Casilla Segura:** Si la coordenada elegida es un número, tomas ese número del `tablero_logico` y lo copias en la misma posición del `tablero_visible`. Luego el turno finaliza y el juego continúa.
5. **Excepción de la Regla del Cero:** Si el jugador descubre un `0` (una casilla sin bombas alrededor), simplemente debes mostrar ese `0` en el `tablero_visible` y el juego continúa con normalidad. *(Nota: En el Buscaminas original de Windows, encontrar un cero destapa en cadena a todos los vecinos vacíos. Para este ejercicio **no es necesario hacerlo**, basta con descubrir solo el 0 para mantener la lógica simple).*

## 4. Instrucciones y Tu Código Final

Deberás programar estrictamente 3 funciones:
1. `crear_tablero(n, cantidad_bombas)`: Contiene la lógica del código base para rellenar la matriz de 0, colocar las minas y contar los vecinos. Debe hacer un `return` del tablero lógico.
2. `mostrar_tablero(matriz)`: Recibe cualquier matriz por parámetro y la imprime en formato cuadrícula. Útil para mostrar tanto el mapa de las 'X' como el resuelto al final.
3. `crear_juego(n, cantidad_bombas)`: Será la función que inicie y controle el juego. Adentro creará la matriz de 'X' (el `tablero_visible`), llamará a la lógica de `crear_tablero`, y ejecutará el ciclo `while` para interactuar con el usuario.
