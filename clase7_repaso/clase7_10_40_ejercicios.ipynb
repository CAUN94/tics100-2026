### El Segundo Al Mando
**Enunciado:** Saber no solo el oro, sino la medalla de plata de una serie.
#### El Segundo Al Mando (simple)
Tienes 3 variables y determinas con lógica manual quién queda al medio.

```python
p1 = 10
p2 = 50
p3 = 30
# Comparacion estatica a mano (simplificada)
if p2 > p3 and p2 > p1 and p3 > p1:
    print("El de en medio es", p3)
```

#### El Segundo Al Mando (con ciclos)
Mantén 2 variables de puntaje en `0`. Recorre la lista con un `for`. Si es mayor al líder (max1), baja el líder a max2 y súbete a max1. Si solo es mayor a max2 (y menor a max1), actualiza max2.

```python
notas = [40, 90, 70, 85]
max1 = 0
max2 = 0
for n in notas:
    if n > max1:
        # Desplazamos jerarquias
        max2 = max1
        max1 = n
    else:
        if n > max2:
            # Supera al segundo pero no al jefe
            max2 = n
print("Primer puesto:", max1)
print("Segundo puesto:", max2)
```

### El Número Primo
**Enunciado:** Identificar la matemática de un primo (Divisible solo por 1 y si mismo).
#### El Número Primo (simple)
Descartar con un if de regla dura simple que no contenga divisibilidad por 2 o 3.

```python
num = 7
if num % 2 != 0 and num % 3 != 0:
    print("Probablemente sea primo o no divisible corto")
```

#### El Número Primo (con ciclos)
Construye bandera `es_primo = True`. `for` va desde `2` hasta el número evaluado (sin tocar los extremos legales). Si hay división 0, bajas la bandera a `False`.

```python
evaluado = int(input("Averigüe si es primo: "))
es_primo = True
# Excluimos 1 y su mismo numero
for i in range(2, evaluado):
    if evaluado % i == 0:
        es_primo = False
        
# Revision de estado final al salir del ciclo
if es_primo:
    print(evaluado, "es un primo de verdad")
else:
    print(evaluado, "NO clasifica como primo")
```

#### Límite de Intentos (con ciclos)
Traba `while intento != clave and fallas < 3`. Suma la falla por error.

```python
llave = "admin"
ingreso = ""
fallas = 0
# Debes fallar y tener margen para quedar en el loop
while ingreso != llave and fallas < 3:
    ingreso = input("Su Clave de Banco: ")
    if ingreso != llave:
        fallas = fallas + 1
        print("Te equivocaste. Van", fallas, "errores de 3 posibles")
# Auditoria de porque se rompio el ciclo
if ingreso == llave:
    print("Retire su dinero")
else:
    print("TARJETA Y CUENTA CONGELADA.")
```

### La hucha de ahorros
**Enunciado:** Continuar trabajando un fondo monetario hasta superar la cuota meta.
#### La hucha de ahorros (simple)
Tengo 500 y meta 1000. Regla dura simple.

```python
mi_plata = 500
if mi_plata > 1000:
    print("Llegue al objetivo")
else:
    print("Sigo siendo pobre")
```

#### La hucha de ahorros (con ciclos)
Pide fondos `while` el acumulador no sobrepase esa pared de meta predefinida.

```python
meta = 5000
chanchito = 0
while chanchito < meta:
    print("Actualmente posees", chanchito, "/", meta)
    moneda = int(input("Ingresar billetes o deposito: "))
    chanchito = chanchito + moneda
print("Meta de vacaciones destruida, ya posees", chanchito)
```
