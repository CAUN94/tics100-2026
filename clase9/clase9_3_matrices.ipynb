# Listas dentro de Listas: Entendiendo las Matrices

## 1. Repaso: Listas Simples, Acceso y Límites

Hasta ahora, has visto las listas como un casillero horizontal de una sola dimensión. Una lista nos permite guardar múltiples datos ordenados.

Para acceder a un valor dentro de una lista, utilizamos su posición (o índice), recordando siempre que en programación empezamos a contar desde el cero (`0`).

**Ejemplo 1: Accediendo a datos a través de una variable**

```python
lista = [1, 2, 3]
print(lista[0])  # Muestra el 1
print(lista[1])  # Muestra el 2
print(lista[2])  # Muestra el 3
```

**Ejemplo 2: Accediendo directamente a la lista (Literal)**

Para el computador, la variable y el dato crudo son lo mismo. Podemos hacer exactamente lo mismo sin usar una variable, aplicando los corchetes directamente sobre la lista escrita:

```python
print([1, 2, 3][0])  # Muestra el 1
print([1, 2, 3][1])  # Muestra el 2
print([1, 2, 3][2])  # Muestra el 3
```

**Modificando un dato en la lista**

Al igual que podemos leer un dato, podemos sobreescribirlo usando su posición y el operador de asignación (`=`):

```python
lista = [1, 2, 3]
lista[1] = 99 
print(lista)  # El resultado será: [1, 99, 3]
```

**¿Qué pasa si quedo fuera de los márgenes?**

Las listas tienen límites estrictos dictados por su tamaño. Si intentas acceder a una posición que no existe, el computador chocará con un muro invisible y tu programa se caerá, arrojando un error.

```python
lista = [1, 2, 3]

# Si descomentas la siguiente línea, verás el error:
# print(lista[5]) 

# ¡ERROR! IndexError: list index out of range
# (Error de Índice: índice de lista fuera de rango)
```

Por esta razón, debemos ser muy cuidadosos de no pedirle a Python que lea o modifique casillas que no existen.

## 2. La creación de las Matrices

Python es tan flexible que dentro de una lista no estás obligado a guardar solo números o textos. **Puedes guardar otras listas completas**.

Cuando haces esto, la estructura deja de ser una simple línea horizontal y se convierte en una cuadrícula plana, como una tabla de Excel o un tablero de ajedrez. A esto lo llamamos **Matriz**.

Al guardar listas dentro de listas, generamos automáticamente dos dimensiones:

* **La Fila (Horizontal):** Cada lista interna es una fila.
* **La Columna (Vertical):** Son las posiciones de los elementos dentro de cada lista interna.

Imagina este código, donde creamos una matriz de 4x4 (4 filas y 4 columnas). Hemos elegido estos números a propósito para que el primer dígito represente la fila y el segundo dígito la columna:

```python
matriz = [ [ 11, 12, 13, 14 ], [ 21, 22, 23, 24 ], [ 31, 32, 33, 34 ], [ 41, 42, 43, 44 ] ]
# Esto es una matriz de 4 filas y 4 columnas, pero también se puede escribir de una forma más clara: 
matriz = [
    [11, 12, 13, 14], # Esto es la Fila 0
    [21, 22, 23, 24], # Esto es la Fila 1
    [31, 32, 33, 34], # Esto es la Fila 2
    [41, 42, 43, 44]  # Esto es la Fila 3
]

# Mientras que la primera forma es correcta, la segunda forma es más fácil de leer y entender, especialmente para matrices grandes.
```

## 3. El uso de `i` y `j` (Las coordenadas)

En matemáticas y en programación, existe un estándar universal para no confundirnos cuando hablamos de dos dimensiones: usamos las letras **`i`** y **`j`**.

Es muy importante entender cómo se mueven estas letras, ya que **funcionan distinto a un plano cartesiano (ejes X e Y)** que sueles usar en geometría:

* La **`i`** siempre representa a la **Fila**. A diferencia del eje Y (que se lee hacia arriba), en programación la coordenada `i` **se lee en dirección hacia abajo**. Es decir, la fila 0 está en la parte superior, la fila 1 está debajo, y así sucesivamente.
* La **`j`** siempre representa a la **Columna**. Esta sí se comporta como el eje X, ya que **se lee hacia el lado** (de izquierda a derecha).

Cuando quieras ubicar un dato, siempre pensarás en las coordenadas `[i][j]` (primero la fila hacia abajo, luego la columna hacia el lado). ¡Es exactamente igual que jugar a la Batalla Naval!

## 4. Accediendo y Modificando datos en una Matriz

Aplicaremos exactamente la misma lógica que aprendimos con las listas simples al principio de la guía, pero ahora utilizando dos coordenadas en lugar de una.

### Ejemplo 1: Accediendo a datos por partes

Vamos a ir paso a paso para ver cómo el computador entra a la matriz.

```python
matriz = [
    [11, 12, 13, 14], 
    [21, 22, 23, 24], 
    [31, 32, 33, 34], 
    [41, 42, 43, 44]  
]

# PASO 1: Usar solo la coordenada de la fila (i)
# Si pedimos la fila 1, el computador nos entrega la lista interna completa
print(matriz[1])  
# Muestra: [21, 22, 23, 24]

# PASO 2: Usar ambas coordenadas (i, j)
# Ahora, dentro de esa fila 1, queremos el dato que está en la columna 2.
# Le agregamos el segundo corchete al lado.
print(matriz[1][2])  
# Muestra: 23
```

### Ejemplo 2: Accediendo directamente a la matriz (Literal)

Nuevamente, podemos hacer esto sin usar variables, escribiendo la matriz cruda en el código y pidiendo las coordenadas al final de la estructura:

```python
print( [ [11, 12, 13, 14], [21, 22, 23, 24], [31, 32, 33, 34], [41, 42, 43, 44] ][1] )
# Muestra la fila 1: [21, 22, 23, 24]

print( [ [11, 12, 13, 14], [21, 22, 23, 24], [31, 32, 33, 34], [41, 42, 43, 44] ][1][2] )
# Muestra el 23
```

### Ejemplo 3: Modificando un dato en la Matriz

Al igual que en las listas simples, si queremos cambiar el valor de una coordenada específica, indicamos su posición `[i][j]` y le asignamos un nuevo valor:

```python
matriz = [
    [11, 12, 13, 14], 
    [21, 22, 23, 24], 
    [31, 32, 33, 34], 
    [41, 42, 43, 44]  
]

# Cambiaremos el número 14 por un 99. 
# El 14 está en la fila 0, columna 3.

matriz[0][3] = 99 

# Si imprimimos la primera fila, veremos el cambio:
print(matriz[0]) 
# El resultado será: [11, 12, 13, 99]

# Si imprimimos el valor específico que cambiamos, también veremos el cambio:
print(matriz[0][3])
# El resultado será: 99
```

## 5. ¿Cómo mostrar los valores de una Columna?

Normalmente, al hacer `print(matriz[0])` vemos una **Fila** completa, porque la coordenada `i` se mantiene fija.

Para mostrar una **Columna** (leer de arriba hacia abajo), debemos hacer exactamente lo inverso: **mantener fija la coordenada de la columna (`j`) e ir cambiando manualmente la coordenada de la fila (`i`)**.

Por ejemplo, si queremos ver todos los datos de la **Columna 2** (que contiene los números terminados en 3):

```python
# Mantenemos el segundo corchete en [2], y avanzamos el primero hacia abajo:

print(matriz[0][2])  # Fila 0, Columna 2 -> Muestra el 13
print(matriz[1][2])  # Fila 1, Columna 2 -> Muestra el 23
print(matriz[2][2])  # Fila 2, Columna 2 -> Muestra el 33
print(matriz[3][2])  # Fila 3, Columna 2 -> Muestra el 43
```

Esta lógica de dejar fijo un índice y mover el otro es la base fundamental que usaremos más adelante cuando incorporemos ciclos (`for` y `while`) para automatizar el recorrido de estas estructuras.
